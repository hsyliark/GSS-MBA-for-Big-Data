1000개정도의 평행세계에서 결과를 관측하고 평균을 내보자. 

````r
doctor<-function(){
    x<-c()
    x[1]<-1
    for(t in 1:9999){
        if(x[t]==0) x[t+1]<-(runif(1)<3/7)
        else x[t+1]<-(runif(1)<1/3)
    }
    x
}
````

````r
X<-rep(0,1000*10000)
dim(X)<-c(10000,1000)
````

````r
for(omega in 1:1000){
    X[,omega]<-doctor()
}
````

````r
apply(X,1,mean)
````

대충 0.393 정도의 확률로 10000일 뒤에 비가 온다고 볼 수 있다.

`–` 풀이1의 결과와 풀이4의 결과가 일치한다는 것은 ensemble mean 과 time average가 일치한다는 것을 의미한다. 

`-` 이는 큰수의 법칙의 확장판이다. 

`-` 큰수의 법칙은 아래와 같다. 

$X_1,X_2,\dots,X_{9999}$이 $i.i.d.$일때 

$\frac{1}{9999}\sum_{i=1}^{9999}X_i\approx E(X_{10000})=E(X_{9999})=E(X_{9998})=\dots=E(X_1)$

우리는 풀이4를 적용하여서 적절한 답을 구했는데 이는 큰수의 법칙이 적용된 것으로 해석해도 된다. 

그런데 우리의 경우는 $i.i.d$ 가 아니다. (일단 확실히 독립은 아니다.)

따라서 큰수의 법칙이 원래는 성립하지 않아야 한다. (그런데 한다?)

`-` 결국에는 $X_1,X_2,\dots$이 서로 독립이 아니어도 아래의 식이 만족하는 특별한 열(array)이 있다. 

$\frac{1}{T}\sum_{t=1}^{T} X_t \approx E(X_T)$

이러한 특별한 열에는 어떠한 것들이 있는지 관심이 있다. 

이 식은 결국 time average(좌변)와 ensemble mean(우변)이 같다는 것을 의미하기 떄문이다. 

참고로 iid 버전에서는 sample mean이 time average 에 대응한다. 

`-` iid가 아닌 열중에서 time average $\to$ ensemble mean 이 되는 열들은 매우 많다.

우리는 그중에서 마코프체인이라는 특별한 열에만 관심을 가진다. 

`-` 마코프체인은 아래의 성질을 만족하는 확률변수열 $\{X_t\}$를 의미한다. 

$P(X_{t+1}|X_t,X_{t-1},X_{t-2},\dots,X_1)=P(X_{t+1}|X_t)$

수식이 어렵게 되어있는데 아래의 느낌을 기억하면 편리하다. 

```
t+1 시점에 비가 올지 안올지 알고 싶다면, 단지 직전 시점(t시점)에 비가 왔는지 안왔는지에 대한 정보만 있으면 된다. 그저께, 그그저께 등에서 비가 왔는지 안왔는지에 대한 정보는 알고 있을 필요가 없다.  
```

`-` 우리의 날씨예제는 마코프체인이다.
