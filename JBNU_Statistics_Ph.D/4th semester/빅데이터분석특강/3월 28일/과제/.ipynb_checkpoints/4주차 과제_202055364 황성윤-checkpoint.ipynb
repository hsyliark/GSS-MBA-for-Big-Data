$y=(x-1)^2$를 최소화 하는 $x$를 경사하강법을 이용하여 찾아라. 수렴과정을 animation으로 시각화하라.  
- x의 초기값은 -3으로 설정한다. 
- 적당한 $\alpha$를 골라서 100번의 반복안에 수렴하도록 하라. 

```python
import tensorflow as tf 
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
import tensorflow.experimental.numpy as tnp 
tnp.experimental_enable_numpy_behavior() 
```

```python
x = tf.Variable(3.0)
```

```python
with tf.GradientTape() as tape:
    y=(x-1)**2
```

```python
tape.gradient(y,x)
```

- 만약 $x$의 값을 $0.01$만큼 왼쪽으로 움직이고 싶다면 learning rate $\alpha$를 다음과 같이 두면 된다. 하지만 수렴하기까지 굉장히 많은 반복횟수가 필요할것으로 보이기 때문에 차후에 증가를 시켜야 할 것 같다.

```python
alpha=0.01/4
```

- 이러한 점을 고려하여 알고리즘을 만들면 다음과 같다.

```python
x=tf.Variable(3.0)
alpha=0.01/4
```

```python
x_lst = []
y_lst = []
```

```python
x_lst.append(x.numpy())
y_lst.append((x.numpy()-1)**2)
```

```python
with tf.GradientTape(persistent=True) as tape:
    tape.watch(x)
    y=(x-1)**2
```

```python
x.assign_sub(tape.gradient(y,x)*alpha)
```

```python
x_lst.append(x.numpy())
y_lst.append((x.numpy()-1)**2)
```

```python
x_lst, y_lst
```

- for

```python
x=tf.Variable(3.0)
alpha=0.025
x_lst = []
y_lst = []
x_lst.append(x.numpy())
y_lst.append((x.numpy()-1)**2)
for k in range(100):
    with tf.GradientTape(persistent=True) as tape:
        tape.watch(x)
        y=(x-1)**2
    x.assign_sub(tape.gradient(y,x)*alpha)
    x_lst.append(x.numpy())
    y_lst.append((x.numpy()-1)**2)
```

```python
_x = np.linspace(-4,6,100)
```

```python
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(_x,(_x-1)**2)
pnts, = ax.plot(x_lst[0],y_lst[0],'or')
```

```python
plt.rcParams["animation.html"]="jshtml"
from matplotlib import animation
```

```python
def animate(i):
    pnts.set_xdata(x_lst[:(i+1)])
    pnts.set_ydata(y_lst[:(i+1)])
```

```python
ani = animation.FuncAnimation(fig,animate,frames=100) 
ani
```
