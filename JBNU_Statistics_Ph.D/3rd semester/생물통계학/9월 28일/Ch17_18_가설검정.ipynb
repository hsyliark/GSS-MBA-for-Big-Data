### Chapter 17. 가설검정

##### **가설검정의 일반적인 절차**

1. 관심연구에 관한 귀무가설(null hypothesis, H0)과 대립가설(alternative hypothesis, H1)을 설정  

2. 표본으로부터 적절한 자료 수집

3. 귀무가설이 사실인 경우 해당하는 검정통계량(test statistic)값 계산

4. 검정통계량 값을 이미 알려진 확률분포의 값들과 비교해서 p-value 계산

5. p-value 해석 및 결론 


##### **귀무가설과 대립가설 설정**

- 예제 : 모집단 내 남녀간 흡연율 비교
> 귀무가설(H0) :  모집단 흡연율은 남녀간에 서로 동일하다 $\to$ 모집단에서 효과가 없다고 가정  
> 대립가설(H1) : 모집단 흡연율은 남녀간에 서로 다르다 $\to$ 우리가 주장하고자 하는 내용으로 설정
  
- **양측검정**(two-sided test) :  군 간 차이의 방향성을 명시하지 않는 검정
  (군 간 차이가 어떤 방향을 향하게 될지 예측하는 것은 어렵기 때문에 대부분 **양측검정**을 사용함)   
- **단측검정**(one-sided test) : 효과의 방향성을 미리 지정하는 검정, 동등성, 비열등성 임상시험 등에 사용함

##### **검정통계량 계산**

- **검정통계량** : 표본의 요약통계량 값과 귀무가설에서 설정된 값 간의 차이를 계산한 값, 검정통계량의 절대값이 클수록 귀무가설이 맞지 않다는 증거가 강해짐

|가설검정|검정통계량|  
|---|---|
|Z-검정|Z-통계량|
|t-검정|t-통계량|
|분산분석|F-통계량|
|카이-제곱검정|카이-제곱 통계량|

- p-value : 귀무가설이 사실인 경우에 현재와 같이 혹은 이보다 더 극단적인 결과들이 얻어지게 될 확률, **값이 작을수록 귀무가설에 반하는 증거가 커짐**
- 일반적으로 p-value가 유의수준보다 작으면 귀무가설을 기각하고, 해당 유의수준에서 유의하다는 결론을 도출함
> EX) 유의수준이 5%일 때, P<0.05이면 귀무가설 기각하고 해당 결과는 5% 수준에서 유의하다.  
  P>=0.05이면 귀무가설 채택하고 해당 결과는 5% 수준에서 유의하지 않다.  
- p-value 해석 시 유의사항
> p-value가 작다고 하여 귀무가설이 사실이 아닐 확률이라고 해석할 수 없음  
  자료가 **귀무가설을 기각할 만큼 충분한 증거를 제공하지 않는다**는 의미로 해석함  
  또한, P<0.05로 표기하기 보다 p-value 전체를 제시하는 것이 좋음

##### **비모수적 검정**

- 모수적 검정(parametric test) : 자료가 따르는 **확률분포와 관련된 정보를 사용해 가설검정**을 실시함  
- 비모수적 검정(nonparametric test) = 분포-무관 방법(distribution-free method) = 순위-기반 방법(rank-based method)  
  : 자료가 따르는 확률분포에 대한 가정없이 사용함 $\to$ 모수적 검정을 실시하기 위한 가정을 만족하지 못할 때 사용함  
  : 크기 순으로 나열된 자료에 대해 순서를 매긴 것을 사용하는 방법  
  : 표본 수가 작을 때, 자료가 범주형일 때 효과적임 
  
> **모수적 검정을 실시하기 위한 가정 (chapter.35)**  
      1. 정규분포를 따르는가? (정규성)  
      2. 분산이 동일한가? (동분산성)  
      3. 변수 간 관련성은 선형적인가? (선형성)  
    


##### **신뢰구간**

- 신뢰구간 : 실제 효과의 크기에 대한 예상범위를 제공하여, 결과가 내포한 의미를 임상/보건학적으로 평가할 수 있게 해줌
> EX) 효과에 대해 가정한 값이 `0`이라고 할 때, `0`이 95% 신뢰구간 밖에 존재하면 귀무가설 기각 $\to$ 아마 p-value도 0.05보다 작을 것

##### **동등성시험 및 비열등성시험**

- 동등성 시험(equivalence trial) : 새로운 처리가 기존 처리에 비해 **적어도 그 효능이 같다**는 것을 보이는 시험  
- 비열등성 시험(non-inferiority trial) : 새로운 처리가 기존 처리에 비해 **실질적으로 더 나쁘지않다**는 것을 보이는 시험
- 생동등성 시험(bioequivalence trial) : **동일한 용량**을 투여했을 때 새로운 약의 제형이 인체 내흡수율, 흡수량 등에 있어서 기존 약의 제형과 **차이가 없는지** 보기 위한 특별한 형태의 **무작위 임상시험**  

> **사용 예시**  
    1. 기존 약제에 비해 부작용이 적은지 확인  
    2. 일회 복용 알약 수가 적어서 환자의 복용 부담 및 비용 감소 확인  
    3. 이외 우월함을 증명하기 보다 상대적인 장점이 있다는 것을 보이고자 할 경우

- 동등성 시험의 가설검정  

  - 귀무가설(H0) : 두 처리의 효과에 차이가 있다.  
  - 대립가설(H1) : 두 처리의 효과가 동일하다.  $\to$ 기존 가설검정의 귀무가설/대립가설과 반대되게 설정함  


  1) 관심효과(ex. 두 처리집단 간 평균 차이)에 대한 신뢰구간 계산
  2) 신뢰구간이 미리 설정된 **동등성 범위**(equivalence range)에 `전체 포함`, `부분 포함`, `전혀 포함되지 않음` 인지 확인  
  3) **신뢰구간 전체가 동등성 범위에 포함**된다면 두 처리는 동등한 것으로 판단 $\to$ **귀무가설 기각**

- 비열등성 시험의 가설검정  

  - 귀무가설(H0) : 새로운 처리가 기존 처리에 비해 열등하다.
  - 대립가설(H1) :새로운 처리가 기존 처리에 비해 비열등하다.    


  1) 관심효과(ex. 두 처리집단 간 평균 차이)에 대한 신뢰구간 계산
  2) `신뢰구간의 하한` > `비열등성 한계`(non-inferiority margin) 이면, 새로운 처리가 기존 처리에 비해 비열등한 것으로 판단 $\to$ **귀무가설 기각**

### Chapter 18. 가설검정에서의 오류

##### **의사결정의 오류**

- 유방암 환자의 5년 사망률을 낮추기 위한 두 처리의 효과를 비교해보자
  - 관심효과 : 5년 사망률 차이
  - 귀무가설 : 두 처리집단 간 유방암 5년 사망률의 차이가 없다.  
  - 대립가설 : 두 처리집단 간 유방암 5년 사망률의 차이가 있다.

- 이 때 우리는 `귀무가설 기각`, `귀무가설 채택` 2가지 의사결정을 하게 됨

- 우리는 표본이라는 제한된 정보를 사용하기 때문에, 귀무가설에 대해 어떤 의사결정을 하던 잘못된 결정을 내릴 가능성이 항상 존재함

- 제1종 오류(type 1 error)
  - 귀무가설이 참인데 기각하는 경우 
  - 제1종 오류를 범할 확률은 $\alpha$(알파)로 표기하고, 해당 검정의 유의수준이라 함

- 제2종 오류(type 2 error)
  - 귀무가설이 거짓인데 채택하는 경우 
  - 제2종 오류를 범할 확률은 $\beta$(베타)로 표기함
  - $1-\beta$ : 검정력(귀무가설이 거짓일 때 이를 기각하게 될 확률)  
  $\to$ 제2종 오류를 범할 확률은 항상 존재하기 때문에 검정력은 100%가 될 수 없음

|구분|H0 기각|H0 채택|
|---|---|---|
|H0 참|제1종 오류|오류 없음|
|H0 거짓|오류 없음|제2종 오류|

##### **검정력 및 그 크기를 결정하는 요인들**

- 표본 수 (sample size) : 표본 수가 커질수록 검정력 증가함

- 유의수준 (significance level) : 유의수준이 클수록 검정력 증가함 $\to$ $\beta$가 작아질 수록 $\alpha$가 커짐

##### **다중 가설 검정**

- 다중검정의 문제 : 한 개의 데이터셋에 여러 번 유의성 검정을 실시하는 경우, **비교횟수 증가에 따라 제1종 오류율이 커짐**

- 다중 비교 실시 예시


  1) 부집단 분석 (subgroup analysis)
     - 부집단 내에서 처리들을 비교하면 분석에 사용되는 표본 수가 작아져 검정력 감소함 $\to$ 통계적으로 유의한 처리 효과를 발견하지 못할 수 있음  
     - 연구설계 단계에서 생물학적/임상적/보건학적 논리에 근거해 부집단을 정의하지 않고 분석된 결과에 의해 결정함 $\to$ 가짜 결과들이 얻어질 수 있음
     - 부집단 내 군 간 연구대상자들의 특성은 불균형할 수 있으며 이로 인해 편향이 발생할 수 있음
  
  
  2) 하나의 결과변수에 대한 다중비교 (multiple comparisons for a single outcome variable)
     - 3개 이상의 처리군들을 비교하고자 하는 경우  
     - 여러 시점에서 측정되는 반응변수에 대해 시점들 간 비교를 하는 경우
  
  
  3) 다중 결과변수 (multiple variable)
     - 여러 개의 종료점들을 사용해 처리 효과를 평가하고자 하는 경우
  
  
  4) 중간 분석(interim analyses)
     - 연구 종료 시점 외 미리 지정된 중간 시점들에서 처리를 비교하는 경우
  
  
  5) 자료 낚기 (data fishing)
      - 유의한 관련성이 얻어질 때까지 비교를 계속하는 것  
  

- 해결방안

  최고의 방법 : 연구의 목적과 관련이 있고, 연구설계 단계에서 명시된, **적은 횟수의 통계 검정만 실시하기**
  
  1. 각 검정에서 얻어지는 p-value 보정하기 (*본페로니 접근법(Bonferroni's approach) : 각 p-value에 수행된 검정 횟수를 곱해주는 방법*)
  
  2. 각 검정에 대해 보다 엄격한 유의수준 적용하기  
  
  3. 부집단을 정의하는 요인과 처리군들 간의 상호작용 효과에 대한 검정 결과가 유의한 경우에만 부집단 분석 실시하기 $\to$ 연구계획서에서 계획된 부집단만 분석하기  
  
  4. 전반적인 처리효과가 유의한 경우에만 처리들 간 짝진 비교(paired-comparison) 실시 후, 얻어진 p-value들을 다중비교방법을 사용해 사후적으로 보정하기  
  
  5. 반복 측정이 이루어진 경우 반복측정된 자료들이 서로 상관되어 있다는 점을 반영하는 분석방법 사용하기  
  
  6. 결과변수(종속변수)가 여러 개인 경우, 이들을 적절히 결합하여 하나의 복합사건 종료점으로 변환하기, 혹은 다변량 분석을 실시하기  
 
  7. 중간분석을 실시하고자 하는 경우, 전체 분석들에 대한 전반적인 유의수준이 0.05로 유지되도록 연구 설계하기 $\to$ 중산분석 시점에서는 낮은 유의수준을 사용하기
