# Part 2 표본추출과 추정

## 10 표본추출 및 표본추출 분포

### 왜 표본을 뽑는가?

- 모집단(population): 연구 관심대상 전체 집단, 전수조사
- 표본(sample): 모집단을 대표(representative)할 것으로 믿어지는 모집단에서 추출된 집단
- 추론(inference): 추출된 표본자료를 사용해 모집단에 대한 결론을 내림
- 표본추출오차(sampling error): 표본의 정보가 모집단을 완전히 대표하지 못할 수 있음

Sampling error를 간단히 **'표본오차'** 라 하기도 하는데, 표준 오차(standard error)와는 다른 개념임

### 대표성이 확보된 표본을 뽑는 방법

- 무작위표본(확률표본, random sample): 표본이 모집단에 대한 대표성을 확보하기 위한 이상적 방법
 1. 무작위추출(random sampling): 모집단에 있는 모든 사람들에 대한 목록[표본추출틀(sampling frame)]을 만든 다음, 이 목록에서 무작위로 추출하는 방법이다.
 2. 편의추출(convenience sampling): 무작위추출의 비용 문제 등을 해결하기 위한 추출방법이며 예를 들어 먼저 한 병원을 선택한 다음, 그 병원에서 해당 상태를 가진 환자들 전체 혹은 일부분을 조사함.
- 비-무작위적인 방법
 1. 할당추출(quota sampling)
 2. 계통추출(systematic sampling)
- 그 외 추출법: https://www.outsourcing.co.kr/news/articleView.html?idxno=87201
 
무작위(randomly)로 추출한다는 의미는 모집단내 같은 크기의 모든 가능한 표본들이 동일한 확률로 선택되도록 추출한다는 의미

### 점추정값(point estimate)과 구간추정값(interval estimate)
- 모수(parameter): 일반적으로 그리스 문자로 표기. ex) 모평균-$\mu$, 모표준편차-$\sigma$

예시의 추정값은 단일 값으로 표현되기 때문에 모수에 대한 **점추정값** 이다. 구간 형태의 값들은 **구간추정값**이라 함.

모수는 표본에 수집된 자료를 사용해 추정함. 이 추정값을 표본 통계량(sample statistic) 혹은 **통계량(statistic)** 이라 함.

### 표본추출의 변동(variation)

모집단으로부터 동일 크기의 표본을 여러 번 반복적으로 추출 할 때 모수에 관한 추정값은 일치하지 않으나 비슷한 값들이고 실제 모수에 가까운 값일 것이다. 이로부터 추정값의 정밀성과 표본추출로 인한 오차를 평가할 수 있으나, 현실적으로 표본을 딱 한번 추출하게 된다. 하지만, **표본 추정값의 이론적 분포에 관한 지식을 활용한다면 모수에 관한 추론이 가능하게 된다.** 

### (얻어진) 표본평균의 표본추출 분포(sampling distribution of the mean)

모평균을 추정하기 위해 다음과 같이 작업을 한다고 가정해보자
- 모집단에서 크기가 n인 표본을 반복추출 
- 각각의 표본으로부터 표본평균 계산: 모평균에 대한 추정값
- 모평균에 대한 추정값들의 히스토그램: 표본평균들의 분포 -> **(얻어진) 표본평균의 표본추출 분포**

표본평균의 표본추출 분포의 특징
- 표본 수가 충분히 크다면 모집단의 분포가 어떤 형태이든 간에 **정규분포**임. [중심극한정리(Central Limit Theorem)]
- 표본 수가 작은 경우, 모집단 자료가 정규분포일 때만 정규분포를 따름
- 불편추정값(unbiased estimate): 표본평균들의 평균 = 모평균
- 표본평균의 표준오차(standard error of the mean; SEM): 표본추출 분포의 변동 크기는 표본평균들의 표준편차로 측정

$$SEM = \frac{\sigma}{\sqrt{n}}$$

현실적으로 표본을 여러번 추출하지 않고 한번만 추출하게 될 것이며, 이러한 경우에 추출된 표본으로부터 얻어진 표본 평균이 점추정값으로 가장 좋다. 또한 모표준편차($\sigma$)는 표본표준편차($s$)를 사용해 추정하는 것이 가장 적절함

$$\hat{SEM} = \frac{s}{\sqrt{n}}$$

### 표준오차들의 해석

표본으로부터 계산되는 모든 종류의 통계량들은 표준오차가 있다.
- 표준오차가 크다: 추정값이 정밀하지 않다
- 표준오차가 작다: 추정값이 정밀하다.

표본오차를 줄이는 방법
- 표본의 크기($n$)을 증가
- 변동량($s$)이 적은 자료들을 얻는다

### 표준편차(SD)와 표본평균의 표준오차(SEM) 중 어떤 값을 사용할 것인가?

- 표준편차(SD): 자료의 변동량을 알려주는 척도
- 표준오차(SE): 통계량의 정밀성을 알려주는 척도
- 표본평균의 표준오차(SEM): 표본평균의 신뢰도의 척도

### (얻어진) 비율의 표본추출 분포(sampling distribution of the propotion)

모집단에서 $n$명의 표본, 특성을 가진 사람 수 $r$명, 모비율 $\pi$에 가장 좋은 점추정값: $p$

$$p = \frac{r}{n}$$

모집단에서 크기가 $n$인 표본을 반복적으로 추출해서 얻어지는 **표본비율($p$)들의 표본추출 분포는**는 평균이 $\pi$인 정규분포에 근사하게 됨.

- 표본비율의 표준오차(standard error of the proportion): 표본비율들의 표준편차

$$SE(p) = \sqrt{\frac{\pi(1-\pi)}{n}}$$

- 모비율을 알고 있는 경우는 드물며, 표본비율의 표준오차의 추정값을 사용함

$$\hat{SE(p)} = \sqrt{\frac{p(1-p)}{n}}$$

- 값 $\hat{SE(p)}$는 모비율 $\pi$에 대한 추정값인 $p$의 정밀성에 관한 척도이며, 이 값이 작으면 작을수록 얻어진 표본비율은 모비율을 추정하기 위한 척도로 사용되기에 더 정밀한 추정값이란 의미임.

## 11 신뢰구간

- 정밀성 척도는 **구간추정**을 하는데 매우 효과적임.
- 신뢰구간(confidence interval; CI): 점추정값을 중심으로 좌우로 표준오차의 몇 배수만큼 떨어진 값들을 계산한 뒤, 이를 구간 형태로 표현
- 신뢰한계(confidence limits): 계산된 좌우 두 개의 구간 끝 값

### 모평균에 대한 신뢰구간(confidence interval for the population mean)

- 정규분포를 이용하는 방법
 1. 모집단을 반복 추출한 표본의 평균을 계산하고 표본 수가 크다면, 표본평균들은 정규분포를 따름
 2. 모평균에 대한 신뢰구간을 구하기 위해 표본평균을 사용하는 경우 정규분포의 특성을 이용
 3. 표본평균들 중 95%는 모평균을 중심으로 좌우 1.96배의 표준편차 내에 존재하고 이 표본평균들의 표준편차를 표본평균의 표준오차(SEM)이라 함
 4. $SEM = \sigma/\sqrt{n}$와 같고, 따라서 **모평균에 대한 95% 신뢰구간**은 다음과 같이 계산됨

$$(\bar{x} - 1.96 \times SEM,~ \bar{x} + 1.96 \times SEM)$$

일반적으로 모표준편차는 알 수 없으므로 근사값을 계산하면

$$(\bar{x} - 1.96 \times \frac{s}{\sqrt{n}},~ \bar{x} + 1.96 \times \frac{s}{\sqrt{n}})$$

- t-분포를 이용하는 방법
 1. 신뢰구간을 구할 때 모분산 $\sigma^2$을 알고 있는 경우에만 정규분포를 사용.
 2. 모분산을 모른다면, 표본 수가 충분히 큰 경우에 한해 중심극한정리를 적용해 정규분포를 구함.
 3. 만일, 모분산도 모르고 자료가 정규분포를 하지도 않는다면 (표본 수가 적다면) $\sigma^2$에 대한 추정값으로 $s^2$을 사용, 해당 경우 표본평균은 $t$-분포에 따르게 되며, 따라서 모평균에 대한 95% 신뢰구간은 다음과 같이 계산됨
 4. $t$-분포를 이용한 신뢰구간은 정규분포에 기초한 신뢰구간보다 약간 넓으나 표본 수가 커지면 정규분포와 차이가 없게 된다.
 
$$(\bar{x} - t_{0.05} \times SEM,~ \bar{x} + t_{0.05} \times SEM)$$

즉, $(\bar{x} \pm t_{0.05} \times \frac{s}{\sqrt{n}})$로 추정. $t_{0.05}$는 자유도가 $n-1$인 $t$-분포에서 양쪽 꼬리확률이 0.05인 경우 해당하는 백분위수(percentile)임.

### 모비율에 대한 신뢰구간(confidence interval for the population pro-portion)

- 표본비율의 표본추출 분포: 이항분포
- 표본 수 $n$이 충분히 크다면 표본비율의 표본추출 분포(sampling distribution of the proportion)는 근사적으로 평균이 $\pi$인 정규분포에 따르게 됨.
- 모비율 $\pi$를 추정하기 위해 표본비율 $p = r/n$을 사용한다면 해당 표본비율의 표준오차 추정값은 $\sqrt{\frac{p(1-p)}{n}}$
- 모비율에 대한 95% 신뢰구간

$$(p - 1.96 \times \sqrt{\frac{p(1-p}{n}},~ p + 1.96 \times \sqrt{\frac{p(1-p}{n}})$$

### 신뢰구간의 해석

- 신뢰구간이 얼마나 넓은가? 
 1. 신뢰구간이 넓다는 것은 점추정값이 정밀한 값이 아니라는 의미, 반대로 좁다는 것은 정밀하다는 의미
 2. 너비를 결정하는 것은 표준오차의 크기, 따라서 표본 수에 의해 영향
 3. 연속형 변수의 경우 자료의 변동량도 영향, 자료의 변동이 심한 소규모 연구는 작은 변동을 가진 대규모 연구에 비해 상대적으로 넓음
- 신뢰구간으로부터 어떤 임상적/보건학적 의미를 도출할 수 있는가? 신뢰구간이 임상적/보건학적으로 중요한 의미를 가지고 있는지를 평가하기 위해서는 해당 신뢰구간의 상한 및 하한을 사용한다.
![](img2.png)
- 신뢰구간이 특별히 관심있는 값을 포함하고 있는가?
 1. 모수가 어떤 값일 것이라는 가설을 평가하기 위해서는 해당 값이 신뢰구간 내에 포함되는지를 검토
 2. 만일 신뢰구간이 이 값을 포함하면 모수는 그 값일 가능성이 높다.
 
### 자유도(degree of freedom, df)

- 자유도: 표본수 - 제약조건의 수
- 제약조건의 수: 추정이 필요한 모수의 수
- 예시1: 합이 T가 되는 수 3개가 있다고 가정하자. 합이 T라는 제약 조건 하에서 이들 3개의 수 중 2개는 어떤 값도 '자유롭게' 취할 수 있지만, 나머지 하나는 제약 조건 때문에 다른 값을 취할 수 없다. 따라서 수의 합이 T라는 제약조건 하에서는 이들 수 3개의 자유도는 2가 됨
- 예시2: 표본분산 $s^2 = \frac {\sum (x-\bar{x})^2} {n-1}$의 자유도는 표본 수 - 1 (즉, $n$ - 1)이다. 왜냐하면 표본분산 $s^2$을 계산하기 위해서는 모평균에 관한 추정값인 표본평균($\bar{x}$)을 미리 계산해야하만 하며, 따라서 표본평균이 이미 계산된 경우 표본분산의 자유도는 자료의 수 -1이기 때문이다.

### 붓스트랩핑(bootstrapping)과 잭나이핑(jackknifing)
- 붓스트랩핑: 추정값에 대한 표본추출 분포를 가정하지 않고, 고도의 컴퓨터 모의실험(simulation)을 통해 모수에 대한 신뢰구간을 유도하는 방법, 원자료 표본을 복원 추출
- 잭나이핑: 원자료 표본을 추출하는 것이 아닌 표본 크기 n인 원자료로부터 관찰값 하나를 제외한 후, 남아있는 n - 1개의 관찰값들을 사용해 관심 모수에 관한 추정값을 계산함

## 12 연구 설계 I

- 연구의 초기 단계에서 그 목적이 명시
- 목적
  1. 모수 추정

  3. 중재의 효과를 평가
- 자원이 충분한지, 윤리적으로 고려해야할 사항에 의해서도 영향을 받음

### 실험연구(experimental study)와 관찰연구(observational study)

- 실험연구
 1. 연구자가 어떤 방식으로든 결과에 영향을 미치기 위해 개입하는 연구
 2. 연구 관심 요인은 아니지만 결과에 영향을 미칠지도 모르는 다른 요인들을 통제할 수 있음
 3. 수행 자체가 불가능할 수 있고, 비윤리적일 수 있음
- 관찰연구
 1. 연구자가 결과에 영향을 미치는 요인들을 통제할 수 없음
 2. 코호트 연구(cohort study)나 사례-대조 연구(case-control study)
 3. 실험연구보다 낮은 수준의 정보가 얻어지나 유일한 대안일 수 밖에 없을 수도 있음
 4. 역학연구(epidemiological study)는 일반 인구 집단을 대상으로 관심 요인과 질병 간의 관계를 평가하는 연구이며 관찰연구에 해당

### 관찰단위(unit of observation)의 결정

- 관찰단위: 분석의 측면에서 서로 독립적이고 개별적인 '개체' 혹은 '개체들'로 이루어진 가장 작은 집단
 1. 임상시험: 관찰 단위는 개별 환자
 2. 치의학 연구: 관찰 단위는 치아가 아닌 구강 (치아끼리 독립적이지 X)
 3. 동물실험 연구: 각 개체(예를 들면 생쥐)로부터 획득 되는 재료(material)들을 한 묶음으로 간주해야 하는 상황이 발생, 각각을 평가하는건 부절절
 4. 집락 (혹은 군집) 무작위배정 임상시험(cluster randomized trial): 개인이 아닌 집단이 관찰 단위
 5. 생태학적 연구(ecological study): 각 개인이 아닌 커뮤니티 혹은 집단이 하나의 관찰 단위가 됨, 연구결과에 편향이 개입될 소지가 많으므로 주의해야함

### 다기관 연구(multicenter study)

- 두개 이상의 기관이 참여하는 연구
- 다기관 연구자료를 분석할 때에는 기관 효과(center effect)를 고려해야만 함
- 집락 자료에 적절한 분석법, 다변수 회귀분석으로 기관의 효과를 보정하는 방법

### 인과성(causality) 평가

의학 연구의 주된 관심은 대부분 특정 요인에 노출되는 것이 결과의 원인이 되었는가를 평가하는데 있다.

인과성 성립 조건에 관한 몇 가지 기준들이 만족되기만 한다면 관찰연구도 실험연구처럼 인과 관계를 평가하는데 사용될 수 있다.

**Hill이 제시한 기준**

- 원인은 결과에 선행되어야 한다.
- 개연성이 있어야 한다. (생물학적으로 발생 가능한 연관성이어야 한다.)
- 기존의 다수 연구결과들과 일관성이 있어야 한다.
- 원인과 결과 간 연관성 크기가 커야 한다.
- 효과에 대한 용량-반응관계가 성립되어야 한다. (효과의 수준이 올라갈수록 질병의 정도가 더 심해지거나 질병의 발현이 더 빨라져야 한다.)
- 해당 요인이 제거되면 질병의 위험이 감소해야 한다.

### 단면연구(cross-sectional study)와 종적 연구(longitudinal study)

- 단면연구: 설문조사나 센서스 등과 같이 어느 한 시점에서 수행되는 연구
 1. 설문조사(survey): 특정 주제에 대한 개인의 신념이나 성향 등을 파악하기 위해 인구집단으로부터 대규모 표본을 추출해 조사하는 연구
 2. 센서스(census, 인구조사): 설문조사의 특별한 형태 중 하나, 목표 모집단 전체를 전수조사 하는 것

단면연구는 인구집단 내 특정 조건에 대한 시점 유병률(point prevalance)을 추정하고자 할때 효과적임

$$시점~유병률 = \frac{해당~시점에서~질병을~가지고~있는~사람들의~총~수}{특정~시점의~총~연구대상자~수}$$

단면연구에서는 연구 시작 전 어떤 시점에서 사건들이 발생했는지 파악할 수 없기 때문에 관심 요인이 질병의 원인이었는지에 대한 여부는 평가 할 수 없고, 단지 해당 요인과 질병 간의 연관성 여부만을 확인할 수 있다. 또한, 질병의 **발생률(incidence)**도 추정이 불가능하며 경향성(trend)도 평가할 수 없다.

그러나 단면연구는 일반적으로 빠르고 쉽게 수행할 수 있다는 장점을 가지고 있다.

- 반복적 단면연구(repeated cross-sectional study): 시간에 따른 경향성을 파악하기 위해서 여러 시점에 걸쳐 단면연구를 반복적으로 실시하는 것

각 시점 별로 참여하는 연구대상자들이 다르므로 시점들 간 명백한 변화가 관찰되는 경우, 해당 사건의 특성 때문인지, 연구대상자의 특성이 변했기 때문인지 파악할 수 없다.

- 종적 연구: 동일한 연구대상자들을 시간이 흐름에 따라 추적하는 연구
 1. 전향적(prospective) 연구: 대부분의 종적 연구, 특정 시점을 기준으로 연구대상자들을 시간의 흐름에 맞춰 앞으로 조사해 나감
 2. 후향적(retrospective) 연구: 연구 대상자들을 모집한 뒤 이들이 과거에 경험했던 요인들을 거꾸로 추적 조사해 나감

종적 연구는 단면연구보다 시간이 더 오래 걸리기 때문에 상대적으로 더 많은 자원이 필요하게 되며, 연구가 의무기록지나 연구대상자 본인의 기억 등에 의존하는 경우에는 **편향**이 발생할 소지가 많다.

### 대조군(control group)

비교 집단 또는 대조군의 사용 여부는 연구를 설계하고, 그 연구결과를 해석하는 데 있어서 매우 기본적인 사항이다.

### 편향, 비뚤림(bias)

연구로부터 얻어진 결과와 실제 상태 간 구조적인 차이(systematic difference)가 존재한다면 이를 편향이 발생 되었다고 한다.

## 13 연구 설계 II

### 변동(variation)

- 자료가 변동을 보이는 이유는 생물학적 요인 (예를 들면 성별, 연령 등), 측정 '오차(error)'(예를 들면 관찰자 변동), 혹은 **설명이 불가능한 무작위 변동(unexplained random variation)** 등이 복합적으로 작용하기 때문이다.
- 자료의 변동이 모수를 추정하는 데 미치는 영향력의 크기는 표준오차로 측정한다.
- 좀 더 정밀한 추정값을 얻기 위해서는 가능한 한 이 변동의 영향력을 줄이도록 노력해야 한다.


### 반복(replication)의 사용

- 반복: 모든 연구대상자들로부터 해당 변수와 관련해 두 개 이상의 측정값을 얻는 것
- 반복측정된 자료들은 서로 독립적이지 않기 때문에 분석에 이를 반영함 (가장 쉬운 방법: 평균을 분석에 사용)

### 표본 수(sample size) 계산

- 표본 수가 커지면 추정값의 표분오차가 작아짐 따라서, 추정값의 정밀성 및 연구의 검정력(power)이 높아짐
- 반응률(response rate)이 적어도 단면연구나 설문조사 연구에서 관찰되는 반응률보다는 높아야함 (추적실패율을 최소화)

### 특별한 형태의 설계방법을 사용

- 실험단위(experimental unit): 자료 분석 시 서로 독립적인 값으로 간주될 수 있는 '개체' 혹은 '개체들'로 이루어진 최소 크기의 집단. ex) 환자, 혈액량, 피부 부착포, 치아 등
- 완전확률화설계(completely randomized design): 실험단위를 처리(treatment)에 무작위로 배정한, 기타 다른 제약조건이 없는 실험설계
 1. 분석이 매우 간단
 2. 실험단위 간 큰 양의 변동 존재 시 비효율적 - 블록화(blocking) 또는 교차설계(cross-over design)을 사용
- 블록화(blocking): 비슷한 특성을 가지는 실험단위들을 동질적인 블록(block)이나 층(stratum)으로 묶음 후 각 블록 내에서 무작위 배정
- 병행설계(parallel design): 개체-간 비교(between-individual comparison)
- 교차설계(cross-over design): 개체-내 비교(within-individual comparison), 잔류효과 처리 해줘야함
 * 잔류효과(carry-over effect): 각 처리를 받는 시점 사이 먼저 받은 처리로 인해 남아 있을지도 모르는 효과로 세정기간(washout period)를 둔다.

### 적절한 연구 종료점(study endpoints)을 선택

- 연구 종료점
 1. 각 개인으로부터 관찰되는, 명확하게 정의된, 결과변수
 2. 연구계획서 상 미리 명시
 3. 가설과 관련 O
 4. 임상적/보건학적이나 생물학적으로도 의미
 5. 임상적 지표(발병, 재발, 사망), 대리 표지자(surrogate marker)(AIDS의 경우 HIV 바이러스 부하량 측정값) 사용 가능
 6. 대리 표지자 종료점은 임상적 종료점을 사용하기 어려울 때 대안으로 사용
 7. 상황에 따라 복합사건 종료점(composite endpoint) 사용: 연구 대상자가 경험할 수 있는 하나 이상의 종료점의 집합
