# Chapter 14. 임상시험

#### 임상시험(clinical trial)
 - 새로운 처치(treatment)가 임상적 결과에 미치는 효과를 평가하기 위해 인간을 대상으로 설계된 실험연구 형태의 연구 설계
1. 제1상 임상시험(phase I trial): 시험약의 임상 약리에 관한 연구(clinical pharmacologic study)
2. 제2상 임상시험(phase II trial): 안전성 및 유효성을 평가하기 위한 소규모 임상연구(small clinical study)
3. 제3상 임상시험(phase III trial): 새로운 처리에 대한 전반적 평가를 위한 연구(full evaluation of the new treatment)
4. 제4상 임상시험(phase IV trial): 시판이 허가된 약제의 장기간 위험성과 유익성을 평가하는 연구(post-marketing study for long-term risk and benefit evaluation)

#### 대조군(control)
 - 대조군이 없다면 처리를 받는 군에서 나타나는 반응이 전적으로 해당 처리의 효과 때문인지 파악하기란 사실상 불가능
1. 양성대조군(positive control): 표준약제가 있다면 대조군으로 표준약제를 사용
2. 음성대조군(negative control)
 - 위약대조군(placebo-control): 새로운 약제와 비교해 맛과 모양이 완전히 똑같지만 내용물에는 어떠한 종류의 활성 화합물도 들어 있지 않는 처리
 - 무처리대조군: 윤리적인 문제가 없는 경우에 한해서 아무 처리도 시행하지 않는 대조군

#### 일차 및 이차 종료점(primary and secondary endpoints) 
 - 연구시작 단계에서 무엇을 종료점으로 사용할 것인지에 대해 결정하기 전에, 새로운 요법의 효과를 가장 정확히 반영해 줄 수 있는 결과변수가 무엇인지 결정
1. 일차 종료점(primary endpoint): 처리의 효능(efficacy)과 관련이 있는 변수 선택
2. 이차 종료점(secondary endpoint): 약물의 독성(toxicity)과 관련이 있는 변수 선택

#### 처리의 배정(treatment allocation)
1. 연구대상자를 처리군들 중 하나에 무작위로 배정하는 이유: 처리군들 간 기저 특성들(baseline characteristics)를 유사하게 만듦(무작위화를 통한 할당편향(allocation bias)을 방지하고 교란(confounding)을 제거하기 위해
2. 무작위배정 방법
 - 층화 무작위배정(stratified randomization): 연구대상자들의 중요한, 그러나 연구관심 요인은 아닌, 기저 특성들(예를 들면 연령, 성별 등)이 반응값에 미치는 영향을 통제하기 위해, 연구대상자들을 기저 특성들의 수준 별로 층화한 뒤, 각 층 내에서 무작위배정을 실시하는 방법
 - 블록 무작위배정 또는 제한 무작위배정(block or restrict randomization): 처리군 개수의 배수에 해당하는, 비교적 적은 크기의 블록 크기(block size)를 결정한 뒤, 각 블록 내에서 일종의 변형된 형태의 무작위화 방법을 사용해 연구대상자들의 수가 군 간에 동일하게 배정되도록 함
 - 집락 무작위배정(cluster randomization): 집단(group)이나 집락(cluster)을 처리군에 배정하는 방법

#### 눈가림(blinding) 또는 은폐(masking)
 - 환자나 의사가 처리의 배정 내용을 알 수 있게 된다면, 특히 반응값이 주관적으로 평가되는 임상시험의 경우, 평가 편향(assessment bias)가 발생할 수 있음

#### 단일 눈가림과 이중 눈가림의 차이점
1. 단일 눈가림(single-blind): 의료진이나 평가자들만이라도 처리의 배정 내용을 모르게 하는 것
2. 이중 눈가림(double-blind): 환자와 의료진, 평가자들 모두에 대해 처리의 배정 내용을 모르게 하는 것

#### 축차설계(sequential design)의 원칙들
1. 두 개의 처리(새로운 처리 및 대조 처리)를 비교하고자 하는 경우, 개인들을 각 처리에 두 명씩 '짝(pair)'을 지어 무작위배정을 함
2. 각 짝에 속한 두 명의 결과가 순차적으로(sequentially) 얻어지면 그 때까지 누적된 현재 가용한 모든 자료(all the data currently available)을 분석함
3. 특정 통계적 기준을 사용해서(만일 두 처리군 간에 확실한 차이가 있는 것으로 밝혀지는 경우, 또는 향후에도 두 처리군 간에는 차이가 없을 것이라는 점이 명백해지는 경우) 해당 임상시험을 조기종료하는 것으로 만일 그렇지 않으면 다른 짝을 추가로 등록해서 무작위화를 실시하는 것으로 결정

#### 연구계획서 순응(PP)분석과 배정된 대로 처리(ITT) 분석
1. 연구계획서 순응(per-protocol, PP) 분석: 연구계획서 상에 미리 명시된 대로 온전하게 끝까지 연구에 참여한 환자들만을 대상으로 분석
2. 배정된 대로 처리(intention-to-treat, ITT) 분석: 모든 환자는 그들이 실제로 임상시험 중 어떤 처리약제를 투여 받았는지에 상관없이 무작위배정 시 배정된 군에 속하는 것으로 간주하고 분석

#### 연구계획서(Protocol)에 포함되어야 할 내용들
임상시험의 목적 및 목표, 환자 모집에 관한 자격기준(eligibility criteria)-> 포함기준(inclusion criteria) 및 제외기준(exclusion criteria), 약제 정보 및 투여 스케줄, 자료 수집 및 분석 방법, 긴급사태 대처방안, 연구참여자 정보

#### 코호트 연구
코호트 연구(cohort study): 특정 병인학적 요인에 노출되는 것이 질병 발생에 어떤 영향을 미치는지를 파악하기 위한 목적으로, 동일한 특성을 가진 일련의 연구대상자 집단(이를 코호트라 한다)을 시간의 흐름에 따라(일반적으로는 앞 방향으로) 조사해 나가는 연구

#### 고정 코호트(fixed cohort)와 역동적 코호트(dynamic cohort)의 차이
1. 고정 코호트: 연구대상자가 코호트를 떠나는 경우에 이를 새로운 연구대상자로 대체하지 않을 때
2. 역동적 코호트: 코호트에서 탈락하는 연구대상자가 생길 때 이를 자격조건을 갖춘 다른 사람으로 대체할 때 

#### 코호트 관련 용어
1. 역사적 코호트(historical cohort): 후향적 코호트 연구로, 노출 여부와 해당 결과에 관한 현 시점까지의 정보들을 의무기록이나 연구대상자의 기억을 토대로 확인
2. 위험요인(risk factor): 어떤 요인에 노출되는 것이 질병 발생에 영향을 미치는 것으로 밝혀질 때의 이 요인
3. 건강인 진입효과(healthy entrant effect): 건강한 사람들을 대상으로 연구가 실시될 때 연구대상자들이 연구 초반에 보여주는 사망률이 일반 인구집단에서 기대되는 사망률보다 더 낮게 되는 현상
4. 임상 코호트 or 질병 레지스트리(disease registry): 한 곳 이상의 병원에서 동일한 임상 소견을 보이는 환자들을 대상으로 코호트를 구축한 뒤, 이들을 추적조사해서 얼마나 많은 환자들이 치료 또는 완치가 되는지, 혹은 질병이 어떤 방식으로 사망이나 재발로 진행되는지 등에 대한 연구를 수행

#### 코호트 연구의 장점
1. 시간의 흐름에 따른 사건의 진행상황을 파악할 수 있다
2. 다양한 결과들에 대한 정보를 확보할 수 있다
3. 질병의 발생률 혹은 위험도를 직접 측정할 수 있다
4. 여러 요인들에 관한 노출 정보를 매우 상세하게 수집할 수 있다
5. 희귀한 요인에 노출되는 경우에 대한 연구가 가능하다
6. 노출에 관한 정보를 여러 시점에서 측정할 수 있기 때문에 시간에 따른 노출의 변화 효과를 연구할 수 있다
7. 사례-대조 연구에 비해 회상편향(recall bias)이나 선택편향(selection bias) 등의 발생 가능성이 낮아진다.

#### 코호트 연구의 단점
1. 일반적으로 코호트 연구는 연구대상자들을 오랫동안 추적조사하기 때문에 비용이 많이 든다
2. 관심 질병이 희귀질병인 경우에는 매우 큰 표본 수가 필요하게 된다
3. 추적기간이 길어지면 길어질수록 추적에 실패하는 환자 수가 늘어나 편향된 결과를 얻을 가능성이 높아진다
4. 연구가 장기간 진행되기 때문에 측정값 관찰에 일관성을 유지하기가 힘들다. 또한 연구대상자의 행동도 초기 인터뷰 시점과 비교해 변할 가능성이 있다
5. 시간이 흐름에 따라 질병의 결과나 발생률 또는 질병 자체의 병인이 변할 수도 있다

#### 코호트 연구를 관리할 때 유념해야 할 사항들
1. 결과변수 - 결과변수를 명시하고 이에 대한 객관적인 정의를 제시한다. 결과변수를 어떻게 측정할 것인지에 대해서도 명시한다
2. 노출변수 - 어떤 노출변수들을 조사할 것인지 명시하고, 이들에 대한 객관적인 정의를 제시한다. 노출여부를 어떻게 확인할 것인지도 명시한다.
3. 연구대상자들의 모니터링 - 연구대상자들을 어떻게 조사할 것인가?, 추적관찰은 얼마나 자주 할 것인가? 각 시점에서 어떤 정보들을 조사할 것인가? 생물학적 표본을 수집할 것인가? 등을 명시한다
4. 코호트의 표본 수 및 추적기간 - 관심변수에 노출된 경우와 노출되지 않은 경우 각각 결과가 관찰될 빈도는 얼마나 되는가? 충분한 크기의 연관성이 있음을 증명하기 위해서는 어느 정도의 표본 수가 필요한가? 등에 대해 명시한다
5. 잠재적 교란변수(potential confounder) 및 효과 조정자(effect modifier) 들에 대한 정의 및 확인 방법 - 이와 관련된 변수들은 무엇이며, 이들을 어떻게 조사할 것인지에 대한 명확한 정의를 제시한다
6. 통계분석 계획 - 언제 코호트에 대한 통계분석을 수행할 것인지를 명시한다
7. 편향을 줄이기 위한 방법들 - 코호트 내 탈락률을 최소화하기 위한 방법들에 대해 명시한다. 결과변수, 노출변수, 잠재적 교란변수 및 효과 조정자들에 대한 정의와 평가기준들이 시간에 따라 변하지 않도록 하는 방법들을 명시한다
8. 질 관리 계획 - 중간분석 시점에서 아래 사항들을 보장하기 위해 고려한 통계분석 방법들에 대해 명시한다.
 - 추적실패율이 지나치게 높지 않도록 함
 - 노출변수, 결과변수 및 다른 주요 자료들을 측정하고 평가하는 방법이 시간에 따라 변하지 않고 일관성이 유지되도록 함
 - 결과변수가 처음에 예상했던 속도만큼 발생되고 있어서 연구가 계획된 대로 진행될 수 있도록 함
9. 연구윤리심의위원회(IRB)의 승인 및 연구대상자 동의서 - 이와 관련된 사항들이 필요한 지를 명시한다. 연구대상자의 동의가 필요한 경우이면 동의서를 어떻게 받을 것인지에 대해 명시한다

#### 상대위험의 계산과 해석
1. 상대위험(RR): 위험요인에 노출되면 질병 발생 위험이 몇 배나 증가(혹은 감소)하는지에 대한 척도
2. 상대위험= 1: 노출된 집단과 노출되지 않은 집단 간의 질병 발생 위험은 같다
3. 상대위험> 1: 위험요인에 노출된 집단의 질병 발생 위험이 노출되지 않은 집단에 비해 상대적으로 더 높음
4. 상대위험< 1: 요인에 노출되는 경우 질병발생 위험이 줄어든다

### Chapter 16. 사례-대조 연구

<img src="C://Users/Administrator/Desktop/fig16-1.jpg" width="80%">

#### 발생과 유병의 차이
1. 발생 건(incident case): 질병이 진단된 시점에 연구에 등록되는 환자들
2. 유병 건(prevalent case): 연구 등록 전에 이미 질병이 있는 것으로 진단된 환자들

#### 사례-대조 연구에서 대조군의 선택 방법들
1. 자격기준(선정기준/제외기준)이 애매모호하지 않고 정확해야 함
2. 연구시작 시점에 해당 질병을 가지고 있지 않다는 점이 보장되도록 선별되어야 함
3. 가능한 한 사례군과 동일한 장소에서 선택해야 함

#### 오즈비의 계산과 해석

<img src="C://Users/Administrator/Desktop/table16-1.jpg" width="80%">

$$\mbox{오즈비}=\frac{\mbox{노출 집단에서 사례 환자가 나올 오즈}}{\mbox{비노출 집단에서 사례 환자가 나올 오즈}}$$

$$\mbox{노출 집단에서 사례 환자가 나올 오즈의추정값}, \mbox{odds}_{exposed} = \frac{a/(a+b)}{b/(a+b)} = \frac{a}{b}$$

$$\mbox{비노출 집단에서 사례 환자가 나올 오즈의 추정값}, \mbox{odds}_{unexposed} = \frac{c/(c+d)}{d/(c+d)} = \frac{c}{d}$$

$$\mbox{오즈비 추정값}, OR = \frac{odds_{exposed}}{odds_{unexposed}} = \frac{a/b}{c/d} = \frac{ad}{bc}$$

#### 빈도별 짝짓기와 대응별 짝짓기의 차이
1. 빈도 짝짓기(frequency matching): 집단 짝짓기(group matching), 잠재적 위험요인들 각각에 대해 사례군 집단 전체의 평균과 대조군 집단 전체의 평균이 비슷하게 되도록 사례군과 대조군을 선택
2. 대응 짝짓기(pairwise matching): 사례군 한 명에 대해 그와 잠재적 위험요인들이 비슷한 대조군을 한 명 선택하는 것

#### 오즈비를 상대위험에 관한 추정값으로 사용할 수 있는 경우
희귀질병(rare disease)의 경우

#### 사례-대조 연구의 장점
1. 일반적으로 수행하기 쉽고 빠르며 비용도 적게 든다
2. 특별히 희귀질병을 연구할 때 효과적이다
3. 하나의 연구를 통해 다양한 위험요인들을 조사할 수 있다
4. 추적실패가 발생하지 않는다

#### 사례-대조 연구의 단점
1. 사례군과 대조군 간에 과거 사건에 대해 자세한 기억을 해내는 능력이 서로 다르면 회상 편향(recall bias)이 발생함
2. 위험요인에 노출되는 사건보다 질병의 발병이 먼저이면 인과관계를 추론할 수 없음
3. 위험요인에 노출되는 사건이 희귀한 경우에는 사례군을 찾기 힘들고, 이 경우에는 사례-대조 연구를 실시하는 것이 부적절함
