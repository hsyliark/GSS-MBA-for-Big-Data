#### `a)` Interpret the reported correlations between length of stay and age, MMA and albumin, as well as the results of any significance tests which relate to these correlation coefficient. When would it be most appropriate to display the Spearman correlation coefficient rather than the Pearson correlation coefficient when assessing the association between two values? Is there any evidence to suggest that this was appropriate in this situation?

- Spearman 상관계수 분석결과에 따르면 환자가 병원에 머무는 시간(length of stay)은 환자의 나이(age), 그리고 메틸말론산(MMA)의 수치와 유의한 양의 상관관계를 나타내고 있으며, 알부민(albumin), 비타민B6(vitamin B6), 그리고 영양평가점수(Mini Nutritional Assessment (MNA) score)와는 유의한 음의 상관관계를 나타내고 있다. Spearman 상관계수는 순위를 사용하기 때문에 자료에 있는 변수의 분포가 정규성을 만족한다고 보기 어려울 경우에 사용이 가능한 비모수적 통계방법이다. 이 때문에 정규성이 만족된다고 할 때 사용할 수 있는 Pearson 상관계수보다 사용할 수 있는 연구의 영역이 더 넓다.  

#### `b)` Comment on the way in which the authors have chosen to present the P-values for these correlation coefficients.

- 상관계수 분석에 대한 P-value를 제시하는 방법은 양의 상관관계를 설명하고자 하는 경우와 음의 상관관계를 설명하고자 하는 경우로 나눌 수 있고 검정통계량은 $Z=\frac {r_{s}} {\sqrt{\frac {1}{n-1}}} \sim N(0,1^2)$ 이다. 만약 유의수준이 $\alpha=0.05$라고 한다면 검정 $H_{0} : r_{s}=0$ v.s $H_{0} : r_{s}>0$의 경우는 산출된 검정통계량의 값보다 더 큰 검정통계량을 얻을 확률이 0.05보다 작은 경우 해당하는 두 변수 사이에는 유의미한 양의 상관관계가 있다고 결론내리게 되며, 검정 $H_{0} : r_{s}=0$ v.s $H_{0} : r_{s}<0$의 경우는 산출된 검정통계량의 값보다 더 작은 검정통계량을 얻을 확률이 0.05보다 작은 경우 해당하는 두 변수 사이에는 유의미한 음의 상관관계가 있다고 결론내리게 된다.

It has been reported that older individuals experiencing severe burns have a higher mortality rate than younger individuals with severe burns. To asseess whether any differences in mortality can explained by differences in demographic or clinical characteristics of older and younger individuals. Albornoz and colleagures (2011) described a case-control study of 66 patients aged 65 years or older (case: over-65 group) and 22 patients younger than 65 years (control: under-65 years) who were admitted with severe burns to the intermediate and intensive care unit of the National Burn Adult Reference Centre in Santiago, Chile. The investigators collected information on the burn agent (eg. fire or scald), whether inhalation injuries were present, the burn surface area expressed as a percentage of the total body surface area (%TBSA), the deep burn surface area, also expressed as a percentage (%DTBSA), the DTBSA:TBSA proportion and the primary outcomes of the study were the length of stay in the intensive care unit and mortality.

A summary of the characteristics of patients in the two groups is shown in Table 2. In unadjusted analyses, the mortality rate was significantly higher in the over-65 group (odds ratio 2.9, 95% confidence interval 1.6-5.2, $P<0.001$); after controlling for other potential confounders in a multivariable logistic regression analysis, those aged 
≥65 years were over 12 times more likely to die from their burn than those in the younger age group (adjusted odds ratio 12.02, $P<0.001$). The adjusted odds ratio for %TBSA was 1.06 per unit increment ($P<0.001$), where as that for the DTBSA:TBSA proportion was 7.6 per unit increment ($P<0.001$). After adjustment using a multivariable logistic regression analysis, there were no significant associations between mortality and gender, the presence of inhalation injuries or any other comobidities (eg. epilepsy, obesity, smoking or cardiovascular disease).
