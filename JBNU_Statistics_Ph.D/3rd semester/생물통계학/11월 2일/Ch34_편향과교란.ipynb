# Chapter 34. 편향(비뚤림)과 교란(혼란)

## 1. 편향(비뚤림, bias)
 
#### 1) 편향
- 연구결과와 실제 상태 간에 구조적인 차이(systematic difference)가 있는 것  
- 효과의 과대추정$\to$ 실제로는 존재하지 않는 관련성을 마치 있는 것처럼 보이게 함  
- 효과의 과소추정$\to$ 실제로 있는 관련성을 없는 것처럼 감출 수 있음  

**편향의 종류**

+ 선택편향  
+ 정보편향  
+ 연구비편향(funding bias): 연구결과를 (제약회사 등) 재정을 지원을 하는 측에 유리한 방향으로 보고하는 경향을 의미  
+ 출판편향(publication bias): 통계적으로 유의하게 나온 결과나 시사성이 높은 결과들만을 추려서 학술지에 투고하는 경향성
 
### 2) 선택편향(selection bias)
연구대상자들이 연구결과가 적용될 모집단을 대표하지 못함으로 인해서 발생하는 편향  
> ex) 연구참여에 동의한 환자들은 참여를 거부한 환자들과 비교해 특성이 다를 수 있음  

+ 확인편향(ascertainment bias)
  - 표본이 모집단으로부터 무작위로 추출되지 않았기 때문에 표본 내에서 모집단에 관한 주요 특성들이 서로 다르게 되는 경우
  -  확증편향이라고도 함 
  > ex) 특정 신체조건의 유전 현상에 관심이 있는 임상의가 연구대상자를 모집단으로부터 무작위추출하지 않고 본인의 병원에 내원하는 환자들만을 대상으로 자료를 모으는 경우  
+ 마모편향(attrition bias)
  - 종적 연구에서 추적실패가 된 사람들과 연구종료 시까지 추적관찰된 사람들의 특성이 서로 다른 경우  
+ 건강인 진입효과(healthy entrant effect)
  - 종적 연구에서 연구 초반에 질병이 없는 사람들이 포함되면 추적의 초기 단계에 일반적으로 예상되는 것보다 더 낮은 사망률이나 이환율을 보이게 됨  
+ 반응편향(response bias)
  - 연구에 자발적으로 참여한 사람들의 특성이 연구참여를 거부한 사람들과 다른 경우  
+ 생존편향(survivorship bias)
  - 특정 치료를 받는 환자들과 치료를 받지 않는 환자들의 생존 시간을 비교하고자 하는 경우
  - 만일 해당 치료가 연구시작 후 일정 시점이 지난 후에 이루어지는 것이라면 오래 생존한 사람만이 치료를 받을 수 있음
   


### 3) 정보편향(information bias)
자료수집 단계에서 발생될 수 있는 편향으로, 노출여부에 관한 측정이나 결과발생과 관련된 정보가 구조적으로 잘못 기록되기 때문에 발생하는 것  
+ 중심경향성 편향(central tendency bias)
  - 리커트 척도(Likert scale)을 사용하는 경우, 응답자들이 해당 척도의 가운데를 선택하는 경향  
+ 조기발견편향(lead-time bias)
  - 생존시간에 관한 연구를 하는 경우, 
  - 시간이 지남에 따라 진단기기의 정확성이 향상되고, 따라서 연구 초기에 등록된 환자들보다는 연구 후반부에 등록된 환자들의 질병상태가 좀 더 조기에 발견
  - 이로 인해 연구 전반부에 등록된 환자들에 비해 후반부에 등록된 환자들의 진단 후 생존시간이 더 길어지게 되는 현상  
+ 측정편향(measurement bias)
  - 측정기기가 부정확해서 구조적인 오차가 발생한 경우
  - 특정 값에 대한 선호도(digit preference) 또는 반올림 오차(rounding error) 등도 측정편향을 발생시키는 원인들 중 하나  
+ 오분류편향(misclassification bias)
  - 범주형 노출변수나 결과변수에 대한 분류가 잘못되었기 때문에 발생되는 편향  
+ 관찰자편향(observer bias)
  - 한 관찰자가 특정 변수를 일관되게 높거나 혹은 낮게 평가/보고함으로 인해서 발생되는 편향
  - 평가편향(assessment bias)이라고도 함  
+ 회귀희석편향(regression dilution bias)
  - 하나 이상의 설명변수가 부정확하게 측정됨으로 인해 발생하는 편향
  - 희석편향(attenuation bias)이라고도 함 
  > ex) 한 설명변수의 측정오차가 매우 크면 이 변수의 회귀계수에 관한 통계적 유의성이 과소추정될 수 있음  
+ 보고편향(reporting bias)
  - 연구대상자들이 연구자가 원하는 방향으로 응답하거나, 혹은 사회적으로 받아들여지지 않는 부끄러운 행동 또는 질환들에 대해서는 축소 보고를 하기 때문에 발생되는 편향  
+ 평균으로의 회귀(regression to the mean)
  - 한 사람의 첫 번째 관찰값이 유난히 낮았다면 이 사람을 다시 측정했을 때는 처음보다 더 높은 값이 측정되게 되고, 반대로 첫 관찰값들이 유난히 높은 사람의 두 번째 관찰값은 처음보다 더 낮게 측정되어서, 전체적으로는 평균을 맞추려는 경향성을 보이는 것  
+ 생태학적 오류(ecological fallacy)
  - 모집단 내 집단 별 요약통계량 값을 사용해서 분석하는 경우에 발생되는 편향
  - 분석단위로 개인이 아닌 집단을 사용해 분석했을 때 얻어진 집단 요약통계량 값들 간의 상호 관련성이 같은 모집단 내 개인별 자료 수준에서도 동일하게 적용될 것이라고 보는 것이 잘못 
  > ex) 사회경제적 수준에 관한 변수로 지역-단위로 측정된 자료를 사용하는 경우
  > - 사회박탈지수가 상대적으로 더 높은 지역에 거주하는 사람들의 병기 III이나 IV 단계 유방암에 걸릴 가능성이 더 높게 나타났다고 할지라도, 이 결과를 해당 지역에 살고 있는 개별 여성들에게 적용할 수 없음
  




 

## 2. 교란(혼란, confounding)

### 1) 교란은 무엇인가?  
잠재적인 위험요인과 질병발생 간의 관계를 파악하고자 할 때 교란변수의 효과를 적절히 보정하지 못해서, 실제로는 존재하지 않는 관련성이 마치 있는 것처럼 나오거나, 실제로 존재하는 관련성을 놓치게 되는 현상  
+ 교란변수(confounding variable 또는 confounder)
  - 종속변수와 관련이 있을 뿐만 아니라, 이와 동시에 모형 내 다른 설명변수들과도 관련이 있는 설명변수  

  > ex) 중년남성 코호트를 사용해 흡연이 관상동맥 질환(coronary heart disease, CHD)의 발생에 어떤 영향을 미치는지 조사  
 > - 만약 흡연과 더불어 음주도 CHD 발생과 관련이 있고, 또한 음주는 흡연하고도 관련이 있다는 것을(즉, 음주를 하는 남성은 음주를 하지 않는 남성에 비해 흡연을 할 가능성이 더 높다는 것을) 알고 있다고 가정  
  > - 이 경우 음주의 효과를 보정/조정(adjust, control)하지 않고 분석하면 음주(교란변수)는 흡연과 CHD 발생 간의 명백한 관련성을 교란 가능
  

### 2) 교란을 처리하는 방법
#### 1. 층화(stratification)
 교란변수의 각 수준 별로 자료들을 층화해서 부집단(subgroup)들을 만든 다음(예를 들어 음주자와 비음주자 등 두 개의 부집단을 만든 다음), 각 부집단 별로 분석 실시  

<장점>  
  - 결과를 간단하게 시각화 할 수 있으며, 결과의 해석 또한 쉽다  
  - 직관적이다  
  - 교란변수의 분포가 노출집단들 간에 충분히 겹치는 지에 대한 검토를 할 수 있다  
  - 분석결과가 교란변수와 결과변수 간의 관련성 형태(예를 들면 선형성)에 관한 어떠한 가정에도 영향을 받지 않는다  

<단점>  
  - 교란변수의 수가 적은 경우에만 사용이 가능하다  
  - 작은 크기의 층이 형성될 수 있기 때문에 특정 층 내에서는 검정력이 낮아질 수 있다  
  - 다중검정으로 인해 가짜로 유의한 결과가 얻어질 수 있다  
  - 처리효과에 관한 단일 추정값(결합추정값)을 제시하기 힘들 수 있다  
  




#### 2. 짝짓기(matching)
 모든 교란변수들의 각 수준들 조합 별로(교란변수들을 사용해) 노출변수의 각 범주에 해당하는 사람들(예를 들어 흡연자 및 비흡연자) 간에 짝을 짓는다
  
<장점>
  - 직관적이며, 결과의 해석도 쉽다
  - 두 개 이상의 교란변수를 처리할 수 있다
  - 분석결과가 교란변수와 결과변수 간의 관련성 형태(예를 들면 선형성)에 관한 어떠한 가정에도 영향을 받지 않는다
  
<단점>
  - 모든 사람에 대해 짝을 찾지 못할 수도 있다; 분석시 짝지어지지 않은 환자들은 제외되기 때문에 검정력이 낮아질 수 있다
  - 교란변수가 결과에 미치는 효과에 대해서는 추정할 수 없다
  - 교란변수의 수가 많으면 짝짓기 작업이 매우 어려워 질 수 있다  

#### 3. 보정(adjustment)
 각 교란변수들을 다중 선형회귀모형이나 다중 로지스틱 회귀모형 또는 다중 포아송 회귀모형 등 다변수 회귀모형(multivariable regression model)에 설명변수로 포함시켜 이들의 효과를 보정
  
<장점>
  - 교란변수의 수가 많은 경우에 효과적이다. 단, 이 경우 표본 수는 적절히 커야 한다
  - 교란변수가 결과에 미치는 효과를 추정 가능
  - 계산이 복잡하지 않다
  
<단점>
  - 교란변수의 수가 매우 많은 경우에는 연구의 검정력이 낮아질 수 있다
  - 노출변수 집단들 간에 각 교란변수의 분포가 균형되어 있어야만 의미 있는 결과를 얻을 수 있다
  - 자료가 존재하는 교란변수들에 대해서만 보정 가능 





#### 4. 성향점수(propensity score) 계산
 각 개인이 노출변수의 관심범주에 속할(예: 비흡연자 대비 흡연자가 될) 성향(propensity)(또는 확률)에 관한 점수를 계산
  
<장점>
  - 노출변수가 두 범주인 경우에는 성향점수를 계산하기가 상대적으로 쉽다
  - 노출변수 집단들 간에 분포의 균형이 이루어지지 않는 교란변수들이 있더라도 성향점수의 분포는 균형될 수 있다
  
<단점>
  - 노출변수가 세 범주 이상인 경우에는 성향점수를 계산하기가 쉽지 않다
  - 노출변수가 시간에 따라 변하지 않는 경우에만 사용할 수 있다
  - 이 방법이 효과적이기 위해서는 표본 수가 매우 커야 한다  



#### 5. 성향점수(propensity score)의 활용
  
1) 성향점수 보정(다변수 회귀모형을 사용해서 통계적으로 보정)
  - 장점: 모형에 사용되는 공변량의 수가 일반적인 다변수 회귀모형을 사용하는 경우보다 더 적다
  - 단점: (교란변수와 결과변수 간의 연관성을 추가적으로 파악하기 위해) 교란변수들도 모형에 포함시키면 이 교란변수들과 성향점수 간에 다중공   선성이 발생할 수 있다
  
2) 성향점수 층화
  - 장점: 각 층 내에서, 잠재적 교란변수들의 영향력을 제거할 수 있다
  - 단점: 성향점수 층들 간 결합추정값을 계산하기 어려울 수 있다
  
3) 성향점수 짝짓기
  - 장점: 잠재적 교란변수들을 보정하지 않고도 이들의 효과를 제거할 수 있다
  - 단점: 짝짓기가 어려울 수 있다
 

 
### 3) 비무작위 연구에서의 교란
#### 1. 교란은 
모집단 내 위험요인들이 무작위로 분포되지 않게 되는 코호트 연구에서 특히 문제가 됨
  - 특별히 중재(처리)가 결과에 미치는 영향에 관한 연구를 하는 경우, 각 개인이 혹시 질병력이나 인구학적 요인 또는 생활습관 요인 등으로 인해   이러한 중재를 선택하게 되는 것은 아닌지, 그리고 이 요인들 중 결과에 영향을 주는 요인은 있는지 등에 대해 주의 깊게 살펴봐야 함
  - 중재를 받는 환자들의 특성이 중재를 받지 않는, 혹은 다른 중재를 받는 환자들과 다르다면 할당편향(allocation bias)또는 경로편향             (channeling bias) 등이 발생하게 됨  
  
  > ex) 중년 남성 코호트에서 코호트 등록 초기에 스타틴(statin) 계열의 약제를 복용하고 있었던 사람들과 피브레이트(fibrate) 계열의 약제를 복     용하고 있었던 사람들 간에 심장병 발생이 예방되는 차이가 있는지를 연구
  > $\to$ 각 개인이 어느 계열의 약제를 복용할 것인지는 다양한 요인들(예를   들면 본인의 중성지방 수준 등)에 의해 결정되게 될 것이며, 또한 이 요인들 중 대부분은 심장병 발생과 관련이 있음
 

 
### 4) 인과적 경로(causual pathway)와 교란
#### 1. 인과적 경로
 선(chain)들을 사용해 일련의 사건들이나 요인들이 어떤 경로를 거쳐 관심결과에 도달하게 되는지를, 즉, 선 상에 있는 각 단계의 효과가 전 단계에 있는 어떤 사건(들)에 의해 영향을 받게 되는지를 논리적으로 연결한 것
#### 2. 매개변수(intermediate variable)
 어떤 변수(B)가 노출변수(A)와 결과변수(C) 간의 인과적 경로 상에 놓이게 될 때 이 변수(B)  

> ex) 새로운 지질 강하제가 CHD 발생(결과변수 C)에 미치는 효과를 평가하기 위한 무작위배정 위약 대조 임상시험
> - 노출변수(A)는 각 개인이 새로운 지질 강하제를 받는지의 여부에 관한 두 범주 변수
> -  콜레스테롤 수치(B) 상승은 CHD 발생에 영향을 미치는 것으로 알려진 위험요인들 중 하나. 
> - 임상시험 시작 시점에서 두 처리군 간에 콜레스테롤 수준의 차이가 있다면, 이 차이는 보정(adjust)되어야 하고, 임상시험 실시 도중에 발생하는 콜레스테롤 수치(B)의 변화는 보정하지 않아야 한다.

 
### 5) 시간-의존적 교란(time-varying confounding)
#### 1. 시간-의존적 교란변수

어떤 변수가 관심 노출변수에 관한 잠재적인 교란변수이면서, 동시에 노출변수와 결과변수 간의 인과적 경로 상에 위치하게 되고, 노출변수 그 자체가 시간에 따라 변하는 경우
  
> ex) 항바이러스제 치료(관심 노출변수)가 HIV에 감염된 환자들의 생존기간(결과변수)을 늘려주는지의 여부를 파악하기 위한 목적으로 조사된 코호트 자료
> -  HIV는 개인의 면역체계를 점진적으로 무너뜨리게 될 것이며, 
> - 이는 CD4 세포 수(CD4 cell count)로 측정될 수 있다(이는 잠재적 교란변수)
> - HIV- 양성인 사람은 시간의 흐름에 따라 CD4 세포 수 감소(잠재적 교란변수가 인과적 경로 상에 있게 된다.)
> - 항바이러스 치료제는 주로 CD4 세포 수가 일정 수치 이하(일반적으로는 350 이하)로 낮아지는 HIV- 양성환자들에게만 제공(각 개인     내 노출변수의 값이 시간에 따라 변한다). 
> - 처치가 시작되고 나면 대다수 환자들의 CD4 세포 수는 급격히 상승하게 되고, 이러한 상승은 생존기간의 연장과 밀접한 관련. 
> - (감염기간 동안 주기적으로 측정되는) CD4 세포 수는 초기 치료의 예측인자이면서, 동시에 항바이러스 치료  의 개시와 사망 간의 인과적 경로 상에 있기 때문에, 
> - 시간-의존적 교란변수에 해당
 
  + 타당한 처리효과 추정값을 얻기 위해 인과적 모형화(causal modeling), 주변 구조적 모형(marginal structural model, MSM), G-추정법(G-         estimation) 사용 필요
  + 의학통계학자의 도움 필요
