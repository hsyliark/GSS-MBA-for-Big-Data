## Ch.36 표본 수 계산

### 표본 수(sample size)의 중요성

- 표본 수가 너무 적으면 실제 존재하는 중요한 효과를 파악해낼 수 있는 검정력이 상대적으로 낮아지게 됨. $\rightarrow$ 자원의 낭비(시간, 비용 등)

- 표본 수가 지나치게 많으면 불필요한 시간이 소모되고, 비윤리적일 뿐만 아니라 연구대상자들이 더 좋은 처치를 받을 수 있는 기회를 뺏음.

<b>$\rightarrow$</b> 연구설계 단계에서 제1종 오류와 제2종 오류를 범할 가능성의 균형을 맞춘 최적의 표본 수를 계산한 후 연구를 수행
 
  `-` 제1종 오류 : $H_{0}$가 참이나  $H_{0}$를 기각하는 오류
  
             (암이 맞으나 암이 아니라고 진단)
                    
  `-` 제2종 오류 : $H_{0}$가 거짓이나 $H_{0}$를 기각하지 않는 오류
  
             (암이 아니나 암이 맞다고 진단)
             
  `-` 제1종 오류가 제2종 오류보다 더 심각한 오류

`but`, 연구설계 단계에서 적정 표본 수를 계산하기 위해서는 해당 연구를 통해 기대되는 결과에 대한 예상값들이 필요하다는 모순이 존재

### 표본 수를 계산하는 데 필요한 정보들

- 최적의 표본 수를 계산하는 방법의 단순한 예:

  - 연구설계가 복잡 $\rightarrow$ 간단한 연구설계의 형태로 변환하여 표본수를 계산
  
  - 여러 개의 통계 검정을 실시하고자 하는 연구 
  
    - 가장  중요한 것을 선택해서 표본수를 계산
    
    - 각각의 검정에 필요한 표본 수들을 계산 한 다음, 가장 큰 값을 표본 수로 사용

- 일반적으로 연구의 적정 표본 수는 해당 연구에서 실시하고자 하는 가설검정 방법에 따라 달라짐

- 상황에 따라 연구의 다른 측면, 즉, 추정값의 정밀성이나 신뢰구간의 넓이 등을 사용해 표본 수를 계산 (비열등성 임상시험, 동등성 임상시험)

- 가설검정 방법을 사용해 최적의 표본 수를 계산하기 위해서는 `연구설계 단계`에서 다음 값을 결정

  - 검정력($power$; $1-\beta$) : 미리 설정된 크기만큼의 효과가 얻어지는 경우, 이 효과를 통계적으로 유의하다고 판정할 수 있는 확률. ($1-\beta=0.8$ 또는 $1-\beta=0.9$)<br><br>
  
  - 유의수준(significance leve; $\alpha$) : 귀무가설을 기각하기 위한 판단 기준으로 사용할 확률. ($\alpha=0.05$ 또는 $\alpha=0.01$)<br><br>
  
  - 관찰값의 변동(variability) : 연속형 변수의 경우 표준편차, 이산형 변수의 경우 범위<br><br>
  
  - 최소 유효크기(the smallest effect size) : 임상적으로 중요한, 그래서 간과할 수 없는, 관심효과의 크기. 일반적으로 군 간 차이값. 때로는 유효크기를 표준편차로 나눈 표준화된 차이(standardized difference)로 표현 (=MCID (Minimally Clinically Important Difference))

- 연구에 적절한 검정력과 유의수준의 크기 결정. 즉, 적용할 제1종 오류와 제2종 오류를 고려해서 결정

  - 의약품 허가를 받기 위한 연구들은 관계당국이 미리 해당 값들을 지정하기도 함

- 유효크기는 가상적인 실제 임상 상황 하에서 임상적으로 중요하다고 판단될 수 있는 값 선택

- 관찰값의 변동은 현실적으로 가장 결정하기 어려움: 즉, 연구 자료를 모으기도 전에, 적정 표본 수를 계산하기 위해서 연속형 변수에 대한 자료의 변동량 추정값을 결정해야 하는 상황.  
  - (1) 유사한 결과변수를 관찰했었던 기존 연구결과 사용
  - (2) 본 연구 이전에 예비연구(pilot study)를 실시해서 예상되는 변동량 크기
- 만일 연구계획서에 충분히 명시하기만 한다면, 내적 예비연구(internal pilot study)를 실시한 뒤, 이로부터 얻어지는 자료를 본 연구에 포함시킬 수도 있음(adaptive design in clinical trials).
- 적정 표본 수를 계산하기 위해서는 
  - 먼저, 비록 제한적이긴 하지만, 연구설계 단계에서 얻을 수 있는 가능한 모든 정보 사용
  - 다음에 예비연구에서 얻어진 정보들을 사용해 본 연구에 대한 표본 수 조정 (예를 들어, 이미 지정한 값의 크기가 너무 크다고 생각되면 실용적인 측면을 고려해 값을 조정). 
  - 이 경우, 내적 예비연구에서 얻어진 정보들은 최종 자료분석 시에 활용할 수도 있음.

`주의: 적정 표본 수는 예비연구에서 관찰된 효과크기가 아니라 처음에 가정한 최소유효크기를 사용해 계산해야 하며, 예비연구 결과를 사용한 표본 수 조정은 예비연구에서 관찰된 효과크기가 처음에 가정했던 최소유효크기보다 더 큰 경우에 한해서만 실시` 

### 표본 수 계산 방법

- 표본수를 계산하는 방법은 매우 다양. 각 방법이 요구하는 정보들은 기본적으로 동일
  - 정확한 공식 : 사용 방법이 복잡하긴 하지만 이 공식이 필요한 경우들이 있다.(집락 무작위 임상시험)
  <br><br>
  - 빠른 공식 : 특정 검정력 및 유의수준 하에서 일부 가설검정들에 대해서만 사용될 수 있는 공식 (레어(Lehr)의 공식)
  <br><br>
  - 표본 수 관련 표 : 여러 상황들에 적용할 수 있는 표본 수 들을 계산해 놓은 표가 있다. (t-검정, 카이제곱 검정, 상관계수에 관한 검정, 두 집단의 생존곡선을 비교하기 위한 검정, 비열등성 검정 등)(Machin, et al.(2018) Sample Size Tables for Clinical, 4th ed., Wiley-Blackwell)
  <br><br>
  - 알트만의 노모그램 : 여러 검정방법들에 대해서 사용하기 쉽게 그림의 형태로 만든 도표
  <br><br>
  - 통계 소프트웨어 : 적정 표본 수를 산출해주는 통계 소프트웨어를 사용하면 각 요인들(예를 들면 검정력이나 유효크기 등)의 수준이 변함에 따라 적정 표본 수의 크기가 어떻게 변하는지를 쉽게 파악할 수 있을 뿐만 아니라, 해당 계산 결과를 표나 그림의 형태로 알기 쉽게 표현할 수도 있다. (G*Power 등)

### 알트만의 노모그램(Altman's nomogram)

#### `기호`

표 36.1 (너비가 깨질 경우 삭제)

| 가설검정 <br>| 표준화된 차이 <br>| 노모그램에서 얻어지는 $N$의 의미 <br> | 용어 <br> |
| :--- | :--: | :----: | :----------------------- |
| 두 독립표본 t-검정(21장) | $\displaystyle \frac {\delta}{\sigma}$ | 전체 표본 수. <br>따라서 군별로는 $N/2$개의<br> 관찰값이 필요 | $\delta$: 임상적으로 중요하다고 판단할 수 있는 두 집단 간 평균차이의 최소값<br> $\sigma$: 각 집단 내 관찰값들의 예상 표준편차. 두 집단의 표준편차는 동일한 크기라고 가정.<br>  과거에 실시되었던 유사한 연구나, 혹은 이미 출판된 다른 연구들을 사용해서 예측 또는<br> 예비연구를 실시하여 추정. 다른 방법으로는 $\sigma$를 표준편차의 배수(표준화된 유효크기)로 <br> 표현 |
| 짝진 t-검정 (20장) | $\displaystyle \frac {2\delta}{\sigma_{d}}$ | $N$개의 짝의 관찰값수 | $\delta$: 임상적으로 중요하다고 판단할 수 있는 짝 간 평균차이의 최소값<br> $\sigma_{d}$: 짝 차이 값들의 표준편차. 일반적으로는 예비연구를 통해 추정 |
| 카이제곱 검정 (24장) | $\displaystyle \frac {p_{1}-p_{2}}{\sqrt{\bar{p}(1-\bar{p})}}$ | 전체 표본 수. <br> 따라서 군별로는 $N/2$개의<br> 관찰값이 필요 | $p_{1}-p_{2}$: 임상적으로 중요하다고 판단할 수 있는 두 집단 간 '성공' 비율 차이의 최소값<br> 일반적으로 두 비율 중 하나(주로 대조군의 비율)는 알려져 있는 경우가 많으며, 따라서<br> 나머지 하나의 값을 결정해서 이들 두 집단 간 차이가 중요한 차이라고 할 수 있는 값을 계산<br> $\bar{p}= \frac{p_{1}+p_{2}}{2}$|

- 표 36.1은 평균이나 비율을 비교하고자 할 때 자주 사용되는 세 종류 가설검정들에 대해서, 알트만의 노모그램(page.306)을 사용해 동일 크기의 두 집단을 비교하기 위한 표본 수를 계산할때 필요한 기호들을 설명

#### `방법`

1) 각 검정 방법에 대해 표준화된 차이를 계산

2) 노모그램의 왼쪽 수직 축에서 이 값에 해당하는 점을 찍음

3) 오른쪽 수직 축에서는 원하는 검정력에 해당하는 점을 선택

4) 이들 두 점을 직선으로 연결

5) 연결된 직선이 노모그램의 가운데 사선과 교차하는 지점에 해당하는 $N$값이 필요로 하는 전체 표본 수

-  주어진 표본 수에 대한 가설검정의 검정력을 평가하는 데에도 사용
  - 유의하지 않은 결과가 얻어졌을 때 그 이유가 부족한 표본 수 때문은 아니었는지를 후향적으로 평가하는데 도움이 됨
  
- **`사후 검정력 계산(post-hoc power calculation)`**은 사전, 즉 연구설계시에 가정했던 MCID 값을 사용해야 함. *`절대로 해당 연구에서 관찰된 처리효과 값을 사용해서는 안됨`*
- 관심효과에 대한 신뢰구간의 범위가 넓은 것 역시 연구결과로 얻어진 추정값이 정밀하지 않다는 의미. $\rightarrow$ 대부분 표본 수가 부족했기 때문

### 간단한 공식

- 두 독립표본 t-검정 또는 카이제곱 검정의 경우

**`레어의 공식(Lehr's formula)`** : $1-\beta$= 80, $\alpha$= 0.05

<center> $\displaystyle n = \frac{16}{(\Delta)^{2}} $</center>

- $n$: 군 당 표본 수, $\Delta$: 표준화된 차이
- `표준화된 차이`의 크기가 작을 때 표본 수를 과대추정하는 경향이 있음
- 분자에 16 대신 21을 사용하면 검정령 90%에 해당하는 공식

### 검정력에 관한 고찰

- 본인의 연구가 연구설계 단계부터 이미 적정 표본 수와 관련된 사항을 충분하고도 조심스럽게 고려했었다는 점을 효과적으로 알려주기 위해 연구계획서나 논문 내의 연구방법론에 관한 절(methods section;page.321)에 해당 연구의 검정력에 관한 내용(power statement)을 포함시키는 것은 매우 기본적인 사항
- 검정력 고찰에 관한 일반적인 예문 (예제 1 참조) : 양측 유의수준 5% 하에서, 두 독립표본 t-검정을 사용해 조기 물리치료군과 일반적인 치료군 간 3개월째 오스웨스터리 장에지수(ODI) 평균 7점(표준편차 16점) 차이를 평가할 수 있는 검정력 90%를 확보하기 위해서는 각 군 별로 110명의 환자가 필요하다.
  - ODI(Oswestry disability index) - 오스웨스트리 요통 장애 지수, 총 10개 문항, 문항별 6점 척도, score= (총 점수 $\times$ 100) / (5 $\times$ 답한 항목 수), score가 높을 수록 요통 장애 정도가 심함.

### 조정

- **추적실패(loss to follow-up)**가 발생할 것으로 예상되면 이를 반영하기 위해 연구 시간단계에서 환자들을 추가적으로 더 모집해야 하며, 따라서 표본 수 조정이 필요. 만일 계산된 전체 표본 수가 $N$이고 탈락률이 $r$(즉, 100$r$%) 정도 될 것이라고 가정한다면 예상 탈락률을 반영한 표본 수(조정된 표본 수)는 

<center> $\displaystyle N^{*} = \frac{N}{1-r}$</center>

- **군 별로 서로 다른 표본 수를 사용**하고자 하는 경우; 한 집단의 연구대상자 모집에 제한이 있어 각 군의 표본수를 동일하게 하기 힘든 경우 표본수 조정 필요
   - 예) 희귀질병에 관한 사례-대조 연구나 개발된 신약의 재고가 충분하지 않은 경우
   - 군간에 표본 수가 불균형인 설계의 경우, 균형설계와 동일한 크기의 검정력을 유지하기 위해서는 일반적으로 더 많은 수의 표본이 필요하게 된다.
   - 표본 수의 조정 : 두 군 간 표본 크기의 비율을 $k$라고 하면, 두 집단이 동일한 크기의 표본수를 가진다는 가정 하에 조정되지 않은, 전체 표본수 $N$을 계산한 뒤, 이 값을 음과 같이 조정
   <br><center> $\displaystyle N' = \frac {(k+1)^{2}}{4k}N$</center><br>
   그런 다음 작은 크기의 집단에는  $\displaystyle \frac {1}{k+1}N'$명을, 큰 크기의 집단에는  $\displaystyle \frac {k}{k+1}N'$명을 할당

### 표본 수가 정해진 상태에서 검정력을 증가시키기 위한 방법

표본 수가 결정된 상태에서, 만일 유의수준 및 군 간 차이(유효크기)가 고정 되었고`(검정력 증가의 목적으로 이 들 값의 변형은 논리적으로 인정되지 않음.)` 양측검정을 사용하는 경우`(단측검정을 사용하면 검정력이 증가하지만 일반적으로 부적절)`, 검정력을 증가시킬 수 있는 방법들은 다음과 같다.

- 결과변수를 변경, 즉, 보다 많은 정보를 제공하는 반응변수를 사용
    - 저혈압/정상혈압/고혈압 등과 같은 범주형 반응 변수 대신, 수축기 혈압 과 같은 연속형 반응변수 사용

- 주 분석 방법을 변경
    - 비모수적 방법 대신 모수적 방법 사용

- 자료수집 시 자료의 무작위 변동이 최소화되도록 실시
    - 자료수집 절차의 표준화 또는 관찰자에 대한 교육 등을 실시

- 연구계획서를 측정값의 변동을 줄일 수 있는 방향으로 수정
    - 층화를 실시하거나, 두 독립집단 대신 짝짓기 실시(교차설계)

`예제1` **두 독립표본 t-검정을 사용해 서로 독립적인 두 집단의 평균을 비교**

**목적**  요통(low back pain, LBP) 환자들의 기능장애를 완화시키기 위한 목적으로 조기물리치료(교정과 운동요법)를 시행하는 것이 일반적인 물리치료와 비교해 얼마나 더 효과적인지 평가. 

**자료수집 방법** 
- 기저시점에서 선정/제외기준을 만족하는 모든 환자들에게 가능한 한 최대한으로 일상적인 활동을 지속해 나가도록 권고 
- 연구진이 검토한 LBP 가이드라인 관련 문서 제공. 
- 조기물리치료군은 연구 등록 후 첫 4주 이내에 물리치료사로부터 계획된 일정표에 따라 총 4회의 치료받은 군. 
- 일반치료군은 첫 4주 동안 아무런 치료도 받지 않은 군.

**설계**  무작위배정, 단일 눈가림, 병행설계 임상시험(randomized, single-blind, parallel group clinical trial).

**표본 수를 계산하기 위해 사용한 주 결과변수** 기저시점 대비 3개월 후, 요통 기능장애 평가 지수 중 하나인 오스웨스터리 장애지수(Oswestery Disability Index, ODI; 0~100점 범위이며 값이 클수록 장애정도가 더 심하다) 값의 변화량.

**질문** - 양측 유의수준 5% 하에서, 조기물리치료군과 일반치료군 간 기저시점 대비 3개월째 ODI 점수 변화량의 평균 7점 차이를 두 독립표본 t-검정을 사용해 평가하고자 할 때, 검정력 90%를 확보하기 위해서는 몇 명의 환자들이 필요한가? 이 변화량에 대한 표준편차는 근사적으로 16점 정도 될 것이라고 예상한다.

**노모그램을 사용하는 경우**: 
- $\delta = 7$ $\sigma = 16\to$ 표준화된 차이(표준화된 유효크기) $\displaystyle \frac{\delta}{\sigma} = \frac{7}{16} = 0.44$
- 표준화된 차이 0.44와 검정력 90%에 해당하는 점 연결 $\to$ 유의수준 5%에 해당하는 표본 수 축은 대략적으로 값 220을 통과 (즉, $N = 220$). 
- 군 당 약 110명의 환자 필요 (n=110).

> **참고**
> - (i) 만일 $\delta$를 6으로 낮추면 표준화된 차이는 0.38이 되고,  이때 필요한 표본 수는 전체 300명, 즉, 군당 150명으로 증가 
> - (ii) 첫 번째 가정($\delta = 7$)을 사용하되, 일반치료군보다 조기물리치료군에 두 배 더 많은 환자들을 할당하고자 한다면 (즉, $k = 2$)
$$N' = \frac{(1+k)^{2}}{4k}N = \frac{(1+2)^{2}}{4\times2}(220) = 247.5 \approx 248$$
즉, 일반치료군으로 $\displaystyle \frac {1}{k+1}N' = \frac{1}{3}(248) = 82.7 \approx 83$명을, 그리고 조기물리치료군으로 $\displaystyle \frac {k}{k+1}N' = \frac{2}{3}(248) = 165.3 \approx 166$명. (목표 등록 환자수는 249명.) 

그림 18.1(18장) : 이 예제에 관한 검정력 곡선(power curve)

**간단한 공식을 사용하는 경우** - 검정력 90%를 만족시키기 위해서는 (첫 번째 가정을 사용하는 경우) 군 당 109명.<br><center> $\displaystyle n = \frac{21}{(Δ)^{2}} = \frac{21}{(0.44)^{2}} = 108.5 \approx 109$ </center>

`예제 2` **카이제곱 검정을 사용해 서로 독립적인 두 집단의 비율을 비교**

**목적** - 땅콩 알레르기 고위험 유아들을 대상으로 땅콩 식이 전략과 식이회피 전략의 효과 비교.

**설계** - 심한 아토피피부염 그리고/또는 계란 알레르기가 있고, 검사 결과 지금 1$\sim$4mm의 팽진(wheal, 진피의 부종에 의해서 피부가 약간 편평하게 부어오르는 피부병) 소견이 있는 4~11개월 사이 유아들을 60개월까지 땅콩을 섭취하도록 하는 군과  섭취하지 않도록 하는 군 중 하나에 무작위배정한 임상시험(RCT).

**표본 수를 계산하기 위해 사용한 주 결과변수** - 60개월 시점에 경구유발시험(oral food challenge)을 시행해서 땅콩 알레르기가 있는 것으로 확인 된 유아들의 비율.

**표본 수에 관한 질문** - 양측 유의수준 5% 하에서, 두 군 간 60개월째 땅콩 알레르기를 보이는 비율 차이가, 임상적으로 중요한 차이라고 생각되는, 30% (땅콩 식이군 50%, 땅콩 식이회피군 20%) 임을 카이제곱 검정으로 평가하고자 할 때, 검정력 80%를 확보하기 위해서는 몇명의 유아들이 필요한가?

**노모그램을 사용하는 경우**:
-  $p_{1} = 0.5$, $p_{2} = 0.2\to$ $\displaystyle \bar{p} = \frac{0.5+0.2}{2} = 0.35$
- 표준화된 차이(표준화된 유효크기)= $\displaystyle \frac{p_{1}-p_{2}}{\sqrt{\bar{p}(1-\bar{p})}} = \frac{0.5-0.2}{\sqrt{0.35(1-0.35)}} = 0.63$
- 표준화된 차이 0.63과 검정력 80%에 해당하는 점을 연결
- 유의수준 5%에 해당하는 표본 수 축은 대략적으로 값 80을 통과($N =80$). 
- 군 당 약 40명의 유아 ($n=40$).

> **참고**  
> - (i) 검정력을 90%로 증가시키면 표본 수는 대략 전체 102명(즉, 각 군당 51명)
> -  (ii) 중도탈락률 10%를 가정하면 표본 수는 (검정력 80% 하에서) 전체 $\displaystyle N = \frac{80}{1-0.1} = 88.9 \approx 90$명, 즉, 군 당 45명으로 상향조정

그림 18.2: 이 예제에 관한 검정력 곡선

**간단한 공식을 사용하는 경우** - 검정력 80%를 확보하기 위해 필요한 군당 표본 수 <br><center> $\displaystyle n = \frac{16}{(Δ)^{2}} = \frac{16}{(0.63)^{2}} = 40.3 \approx 41$ </center><br> 즉, 군 당 41명.

## Ch.37 연구결과의 제시

### 서론

분석결과의 주요 특성들을 효과적으로 요약해야 함

독자들이 적절하고 중요한 정보를 쉽게 파악해서 올바른 결론은 내릴 수 있도록 하기 위해 연구자는 분석 결과들 중 어떤 것을 제시하고 어떤 방식으로 표현할 것인지를 결정

### 수치적 결과

- 적절한 수준의 정확도가 확보된 그림들 만을 제시<br>
- 계산기를 사용해 분석하는 경우, 결과값의 반올림은 계산의 맨 마지막 단계에서만 시행<br>
- 요약값(예; 백분율)을 제시할 때에는 해당 요약값의 기초가 되는 항목들(예; 분모의 수)도 함께 제시<br>
- 각 분석에 사용된 전체 인원 수와 집단 크기를 제시<br>
- 이상값들에 대한 정보 및 이들을 처리한 방법에 대해 제시<br>
- 측정단위를 명시<br>
- 결과 제시의 주된 목적이 모수(예; 평균, 회귀계수 등)에 관한 추정값의 제시이면 해당 추정값의 정밀성도 함께 제시<br>
  - 정밀성의 척도로는 가능한 한 신뢰구간을 제시하되, 상황에 따라 표준오차를 제시<br>
  - 평균 $\pm$ 표준오차(mean $\pm$ SE) 사용은 피함; 평균 $\pm$ 표준오차는 67% 신뢰구간을 의미하는데 마치 95% 신뢰구간인 것처럼 오해 가능<br>
  - 평균(표준오차);mean(SE) 로 표준오차임을 명시하면서 괄호를 이용하는 것이 바람직<br>
- 결과 제시의 주 목적이 관찰값들의 분포를 제시이면 자료의 '흩어진 정도'에 관한 정보도 함께 제시<br>
  - 이상값을 제외한 자료의 범위(일반적으로는 관찰값들의 가운데 95%를 포함하는 자료의 범위)와 최소 및 최대값을 함께 제시<br>
  - 만일 자료가 정규분포를 한다면 이 범위는 근사적으로 표본평균 $\pm$ 19.6 $\times$ 표준편차<br>
  - 평균 $\pm$ 표준편차 만을 제시한 뒤 독자로 하여금 해당 범위를 스스로 평가하도록 함

### 표

- 표 하나에 너무 많은 정보가 들어가지 않도록 주의<br>
- 표의 제목은 가능한 한 간결하고 효과적이며 정확하게 작성<br>
- 표 제목은 해당 표를 통해 전달하고자 하는 주된 내용을 요약해서 작성. 단, 표 제목에는 분석에 사용된 통계방법의 명칭을 가능한 한 사용하지 않음<br>
- 각 행과 열에 대한 내용을 정확하게 설명
- 정보는 행의 방향보다는 열의 방향으로 제시하는 것이 더 효과적이고 표를 작성하기도 쉬움

### 그림

- 가능한 한 단순하게 만들고 불필요한 장식은 달지 않음(예; 원도표를 3차원으로 만들지 않음)<br>
- 그림의 제목은 간결하고 효과적이며 정확하게 작성<br>
- 모든 x-축, y-축, 조각들, 막대들에 대한 설명(label)을 달고, 사용된 모든 기호에 대해서도 설명<br>
- 축의 표현 단위를 과장/변형해서 왜곡된 결과가 제시되지 않도록 함<br>
-  산점도 상에서 동일 위치에 두개 이상의 관찰값이 존재하는 경우에는 구분되는 기호를 사용하는 등의 방법으로 가능한 한 하나만 있는 값들과 서로 구분<br>
- 그림 내에 모든 적절한 정보들이 제시되도록 함(예; 짝진 관찰값들은 직선을 사용해 서로 연결)

### 논문에서 결과 제시

- 논문 내에 분석결과를 기술할 때에는 어떤 분석이 수행되었는지 독자들이 충분히 이해할 수 있도록 모든 정보를 포함<br>
  - 만일 독자들에게 동일한 자료와 동일한 통계분석 소프트웨어가 제공된다면 그들 또한 동일한 결과를 얻을 수 있어야 함<br>
- 연구설계 및 사용된 통계방법과 관련된 모든 내용을 자세하고 완벽하게 제시<br>
  - 연구의 주된 목적(primary objective of the study)에 대한 기술<br>
  - 분석에 사용된 변수들에 대한 요약된 기술통계량 제시<br>
  - 주 분석 및 이차 분석들에 사용된 통계분석 방법들을  각각 충분히 기술(동시에 한꺼번에 나열 X). 필요하다면 다중비교가 몇 회 이루어졌는지도 언급<br>
  - 가설검정에 사용된 유의수준의 크기에 대한 명시<br>
  - 분석에 사용된 통계 소프트웨어의 이름 및 버전 명시<br>
  
- EQUATOR Network(부록 D)은 의학/보건학 관련 연구결과들을 제시하는 방법을 훈련 + 자원을 제공하기 위해 만들어진 기구 (www.equator-network.org) <br>
  - 무작위배정 임상시험, 임상시험계획서 작성, 관찰연구, 진단 정확도 연구, 신뢰도 및 일치도 연구, 체계적 고찰 및 메타분석, 예후모형 개발 및 타당도 평가 등 포함<br>
  - SAMPL 가이드라인은 의생명과학 분야 학술지에서 기본적으로 요구하는 통계분석 결과의 제시 방법들에 대해 설명<br>(www.equator-network.org/reporting-guidelines/sampl/)

#### 모든 통계분석 방법에 공통적으로 적용되는 일반적인 결과 제시 방법

`가설검정 결과의 제시`

- 만일 필요하다면 적절한 형태의 도표(흐름도)도 사용<br>
- 연구가설을 명시<br>
- 사용된 검정방법의 이름을 제시, 단측검정과 양측검정 중 사용한 것을 명시<br>
- 각 검정방법에서 요구되는 가정들(예; 정규성, 등분산성 등)의 타당성에 대해 평가, 자료를 변환 시 명시<br>
- 검정통계량 값과 해당 분포(및 자유도)를 $P-value$와 함께 제시<br>
  - 정확한 $P-value$ 제시; $P<0.05$, 유의수준 별로 구분하여 $*$,$**$,$***$, $P>0.05$인 경우 n.s, NS 사용한 표기 등 X<br>
- 관심효과에 관한 추정값과 표준 오차 또는 신뢰구간 제시(표준오차보다는 신뢰구간 선호)<br>
- 분석결과에 대한 결론, 신뢰구간을 계산한 경우 신뢰구간에 대한 해석 및 내포하고 있는 의미도 설명

`회귀분석 결과의 제시`

단순 선형회귀(27,28장), 다중 선형회귀(29장), 로지스틱 회귀(30장), 포아송 회귀(31장), 콕스의 비례위험 회귀(44장) 및 집락 자료에 관한 회귀방법들(42장)에 해당하는 내용 (자세한 내용은 해당 장들을 참조)

- 필요하다면 적절한 그림을 추가(예; 단순회귀의 경우 적합선이 포함된 산점도(회귀선이 자료의 최소값과 최대값을 벗어나지 않도록 주의))<br>
- 분석에 사용된 종속변수와 설명변수들을 구분해서 명시<br>
- 회귀모형에 대한 기본 가정들을 평가, 필요에 따라 회귀진단 결과도 제시<br>
- 자료를 변환 시 명시, 변환한 이유 기술 <br>
- 이상값과 결측값들의 처리방법 명시<br>
- 범주형 변수는 분석에 사용된 입력값과 가변수들에 대한 내용을 기술, 연속형 변수는 변수명 옆에 자료의 단위를 함께 제시<br>
- 모형의 적합도(goodness of fit)에 관한 정보($R^{2}$나 LRS 값 등) 제시<br>
- (다중회귀 등) 분산분석표(ANOVA table)에서 제공되는 전반적 F-검정(overall F-test)에 관한 결과도 제시<br>
- 다변수 회귀의 경우 설명변수들 간의 다중공선성 존재 여부 및 상호작용 효과 검토 여부에 대해 보고, 최종모형을 구축하기 위한 변수선택법(전진석택, 단계적 선택 등) 언급<br>
- 모형 내 모든 설명변수에 대해서 회귀계수 추정값(유의성 여부 포함)과 해당 계수에 대한 신뢰구간 또는 표준오차를 제시, 회귀계수들에 대한 해석도 실시<br>
  - 로지스틱 회귀, 포아송 회귀, 콕스의 비례위험 회귀 등의 경우 각 회귀계수를 오즈비(odds ratio)나 상대율(relative rate) 또는 위험비(hazard ratio) 등으로 변환해서 신뢰구간과 함께 제시<br>
- 회귀계수들에 대한 가설검정 결과를 (검정통계량과 $P-value$등을 통해) 제시, 각 검정결과에 대한 해석

`복잡한 분석`

- 복잡한 형태의 통계분석 결과를 어떻게 제시할 것인지에 대한 공통적인 기준은 없음<br>
- 기본적으로 연구설계는 포괄적으로 설명되어야 하고, 기본 가정들에 대한 타당성 여부, 적절한 기술통계량, 검정통계량 및 $P-value$ 등의 정보등이 언급되어야 함<br>
- 잘 모르는 사람들을 위해 분석의 내용을 간단히 설명해 주는 것도 좋음(이 경우 반드시 해당 분석법에 대해 자세히 설명한 참고문헌도 함께 제공)
