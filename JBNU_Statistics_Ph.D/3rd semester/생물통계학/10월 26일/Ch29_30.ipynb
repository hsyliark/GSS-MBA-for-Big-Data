## Part 5. 기초적인 자료분석 방법

### Chapter 29. 다중 선형회귀

##### **다중 선형회귀(multiple linear regression)란 무엇인가?**

- **다중 선형회귀**는 **여러 개의 설명변수($x$)** 들이 전체적으로 종속변수($y$) 에 미치는 영향을 파악하는 방법임.

    - $x$들 간에 상호 관련성이 있다면 각각의 $x$를 사용해 단순 선형회귀분석을 시행하는 것보다 동시에 전체 $x$들의 효과를 평가하는 것이 효과적임.
    > 예시) 아이들의 수축기 혈압($y$), 키($x_1$) 회귀분석 시 몸무게($x_2$)를 추가하는 경우
    
- 표본 n개, 변수 k개인 경우 다중 선형회귀방정식은  $Y=a+b_1x_1+b_2x_2+ ... +b_kx_K$  로 표현됨.

    - $x_i$ = i번째 설명변수, 공변량
    - $Y$ = 주어진 변수의 값들에 대한, 발생 가능한 모든 $y$변수 값들의 예측 평균 추정값($y$의 적합값에 대한 추정값)
    - $a$ = 상수(constant), 절편(intercept)에 대한 추정값, 모든 $x=0$일 때 $y=a$
    - $b_1, b_2, ... , b_k$ = 편회귀계수(partial regression coefficient), 각 $x$들의 기울기에 관한 추정값
    > $b_1$은 다른 모든 $x$변수의 값들이 일정하게 유지된 상태에서 $x_1$값만 한 단위 증가할 때 변화하는 $Y$의 평균 증가/감소량  
      기울기 $b_1$은 $x_1$과 $y$간 단순 선형회귀에서 추정된 기울기와는 다름.  
      왜냐하면 단순 선형회귀분석 기울기는 $y$에 대한 $x_1$의 효과를 다른 x들과는 독립적으로 파악한 것이기 때문임.  
      
- 다중 선형회귀분석은 통계 소프트웨어를 사용해 분석해야만 하기 때문에, 각 회귀모수 추정값 공식은 제시하지 않음.

##### **왜 다중 선형회귀분석을 실시하는가?**

- 다중 선형회귀분석을 실시하는 목적
    - 종속변수($y$)와 관련이 있는 주요 설명변수($x$)들을 파악하기 위함.
    - 종속변수($y$)와 관련이 있을 수도 있는 다른 변수들의 효과를 보정한 상태에서, 각 설명변수($x$)들이 종속변수($y$)와 선형적인 관계를 가지고 있는지, 그 크기는 어느 정도 인지를 확인하기 위함.
    - 설명변수($x$)들을 사용해 종속변수($y$)의 값을 최대한 정확히 예측하기 위함.

##### **필요한 가정들**

- 다중 선형회귀분석 필요 가정들은 단순 선형회귀분석에서의 가정과 동일함. $\to$ `x`를 `각 x들`로 대체하면 끝!
- 특히 중요한 가정은 **선형성 가정**, **독립성 가정**으로, 만일 이 가정이 만족되지 않으면 $\to$ 변수 $y$ 그리고/또는 해당 $x$변수들을 변환한 후, 변환된 자료를 가지고 가정검토를 한 후 재분석을 실시함.

##### **범주형 설명변수들**

1. **이진수(binary)변수**인 경우
    - 예시) 성별
    > `남자=0`, `여자=1` 로 입력한 경우  
      $x_1$이 한 단위 증가한다는 의미는 남자에서 여자로 변한다는 의미임.
      $\to$ $x_1$에 관한 기울기 **추정값 $b_1$은** 모든 $x$변수들의 효과가 보정된 상태에서 **남자와 여자 간 $y$의 평균 추정값 차이**를 의미함.

 
      
2. 3개 범주 이상의 반응값으로 이루어진 **명목형 설명변수**
    - 가변수(dummy variable), 지시변수(indicator variable)를 만들어 분석함.
    - 가변수는 k개의 범주로 구성된 명목형 변수일 경우, 이진수 변수 k-1개를 생성한다.
    - 예시) 교육수준(초졸 이하, 중졸, 고졸, 대졸 이상)별 평균 수축기 혈압 비교
    > 교육수준 변수는 4개 범주 변수 $\to$ 기준 범주를 고졸로 선택하고 3개의 가변수를 생성함.         
      
|구분|초졸 이하|중졸|고졸|대졸 이상|
|:---:|:---:|:---:|:---:|:---:|
|O|1|1|-|1|
|X|0|0|-|0|  

>$Y = a+b_1x_1+b_2x_2+b_3x_3$  
      세 개의 가변수가 모두 0인 경우 기준 범주인 '고졸'을 대표하게 됨. $\to$ 회귀모형의 절편 $a$는 고졸인 사람들의 평균 수축기 혈압에 대한 추정값  
      가변수의 회귀계수들은 기준 범주인 '고졸'에 비해 다른 교육수준들의 $Y$값이 평균적으로 얼마나 다른지를 알려줌.  
      
      
3. 설명변수가 3개 또는 그 이상의 범주로 이루어진 **순서형 변수**
    - 범주들 간 의미있는 선형적인 척도를 부여할 수 있다면(예를 들어, 사회경제적 수준 1~5) 연속형 변수로 간주해서 직접 사용함.
    - 명목형 변수로 간주하고 가변수를 생성해 분석할 수 있음. $\to$ 범주들 간의 순서 측면은 무시됨.

##### **공분산분석(analysis of covariance, ANCOVA)**

- 공분산분석은 집단들 간 반응값을 비교하기 위한 방법으로, 집단들을 구분하기 위한 `가변수`를 만든 뒤 이를 다중 선형회귀분석 모형에 포함시켜 분석하는 방법임.
- 예시) 두 처리집단 A, B에 대한 공분산분석
    > 다른 변수 $x_2, x_3, ... , x_k$가 있을 때,  
    $x_1$을 처리집단에 대한 가변수로 놓고 $x_1=1$이면 처리 A, $x_1=0$이면 처리 B로 설정함.  
    이 때 $x_1$에 해당하는 계수 $b_1$은 다른 $x$값들의 효과가 보정된 상태에서 `처리 B` 대비 `처리 A`의 `y에 관한 평균 반응값 차이`를 나타냄.
- 공분산분석은 처리군들 간에 각 연구대상자의 기저값과 처리후 추적조사한 값 간의 차이를 비교하는 `무작위배정 임상시험`에서 많이 사용됨.
    - $y$ = 처리 후 추적조사된 측정값(연속형 변수)
    - $x_1$ = 처리군 여부를 나태는 이진수 범주형 변수(가변수)
    - $x_2$ = 각 환자의 기저값(연속형 변수)
- 이러한 공분산분석은 기저값 대비 추적조사 값의 변화량이나 변화율을 반응변수로 사용하여 분석하는 방법들보다 효과적인(검정력이 높음) 분석방법임.

##### **설명변수들의 선택**

- 다중 선형회귀분석을 실시하기 위해서는 대략적으로 관찰값의 수가 변수 수의 10배 이상이 되어야 함. $\to$ n>10k
- 대부분의 통계 소프트웨어는 단계적 선택(stepwise selection)과 같은 통계적으로 유의한 변수들을 자동적으로 선택해주는 방법들을 제공함.
- 위와 같은 변수선택 방법은 다수의 설명변수들이 상호연관 되어 있을 때 효과적이나,
- 설명변수들 간에 `공선성(collinearity)`=다중공선성(multicollinearity)이 존재할 때 문제가 발생할 수 있음.
    - 공선성 : 설명변수들 간에 아주 강한 상관관계가 나타나는 문제

##### **분석**

1. 적합도(goodness of fit) 평가
    - 적합도는 다중 선형회귀방정식을 예측하기 위한 목적으로 사용하는 경우 특히 중요한 척도임.
    - 수정된(adjusted) $R^2$은 x들과의 관계로 설명할 수 있는 y의 변동량 비율(%)에 대한 값임.
    - 수정된 $R^2$ 값이 너무 낮은 경우 해당 모형의 적합도는 낮다고 판단함.


2. 분산분석표의 F-검정
    - $H_0$ : $\beta_1$, $\beta_2$, ... , $\beta_k$ = 0을 검정해주는 방법으로,
    - 분석결과가 유의하다는 것(귀무가설을 기각한다는 것)은 회귀모형 내에 y와 선형적인 관계를 가지고 있는 x가 최소한 하나는 존재한다는 의미임.


3. 각 편회귀계수 $\beta_i$$(i=1,2,...,k)$에 대한 t-검정
    - 해당 설명변수가 다른 x들의 효과를 보정한 상태에서 y에 유의한 영향을 주는 변수인지 평가하는데 사용됨.
    - $H_0$ : $\beta_1$ = 0 을 검정한다고 할 때,
    - 검정통계량은 $\frac{b_i}{SE(b_i)}$이며 이 검정통계량은 자유도=$(n$-설명변수의 수-1)인 t-분포를 따름. $\to$ 왈드 검정(Wald test)
    - 신뢰구간은 $b_i±t_0.05\times SE(b_i)$임.
    - 모든 통계 소프트웨어 선형회귀분석 결과에는 $b_i, SE(b_i)$, 검정통계량 값, p-value, $\beta_i$에 대한 95% 신뢰구간 등이 제시됨.

##### **이상값 및 영향값**

- 이상값 : 자료 내 대부분의 다른 값들과는 다른 특성을 보이는 관찰값
- 영향값 : 해당 값을 제거하는 경우 모수에 대한 추정값이 변하게 되는 관찰값  
[참고] 모든 이상값이 항상 영향값이 되는 것은 아님.

- 이상값과 영향값 특성 (적어도 하나는 가지고 있음)
    - 매우 큰 잔차(residual)을 보임. $\to$ `스튜던트화 잔차`나 `스튜던트화 제외 잔차`의 값이 ±3 혹은 ±4보다 클 때
        - 잔차란? $y-yhat$
    - 매우 높은 지렛대(leverage)값을 보임.
        - 지렛대란? 개인별 x(들)값이 자료 내 x(들)의 평균으로부터 얼마나 떨어져 있는지를 평가하는 척도
        - 선형회귀모형에 사용된 x의 수=k, 관찰값의 수=n일 때, 분석결과로 통계 소프트웨어에서 제공되는 **지렛대 값이 $2(k+1)/n$보다 크면** 지렛대 값은 크다고 간주하고, 이에 해당하는 관찰값은 **영향값**으로 봄.


- 영향값 탐색 방법
    - `잔차가 크고`, `지렛대 값도 크고`, 그리고/또는 `쿡의 거리(Cook's distance)>1` 이거나 `쿡의 거리>4/n`, `다른 관찰값들에 비해 굉장히 큰 관찰값` 탐색     
    - 영향값들을 시각적으로 확인할 수 있도록 개발된 진단 그림(diagnostic plots) 검토

### Chapter 30. 두 범주 결과변수와 로지스틱 회귀

##### **서론**

- 로지스틱 회귀란? 관심변수가 두 범주 결과변수(binary outcome variable)일 때 진행하는 회귀분석
    - 예시) 증상의 유무, 질병의 유무, 사망여부 등
- 로지스틱 회귀의 목적
    - 결과에 영향을 미치는 설명변수 파악하기.
    - 특정 공변량(설명변수 값들의 조합)을 가진 사람에게 결과가 발생할 확률을 계산하기.
    - 확률을 통해 (1)결과가 발생할 위험이 높은 집단, (2)결과가 발생할 위험이 낮은 집단으로 분류하기.  
      분류기준 값(cut-off value)는 주로 0.5를 사용함.  
    - 짝지지 않은 사례-대조 연구를 분석하는데 사용되며, 결과변수는 사례군(case), 대조군(control)임.

##### **이론**

- 선형회귀분석으로 `종속변수가 두 범주 변수인 경우`를 분석할 수 없는 이유
    -  정규성 가정을 만족하지 않음.
    - 선형회귀식으로 추정되는 예측값이 0,1 이외 다른 값인 경우 해당 값을 해석할 수 없음.
- 위에서 2번과 같은 이유로 로지스틱회귀분석은 
    - 예시) 질병이 있음 = 1, 질병이 없음 = 0 일 때  
  질병이 있음 =1(두개의 값 중 큰 값)으로 분류될 확률 `p`를 사용함.
- 이 때 수학적으로 발생하는 제한점을 피하기 위해 p값 자체가 아닌, 로지스틱 혹은 로짓변환한 값을 종속변수로 사용함.  
  p에 대한 로짓변환은 오즈에 대한 자연로그 변환임.  
  $\mbox{logit}(p)$ = $\ln$$\frac{p}{1-p}$

##### **로지스틱 회귀방정식(logistic regression equation)**

- 로지스틱 회귀방정식

$\mbox{logit}(p)$ = $\alpha$ + $b_1x_1 + b_2x_2 + ... + b_kx_k$

$\log(\rm{odds})$ = $\alpha$ + $b_1x_1 + b_2x_2 + ... + b_kx_k$

- 방정식의 계수는 최소제곱법(OLS)를 사용하여 추정할 수 없고, **최대가능도(maximum likelihodd)** 를 사용해서 계산함.  
    - $x_i$ = i번째 설명변수 $(i=1,2,3,...,k)$  
    - p = 특정 $x_1, x_2, ... , x_k$값을 가진 사람이 질병에 걸리게 될 실제 확률에 대한 추정값, 이항분포에 따르는 것으로 가정함.
    - $\alpha$ = 상수항(절편)에 대한 추정값
    - $b_1x_1 + b_2x_2 + ... + b_kx_k$ = 로지스틱 회귀계수(logistic regression coefficient)에 대한 추정값


- 각 계수의 지수형태 **$e^b$** 는 **오즈비에 대한 추정값**이 되며, 다른 x값들의 효과가 보정된 상태에서 한 개인이 $x_1$인 경우 질병에 걸릴 오즈의 추정값보다 $x_1 + 1$인 경우에 질병에 걸릴 오즈의 추정값이 상대적으로 얼마나 높은지 알려주는 값임.
    - 예시) 한 사람이 나이가 70살($x_1$)일 때 질병에 걸릴 오즈의 추정값보다 나이가 71살($x_1 + 1$)인 경우의 질병에 걸릴 오즈의 추정값이 상대적으로 $e^b$만큼 높다.


- 오즈비에 대한 추정값은 보정(조정)된 오즈비(adjusted OR)라고 함.
    - OR = 1 이면 $x_1$이 한 단위 증가해도 질병에 걸릴 오즈에 아무런 영향을 주지 않음.
    - OR > 1 이면 $x_1$이 한 단위 증가하면 질병에 걸릴 오즈도 증가함.
    - OR < 1 이면 $x_1$이 한 단위 증가하면 질병에 걸릴 오즈는 감소함.
    - *희귀질병*의 경우 오즈비를 상대위험(relative risk, 위험비, risk ratio)로 간주하고 해석할 수 있음.
    - 통계 소프트웨어에 따라 p=0을 모형화 하는 경우 오즈비는 1/OR이 되므로 주의하자.


- 각 개인이 질병에 걸리게 될 확률을 추정하면,  
$\displaystyle p = \frac{e^(\alpha +b_1x_1 + b_2x_2 + ... + b_kx_k)}{1+e^(\alpha+b_1x_1 + b_2x_2 + ... + b_kx_k)}$  
  각 공변량 별로 관찰값들과 확률들 간의 산점도를 그리면 결과 해석에 도움이 됨.


- 로지스틱 회귀모형은 설명변수들을 관심 결과변수 값이 관찰될 오즈에 관한 로그 척도와 적합하는 것이기 때문에 **승법적(multiplicative)효과**를 갖게 됨. $\to$ 각 $x_i$의 개별 효과들의 곱이 $x_i$들의 결합효과가 됨.
    - $\log(\rm{odds})$ = $\alpha$ + $b_1x_1 + b_2x_2$ 일 때  
    질병이 관찰될 오즈의 추정값은 각각 $OR_1$ = $\exp(b_1)$, $OR_2$ = $\exp(b_2)$가 됨.
    $x_1$, $x_2$ 모두 0인 사람에 비해, $x_1$, $x_2$ 모두 1인 사람이 몇 배 더 질병에 걸릴 오즈가 높은지 추정하려면, $OR_1 \times OR_2$를 하면 됨. $\to$ 오즈 척도에서의 승법적 효과

##### **설명변수**

- 로지스틱 회귀분석 결과로 (1)로지스틱 회귀계수 추정값, (2)해당 표준오차, (3)오즈비에 대한 추정값(회귀계수의 지수형태), (4)모집단 오즈비에 대한 95% 신뢰구간 등이 제공됨.
- 이 때 귀무가설은 $H_0 : \beta_i = 0$(설명변수와 관심 결과변수는 관련이 없다)이며, 이를 기각할 수 있는지 검정함.
- 위에서 말한 것은 또 $H_0 : OR_i = 1$과 동일함.


- 이를 위해서는 아래와 같은 두 방법 중 하나를 사용함  
    1. 왈드 검정(Wald test)
        - 검정통계량 = $\displaystyle \frac{로지스틱 회귀계수 추정량(\beta_i)}{해당 표준오차}$
        - 검정통계량은 표준정규분포를 따르며, 이 값을 제곱하면 근사적으로 자유도가 1인 카이제곱 분포를 따름.
    2. 가능도비(우도비)검정(likelihood ratio test)
        - 검정통계량 : 완전모형(설명변수를 모두 고려한 모형)의 이탈도와 축소모형(검정 대상 설명변수를 제외한 모형)의 이탈도 간 차이
        - 이탈도를 가능도비 통계량(likelihood ratio statistic, LRS), -2로그-가능도(-2 log likelihood)라고 함.
        - 검정통계량은 자유도가 1인 카이제곱 분포를 따름.
        * 이탈도 : 모형이 자료를 설명하지 못하는 크기에 관한 척도(deviance)
- 표본 수가 크면 두 검정방법은 비슷한 결과를 제공함.
- 왈드 검정이 가능도비 검정 대비 검정력이 낮고, 자료 내 해당 설명변수의 각 값 별 관찰값 수가 충분하지 않으면 편향된 추정결과를 제공하기도 하지만, 대부분의 통계 소프트웨어들이 기본 설정으로 왈드 검정 결과를 제공함.
- 다중 선형회귀분석의 경우처럼 자동 변수선택 절차를 사용할 수 있으며, 다중 로지스틱 회귀분석 실시를 위해 결과변수의 두 범주 중 적은 빈도를 가지는 범주의 관찰값 개수가 모형에 사용되는 설명변수 수의 10배 이상은 돼야 함.

##### **모형의 적절성(adequacy of the model)**

- 모형의 적합도(goodness of fit) 평가  
    1. 이탈도(D, 혹은 LRS, -2 로그가능도)
        - 제시되는 값 그 자체의 이탈도는 포화모형의 가능도에서 k개의 공변량을 가진 모형의 가능도를 뺀 값
        - 이 검정통계량은 근사적으로 자유도 ($n-k-1$)인 카이제곱 분포를 따름.
        - 검정결과가 유의하다는 것은 해당 모형이 자료를 잘 적합시키지 못했다는 의미임. $\to$ 이탈도는 모형부적합도에 관한 척도
    2. 모형 카이제곱(model chi-square), 공변량들의 카이제곱, G
        - 아무런 설명변수를 사용하지 않은 영 모형(null model)의 이탈도에서 k개 설명변수를 가진 이탈도를 뺀 값임.
        - 이 때 귀무가설은 $H_0 : \beta_1 = \beta_2 = ... = \beta_k = 0$
        - G값은 근사적으로 자유도가 k인 카이제곱 분포를 따름.
        - 검정결과가 유의하다는 것은 k개 설명변수 중 적어도 하나는 종속변수와 유의하다는 것을 의미함.
    3. 호스머-르머쇼우 검정(Hosmer-Lemershow test) 
        - 표본 수 n이 충분히 큰 경우(n>400)에만 사용 가능함. (46장 참조)
        - 검정결과가 유의하다는 것은 해당 모형이 자료를 잘 적합시키지 못했다는 의미임.
    4. 참고
        - 로지스틱 회귀분석에서도 $R^2_L$(Cox-Snell $R^2$), pseudo-$R^2$, max-rescaled $R^2$ 등과 같은 적합도에 관한 여러 지표가 있으나 해석이 쉽지 않아 잘 사용하지 않음.

##### **오즈비와 상대위험의 비교**

- 희귀질병인 경우 오즈비(OR)과 상대위험(RR)이 서로 비슷하게 되기 때문에 오즈비를 상대위험으로 간주해 해석할 수 있음
- 일반적으로는, RR>1 일 때 오즈비>상대위험, RR<1 일 때 오즈비<상대위험임.
- 오즈비는 상대위험보다 해석하기 어렵지만 매력적인 특징들이 많아 더 선호되며, 사례대조 연구의 경우 상대 위험을 직접 추정하는 것이 불가능하기 때문에 반드시 오즈비를 사용해야 함.

##### **다항 로지스틱 회귀(multinomial logistic regression) 및 순서형 로지스틱 회귀(ordinal logistic regression)**

- 세 개 이상의 범주를 가지는 다범주 범주형 종속변수에 대해 사용하는 방법으로는 다항 로지스틱 회귀와 순서형 로지스틱 회귀가 있음.
- 다항 로지스틱 회귀는 순서, 순위 없이 3개 이상의 범주형 종속변수를로 이루어진 경우 사용하며,
- 순서형 로지스틱 회귀는 종속변수가 순서형 또는 순위형 척도로 이루어졌을 때 사용함.

##### **조건부 로지스틱 회귀(conditional logistic regression)**

- 짝진 자료(짝진 사례-대조 연구 등)를 분석할 때 조건부 로지스틱 회귀를 사용함.
- 조건부 로지스틱 회귀는 짝지어진 동일한 집합 내에서 대조군과 사례군을 비교하는 방법임.
- 대부분 콕스의 비례위험 회귀모형(Cox's proportional hazard regression model)을 사용해서 분석함(44장)

##### **예제**

- 관심변수(종속변수) : 인간 허피스 바이러스 제 8형(HHV-8) 감염여부
- 표본 : 동성애/양성애자 남성 271명
- 표본을 대상으로 성매개 감염병(임질, 매독, 제2형 단순 허피스 바이러스, 인간면역결핍바이러스)을 알았던 적이 있는지 설문조사 실시  
$\to$ 설명변수 : 임질, 매독, 단순허피스바이러스, 인간면역결핍바이러스, 연령
- 24장에서 임질 과거력을 가진 남성은 그렇지 않은 남성보다 더 높은 혈청 내 HHV-8양성률을 보인다고 조사됨.


- 다변수 로지스틱 회귀분석 실시
    - $H_0 : b_i = 0$
    - 각 성매개 감염병 과거력 유무를 나타내는 4개 가변수를 만들어 분석
    - 회귀분석 결과 p=0.002로 적어도 하나의 설명변수와 종속변수가 관련이 있다는 결과가 도출됨.


- 분석 결과 표를 보면 HSV-2 양성 여부와 HIV상태는 p-value<0.05로 HHV-8과 관련이 있는 것으로 확인됨.
    - HSV-2가 양성인 사람은 음성인 사람에 비해 HHV-8에 감염될 확률이 $\exp(\mbox{HSV-2의 회귀계수})$ = $e^0.7910$ = 2.21배 높음.  
      HSV-2가 양성인 사람들은 HHV-8이 양성일 확률이 2.21-1.00 = 1.21 $\to$ 121% 증가함.
    - 승법적 효과를 감안하면, 다른 설명변수가 고정된 상태에서 HSV-2=양성(=1), HIV=양성(=1)인 남성은 둘 다 음성인 남성 대비 HHV-8에 감염될 오즈가  
      $2.206\times 5.133=11.3$배 더 높음.
- 매독은 p=0.0935로 0.05보다 크나, 신뢰구간을 보면 오즈비가 13.28까지 올라갈 수 있어서 어느정도 관련이 있는 것으로 해석할 수 있음.
- 이전 24장에서 카이제곱 검정 결과 임질력과 HHV-8 간 관련성이 유의하다고 나타났으나, 해당 회귀분석 결과에서는 미미한 관계성을 보임.
    - 아마 임질을 제외한 다른 성매개 감염병 과거력을 가진 남자의 대부분이 임질력도 있었기 때문으로 해석할 수 있음.


- HHV-8에 걸릴 확률을 계산해보면,
    - 임질 과거력=있음, HSV-2=양성, 매독 과거력=없음, HIV=음성, 연령=51세 인 남성의 경우를 계산하면  
      $\frac{\exp(-2.2242 + 0.5093*1 + 1.1924*0 + 0.7910*1 + 1.6357*0 + 0.0062*51)}{1+\exp(-2.2242 + 0.5093*1 + 1.1924*0 + 0.7910*1 + 1.6357*0 + 0.0062*51)}$ = $0.35$
