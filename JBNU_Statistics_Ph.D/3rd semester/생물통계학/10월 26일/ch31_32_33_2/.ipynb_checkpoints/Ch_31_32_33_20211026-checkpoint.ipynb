# Part 5 기초적인 자료분석 방법

## Chapter 31 율과 포아송 회귀

### 율(rate)

`-` 사건발생 여부를 조사하는 종적 연구(16장, 사례-대조 연구)에서 연구 도중에 탈락하거나 연구 참여 시점이 달라 연구종료 시점까지의 추적기간이 사람 별로 각각 다름.  분석시 이 점을 반영 필요.

`-` 각 시간당 사람당(per person per period of time) 사건 발생 정도를 나타내는 **율** 사용
- 편의상 연(year) 주로 사용
- 율(rate) = $\frac{\mbox{관심사건수}}{\mbox{모든 사람들의 전체 추적 연도 수}}$ = $\frac{\mbox{number of events}}{\mbox{follow-up person-year}}$

`-` 추적기간
- 사건이 발생한 사람: 연구 참여일 ~ 사건 발생일 간 시간
- 사건이 발생하지 않은 사람: 연구 참여일 ~ 연구 종료일 간 시간

`-` 전체 추적기간: 모든 사람의 추적기간들의 합

`-` 새로운 케이스(예를 들면 질병)가 관심사건이면 **발생률(incidence rate)**

`-` 사망이 관심사건이면 **사망률(mortality rate)**.

`-` 율의 특성
- 동일인으로부터 조사된 추적 인-년과 서로 다른 사람들로부터 조사된 추적 인-년 구분하지 않음
  - (ex. 100명을 각각 1년씩 조사, 1명을 100년 조사 모두 100 인-년에 해당한다.)

- 사건이 한 번 이상 발생하는 반복사건(repeated event)의 처리방법
  - 연구의 가설에 따라 달라짐.
  - 첫 번째 사건에만 관심이 있다면 추적기간은 첫 번째 사건이 발생할 때까지의 시간으로 정의.
  - 여러 사건을 관심 사건으로 간주해서 율을 계산하고자 하는 경우(**집락 자료(clustered data, 41장)**) 적절한 통계분석 방법 사용(41,42장).

`-` 율(rate)과 위험(risk)의 차이
- 어떤 사건의 **위험**이란, 사건의 총 수를 추적조사 시작시점에서 연구에 참여한 전체 인원 수로 나눈 값.  추적기간 미고려.
- 추적기간이 길어지면 이들이 해당 사건을 경험하게 될 가능성이 늘어나므로 사건 발생위험 또한 더 커짐.
- **율**은 추적기간을 반영하여 계산하므로 위험에 비해 안정된 값이 얻어짐.

`-` 상대율(relative rate)

$$상대율(relative~rate) = \frac{rate_{exposed}}{rate_{unexposed}} = \frac{노출~집단~질병~발생률}{비노출~집단~질병~발생률}$$

- **상대율**은 **율비(rate ratio, $PR$)** 또는 **발생률비(incidence rate ratio, IRR)**.
  - 상대율 = 1: 두 집단 간 질병 발생률이 같다.
  - 상대율 > 1: 관심요인에 노출되지 않은 집단에 비해 노출된 집단의 질병 발생률이 상대적으로 높다
  - 상대율 < 1: 관심요인에 노출되지 않은 집단에 비해 노출된 집단의 질병 발생률이 상대적으로 낮다
- 상대위험을 상대율에 대한 추정값으로 사용하려면 관심사건이 희귀한 경우에만 사용
    - 희귀질병의 경우에만 상대위험과 상대율의 값이 서로 비슷하게 됨
- 희귀사건이 아니고, 추적기간도 상이하다면, 율은 영향을 받지 않겠지만 위험은 추적기간에 직접적인 영향을 받음.

### 포아송 회귀(Poisson regression)

`-` 포아송 회귀란?
- 포아송 분포 :특정 기간 동안 일정 크기의 평균율 하 무작위로 발생하게 되는 희귀사건의 계수(count) 값들에 대한 확률 분포.
- > 포아송 회귀 : 포아송 분포에 기초하여 서로 다른 추적기간을 가진 사람들에게 특정 사건이 발생하는 율을 추정
- > 로지스틱 회귀  :  사건 여부에 관한 오즈비 추정, 
- 포아송 회귀의 가정 :  전체 연구기간 동안 동일한 설명변수(연령, 성별 등) 값을 가지는 사람들의 사건 율(event rate)이 일정하게 유지됨.
- 포아송 회귀의 목적
  - 사건 율에 영향을 미치는 설명변수들은 무엇인가?
  - 노출 집단은 비노출 집단에 비해 사건 율이 얼마나 높은가?
  - 특정 집단의 사건 율은 어느 정도인가?

`-` 방정식과 해석방법
- 포아송 회귀모형과 로지스틱 회귀모형의 식은 매우 유사하며, 두 모형 모두 설명변수들의 선형결합 형태를 방정식의 오른쪽에 표시.
- 변환 방법으로는 자연로그 변환(natural logarithmic transformation: $ln$, 즉, $log_e$)을 사용.
- 포아송 회귀방정식 
$$\ln(r)=a+b_1x_1+b_2x_2+...+b_kx_k$$
반복계산절차(최대가능도, 32장)를 사용해 추정.
  - $x_i$ : $i$번째 설명변수. ($i=1,2,...,k$);
  - $r$ : 특정 $x_1, x_2, ..., x_k$값들을 가진 사람의 평균율(mean rate) 혹은 기대율(expected rate)에 관한 추정값.
  - $a$ 상수항(절편) :  회귀식 내 모든 $x_i$값들이 0인 경우에 해당하는 로그 율(log rate)의 추정값.
  - $b_1,b_2,...,b_k$ : **포아송 회귀계수(Poisson regression coefficient)**의 추정값.
- 이 계수들의 지수 값, 예를 들어 $e^{b_1}$은 변수 $x_1$과 관련된 **상대율(relative rate)** 추정 값.
  - $e^{b_1}$ :  회귀식 내 모든 다른 변수 $x_i$들의 효과가 보정된 상태에서, $x_1$값 대비 ($x_1+1$)값의 상대적인 질병 발생률 추정값 의미.
  - 상대율 = 1: $x_1$이 한 단위 증가해도 사건의 율은 일정하다는 의미
  - 상대율 > 1: $x_1$이 한 단위 증가 할 때마다 사건의 율도 증가
  - 상대율 < 1: $x_1$이 한 단위 증가 할 때마다 사건의 율이 감소
- 로지스틱 회귀모형은 질병이 발생할 오즈를 설명변수들과 로그 척도로 결합하고,오즈 척도에서의 승법적 효과를 가짐.
- 포아송 회귀모형은 질병이 발생한 율을 설명변수들과 로그 척도로 결합하고, 질병의 율 척도에서의 승법적 효과를 가짐
- 포아송 회귀방정식을 사용해 $x_1,x_2,...,x_k$값들 조합에 따른 사건 율 추정.
  - 공변량 값들이 $x_1,x_2,...,x_k$인 사람의 사건 발생률 추정값
$$r = e^{a+b_1x_1+b_2x_2+...+b_kx_k}$$

`-` 차감(offset)의 사용
- 율(rate) = $\frac{\mbox{사건발생수}}{\mbox{추적 인-년}}=r= \frac{k}{n}\to \ln(r)=\ln(y)-\ln(n)$ 

- 대부분의 통계 소프트웨어들은 종속변수로 $r$이 아닌 $y$를 지정한 후, $\ln()$ 함수를 사용해서 종속변수 $y$와 설명변수들을 연결. 

- $\ln(n)$부분은 offset(차감)으로 따로 지정

`-` 집단 자료(group data)의 입력
- 설명변수들이 모두 범주형인 경우 율을 계산할 때 동일인의 추적 인-년과 다른 사람의 추적 인-년을 구분할 필요가 없다는 점을 활용해 자료입력 방법 단순화 가능.

- 해당 방식은 집단 정의에 연속형 공변량이 사용되거나, 각 집단 내에 개인 별로 각기 다른 값을 가지는 공변량이 있는 경우 사용할 수 없음.

`-` 시간에 따라 변하는 변수의 사용
- 추적기간을 길이가 짧은 구간들로 세분화하면 **시간에 따라 변하는** 설명변수(즉, **시간-의존적 공변량, time-dependent covariate**)를 포아송 회귀모형에 포함 가능.
  > 예: 흡연력이 중년 남성의 폐암 발생률에 미치는 영향을 알아보기 위한 연구의 경우
  >  - 추적기간이 길어지면 길어질수록 금연을 하는 사람 수가 증가, 폐암 발생률은 감소.
  >  - 연구대상자들을 연구시작 시점의 흡연 상태 자료를 사용해 범주화 하면 흡연력이 폐암에 끼치는 영향력을 제대로 평가할 수 없음.
 > - 이런 경우, 본인의 흡연 상태가 유지될 수 있을 정도의 짧은 구간들로 세분화해서 세분화된 구간 내 정보(사건 발생 유/무, 추적기간, 흡연상태) 등을 수집한 후, 이 정보들을 각 구간이 모두 서로 다른 사람들의 구간인 것처럼 간주하고 포아송 회귀 분석을 실시.

`-` 통계 소프트웨어 분석결과
- 제공되는 포아송 회귀분석에 관한 컴퓨터 분석결과
  - 설명변수 별 포아송 회귀계수의 추정값 및 해당 표준오차
  - 상대율 추정값 및 해당 신뢰구간
  - 왈드 검정통계량 값(분석 대상 '질병'의 모집단 상대율이 1이라는 귀무가설을 검정하기 위한 값) 및 $p$-value
- 로지스틱 회귀분석과 마찬가지로 포아송 회귀분석 역시 -2로그-가능도(LRS 또는 이탈도)와 모형 카이제곱 또는 공변량들에 대한 카이제곱 등을 사용해 모형의 적절성 여부 평가

`-` 초과-포아송 변동(extra-Poisson variation)
- 포아송 회귀모형 적합 시 고려해야 할 사항 (과산포)
  - **과산포(over-dispersion)** : 초과-포아송 변동이 있는지 여부.
    - 모형적합을 통해 계산되는 잔차 분산이 포아송 회귀모형에서 기대되는 값보다 더 크게 나타나는 현상.
    - 주된 원인: 이상값, 중요한 설명변수 미포함, 집락 자료의 형태의 자료인데도 분석시 미반영
  - 과산포가 발생한다면,
    - 추정값들의 표준오차는 실제보다 낮게 추정되게 되고,
    - 이로 인해 모수에 대한 신뢰구간이 지나치게 좁아지게 되며,
    - 각 설명변수의 유의성 여부에 관한 $P$-value 역시 작아짐
  - 과산포 존재 여부는
    - -2로그-가능도를 $n-k-1$로 나눈 값으로 확인.
      - $n$= 자료 내 관찰값의 개수,  $k$= 모형 내 적합된 모수의 개수
    - 과산포가 발생하지 않으면 이 값은 1에 가까움
    - 1보다 상당히 크면 자료에 과산포가 존재하는 것으로 판단
  - 과산포 발생한 것으로 확인 되면,
    - **척도 모수(scale parameter)**를 사용한 과산포 포아송 회귀모형(over-dispersed Poisson regression) 적합 또는
    - 포아송 분포 대신 음이항 분포(negative binomial distribution)를 가정하는 음이항 회귀모형(negative binomial regression)을 적합.
    >  척도 모수: 일반적으로 포아송 회귀분석은 과산포가 없는 자료에 대해 척도 모수 1을 가정.

- 포아송 회귀모형 적합 시 고려해야 할 사항 (저산포)
  - **저산포(under-dispersion)** : 잔차 분산이 포아송 모형에서 기대되는 값보다 작은 현상.
  - -2로그-가능도를 $n-k-1$로 나눴을때, 1보다 작으면 저산포 자료임을 의심.
  - 자료 수집 시 큰 자료 값이 누락된 경우 등에 발생.
-  과산포나 저산포의 문제는 로지스틱 회귀를 실시하는 경우에도 발생 할 수 있다(초과-이항 변동(extra-binomial variation).).

`-` 포아송 회귀분석의 대안
- 자료가 어떤 '시작점'(예를 들면 수술)에서부터 관심 결과변수가 관찰되는 시점까지 추적조사한 자료이면 대안으로 **생존분석(survival analysis)** 실시.
- 포아송 회귀분석은 사건율이 전체 연구기간에 걸쳐 일정하다는 가정 필요,
- 생존분석은 추적기간 동안 '위험(hazard)'이 일정하다는 가정 필요 없음(44장).

#### 예제

`-` 바이러스학적 치료실패율과 관련된 요인들에 대한 다변수 포아송 회귀분석
- 치료 구간을 나눈 이유: 바이러스학적 치료실패율이 치료기간 별로 변하는지를 조사하기 위해
- 결과변수: 바이러스학적 치료실패 사건의 율
- 연구의 주된 목적: 이들 세 구간 간에 치료 실패율이 다른지를 평가하고자 함

`-` 단변수 포아송 회귀분석 (부록 C_p.315, 표 31.1_p.177)
-  총 추적기간은 718 인-년
    - $\frac{\mbox{1~2년 범주}}{\mbox{1년 미만 범주}}$, $\frac{\mbox{>2년 범주}}{\mbox{1년 미만 범주}}$의 바이러스학적 실패율을 비교하기 위해 가변수 두 개를 사용하였고, 범주들간에 통계적 유의성이 있는 것으로 나타났다(*P*=0.04).
    - 3개 변수가 바이러스학적 치료 실패율과 유의한 관련이 있는 것으로 나타남.
        - 기타설명변수들 중 환자의 성별(*P*=0.033),
        - 기저시점에서의 CD8 세포 수(*P*=0.03),
        - 동일한 치료를 과거에 받아본 적이 있는지에 관한 치료 과거력 여부 ($\frac{\mbox{과거에 치료를 받아본 적이 있음}}{\mbox{없음}}$*P*=0.0008)   
    - 세 변수 효과를 보정한 상태에서 치료기간과 바이러스학적 실패율 간의 관련성 여부를 평가하기 위한 다변수 포아송 회귀분석 실시

`-` 다변수 포아송 회귀분석
- 바이러스학적 실패율은 초기 반응 이후 첫 해에 비해 증가하는 경향(1.0 $\to$ 1.53, 53% $\uparrow$), 2년 이후부터는 오히려 낮아짐(1.53 $to$ 0.56,44% $\downarrow$). 둘 다 통계적으로 유의한 효과 크기는 아님(*P*=0.1152,*P*=0.2265).
- 모형 내 모든 다른 변수들의 효과가 보정된 상태에서 해당 치료를 처음 받은 환자들은 기존에 치료를 받은 적이 있는 환자들에 비해 통계적으로 유의하게 44%정도 낮은 바이러스학적 실패율을 보이는 것으로 추정($P$=0.0233)
- 남성이 여성에 비해 39% 낮은 실패율을 보였으나 통계적인 유의성은 확보되지 않음($P$=0.0676)
    - 기저 시점에서의 CD8세포 수가 100$cells/mm^3$ 증가할 때마다 바이러스학적 치료실패율은 5%씩 유의하게 낮아짐($P$=0.0369)
- 포아송 회귀모형의 율 척도에서의 승법적 효과 감안
    - 바이러스 부하 초기 반응 후 결과 기간과 CD8 세포 수가 동일한 환자의 경우, 과거에 치료를 받은 적이 없는 남성은 치료를 받은 경험이 있는 여성에 비해 바이러스학적 치료 실패율이 68% 정도 낮다고 할 수 있음.

## Chapter 32 일반화 선형모형

`-` **통계적 모형화(statistical modelling)**란, 단순 또는 선형회귀, 로지스틱 회귀, 포아송 회귀, 생존자료에 관한 콕스 회귀(44장) 등을 사용하는 방법.
- 모두 관심 결과변수와 하나 이상의 설명변수들 간의 관계를 가장 잘 설명해주는 **수학적 모형(mathematical model)**.
  - 목적
      - 관심 설명변수와 결과변수 간의 관련성 크기를 다른 설명변수들의 효과를 보정한 상태에서 파악 
      - 설명변수들을 사용해 결과변수의 값 예측
- **일반화 선형모형(generalized linear model, GLM)**
$$g(Y)=a+b_1x_1+b_2x_2+...+b_kx_k$$
  - $Y$: 알려진 확률분포에 따르는 종속변수의 예측값, 평균 또는 기대값에 대한 추정값
  - $g(Y)$: **연결함수(link function)**라 하며, 역할은 $Y$를 $x_1,x_2,...,x_k$들과 선형적인 관계를 가지도록 변환.
  - $b_1, b_2, ..., b_k$: 설명변수들과 관련된 회귀계수의 추정값
  - $a$: 상수항

-  로지스틱 회귀는 비율에 대한 **로짓**(**logit**, 오즈에 대한 $ln$, $log_e$)을 연결함수로 사용하는 모형,**로짓연결(logit link)**
- 포아송 회귀는 연결함수가 율에 대한 $\log$ 변환, **로그 연결(log link)**
- 단순 및 다중회귀는 종속변수에 대한 어떠한 변환도 하지 않음, **항등연결(identity link)**

### 모형 선택의 기준

`-` 어떤 통계 모형이 적절한지는 관심 결과변수의 형태에 따라 결정 (아래 표)

|**결과 변수의 형태**|**일반적으로 사용되는 GLM의 종류**|**참조 장**|
|---|---|---|
|연속형 변수|단순 또는 다중 선형회귀|28. 29|
|두 범주 변수|||
|질병 발생 여부에 관한 추적연구|로지스틱 회귀|30|
|(단, 환자들의 추적기간이 모두 동일함)|||
|두 범주 결과변수를 가진 단면조사 연구|로지스틱 회귀|30|
|짝지지 않은 사례-대조 연구|로지스틱 회귀|30|
|짝진 사례-대조 연구|조건부 로지스틱 회귀|30|
|세 범주 이상의 범주형 결과변수|다항 또는 순서형 로지스틱 회귀|30|
|사건의 율(event rate) 또는 계수(count) 자료|포아송 회귀|31|
|사건이 발생할 때까지의 시간 자료*|지수(exponential), 웨이블(Weibull) 또는 곰페르츠(Gompertz) 모형|44|

$*$ 사건이 발생할 때까지의 시간(time to event) 자료는 콕스의 비례위험모형(Cox's proportional hazard model).을 사용해 분석.(44장)


- 예
    - 종속변수가 연속형 변수일 때, 이 변수와 관련 요인을 찾아내기 위해서는 단순, 다중 선형회귀를 사용,
    - 결과변수가 두 범주 변수이고 모든 환자의 추적관찰 시간이 동일할 때, 로지스틱 회귀를 선택하는 것이 적절.

`-` 종속변수의 형태를 수정 후 다른 모형 사용 가능.
- 연속형 결과변수가 선형회귀에서 필요로 하는 가정들 중 일부를 만족하지 않을 시, 이 결과변수의 값을 두 집단으로 분류하여 두 범주 결과변수로 만듦.
- 종속변수 이분화
    - 장점: 통계적 모형적합 및 해석이 편이.
    - 단점: 종속변수가 가진 정보 중 일부 손실.

### 가능도(likelihood) 및 최대가능도(maximum likelihood estimates, MLE) 추정

`-` GLM 적합 시 모형 내 모수 추정 위해 **가능도(likelihood)** 개념 사용.
- GLM은 종속변수에 대한 확률분포, 설명변수들, 각 회귀계수들이 취할 수 있는 값들에 의해 형태가 결정,
- **모형의 가능도($L$)**: 해당 모형으로부터 계산된 회귀계수 추정값들 하 현재 관찰된 결과들이 얻어질 확률 .
- **최대가능도 추정법(maximum likelihood estimation, MLE)**: 회귀모형 적합 시, 회귀계수의 추정값으로 이 $L$을 최대로 해주는 값, 즉, 관찰된 결과를 얻울 수 있는 가장 그럴 듯한(most likely)값을 추정하는 방법(얻어진 추정값:  **최대가능도 추정값(maximum likelihood estimates)**)
  -  MLE를 계산하기 위해 반복계산(iteration) 실시$\to$통계 소프트웨어 사용
   - MLE를 계산하지 않아도 되는 유일한 예외는 **최소제곱법**으로 모수를 추정하는 (항등연결 함수를 사용하는) 단순 그리고 선형 회귀 모형 뿐.
  - 최소제곱법을 사용해 추정된 값들은 **보통 최소제곱(ordinary least squares, OLS)** 추정값
  - 선형회귀모형의 경우, OLS 추정값과 MLE 추정값이 서로 일치

### 모형의 적절성 평가

`-` MLE는 주어진 설명변수 집합 내 L을 최대로 해주는 추정값이지만, 모형에 설명변수가 추가되면 $L$ 값은 항상 커짐.
- **포화모형(saturated model)**: 가장 극단적인 경우($L$ 값이 가장 큰 경우), 자료 내 관찰값들의 수와 변수의 수가 같을 때
- 포화모형의 $L$ 값은 자료가 완벽하게 모형화 된 경우에 해당하는 값이므로, 포화모형의 $L$ 값과 상대적으로 더 적은 수의 변수를 가진 축소모형(reduced model)을 적합해서 얻어지는 $L$ 값을 비교하면 축소모형의 **적절성(adequacy of the fit)** 평가 가능.
  - **가능도 비 통계량(likelihood ratio statistic, LRS)**: 두 모형 비교 방법으로 포화모형의 $L$ 값과 축소모형의 $L$ 값 간 비율 계산
$$LRS=-2\times \log\begin{pmatrix}\frac{L_{saturated}}{L_{fitted}}\end{pmatrix}$$
$$~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~=-2\times [\log(L_{saturated})-\log(L_{fitted})]$$
    - LRS를 **-2 로그-가능도(-2 log-likelihood)** 또는 **이탈도(deviance)**라고도 한다.
    - LRS는 근사적으로 카이제곱 분포에 따름, 자유도= 적합된 두 모형에서 사용된 모수 개수의 차이 ($n-k-1$).
    - $H_0$: 축소모형에는 없고 포화모형에만 추가적으로 있는 변수들에 해당되는 모수 모두 0.
        - 축소모형에 없고 포화모형에만 있는 변수들의 모집단 회귀계수가 모두 0이라는 것이므로 포화모형과 비교해서 축소모형의 적합도가 좋다는 가설.
    - LRS 클수록 결과는 유의하게 되고, 이는 곧 축소모형의 적합도(goodness of fit)가 나쁘다는 것을 의미.

`-` LRS가 가장 효과적일 때
- 포화모형은 아니지만, 한 모형이 다른 모형에 **포함된(nested)** 상태인 두 모형(큰 모형이 작은 모형에 있는 모든 변수들을 포함하고 있으면서 추가 변수도 있는 경우)을 비교하고자 할 때 LRS 사용.
  - 검정통계량: (추가 변수들이 있는)큰 모형의 LRS과 (추가변수들이 없는)작은 모형의 LRS의 차이
  - 검정통계량은 카이제곱 분포에 따르며, 자유도는 작은 모형에 비해 큰 모형에만 추가적으로 있는 모수의 수
  - $H_0$: 큰 모형에만 있는 추가 변수들의 모수들이 모두 0
- LRS는 모형 내 특정 공변량들에 관한 모수들이 모두 0이라는 귀무가설을 검정하고자 할 때도 사용.
  - 해당 공변량들이 포함된 모형(큰 모형)의 LRS와 공변량들이 제외된 모형(작은 모형)의 LRS 비교.
  - **모형 카이제곱(model chi-square)** 혹은 **공변량들에 대한 카이제곱(chi-squre for covariates)** : 두 LRS의 차이

### 회귀진단

- 회귀진단의 목적
  - 적합된 회귀모형의 수행력 검토,
  - 모수 추정값과 그 표준오차에 영향을 줄 수 있는 이상값들의 존재 여부를 평가.
  - 모형과 관련된 가정들이 만족되는지, 분석결과에 심하게 영향을 주는 영향값이 있는지 검토.

#### 예제 (31장 이어서)

`-` 가능도 비 통계량을 이용하여 변수들의 $P$-$value$ 계산(부록 C_p.315, 표32.2_p.182)
- 모형 1. 치료 과거력, 성별 및 기저시점에서의 CD8 세포 수 포함
- 모형 2. 모형 1에 치료요법에 초기 반응한 이후의 시간에 관한 2개 가변수를 추가
- 추가로, 이들 두 모형에서 계산된 -2로그-가능도 값들의 차이를 계산
    - 가변수를 포함하지 않은 모형 1에 비해, 가변수 2개가 추가로 포함된 모형 2는 -2로그-가능도 값이 5.53(393.12-387.59)만큼 낮아짐(즉, LRS=5.53).
        - 자유도가 2인 카이제곱에 따름.
    - 검정통계량에 관한 $P$-$value$는 0.063이며, 이는 곧 바이러스학적 실패율과 치료요법에 초기 반응한 이후의 결과시간 간의 관계는 경계선 상에서 유의하지 않음을 의미
    - 모형 2의 적절성 여부는 이 모형의 -2로그-가능도 값 387.59를 자유도가 982인 카이제곱 분포와 비교(척도 모수 추정값은 387.59/982=0.395, 이 값은 1에 비해 상당히 작다.)
    - $P$-$value$은 >0.99로, 모형 2의 적합도는 좋다고 할 수 있음.
    - 변수 5개를 사용하는 모형2는 저산포의 경향이 있는 것으로 보임.
        - -2로그-가능도를 자유도로 나눈 값인 0.39는 1보다 매우 작음.
        - 따라서 적합된 모형의 잔차 변동은 포아송 모형에서 기대되는 변동보다 작다고 할 수 있음.

## Chapter 33 통계적 모형의 설명변수들

### 명목형 설명변수들

- 회귀분석에서 명목형 범주형 변수의 효과를 평가하기 위해서는 **가변수(dummy variable)**를 만들어 분석
- 세 범주 이상으로 이루어진 명목형 변수의 경우,
    - 이 변수가 포함된 모형의 적절성을 평가
    - 이 변수의 유의성 여부를 평가하기 위해서는 해당 변수에 대한 가변수들을 모형에 모두 함께 포함
        - 일부 가변수들만 포함시킨다면 이는 해당 변수가 결과에 미치는 영향력 중 일부분만 평가하는 모형이 되기 때문
- 명목형 변수의 유의성 여부를 평가할 때는 가변수들 각각에 대한 개별 p-value를 평가하는 것보다는 가변수들의 효과를 동시에 평가할 수 있는 가능도 비 검정통계량(LRS)를 사용하는 것이 바람직

### 순서형 설명변수들

- 세 범주 이상을 가진 순서형 변수를 설명변수로 사용할 때 접근방식
    1. 범주형 변수의 각 범주에 수치를 할당할 뒤, 이 변수를 연속형 변수로 간주하고 분석하는 방법
         - 각 범주들 간의 순서가 충분히 반영된 변수,
         - 해당 설명변수와 종속변수 간의 선형성 여부를 평가하는 분석.
    2. 범주형 변수를 명목형 설명변수로 간주하고 이에 해당하는 가변수들을 사용하는 방법
         - 범주들 간의 순서가 반영되지 않기 때문에 자료의 정보 중 일부분이 손실.
         - 하지만, 해당 설명변수와 종속변수 간의 선형적인 관계를 가정하지 않기 때문에 더 선호되는 방법.
- 위 두 모형의 LRS 차이가 해당 설명변수의 **선형 추세 검정(linear trend test)**에 관한 검정통계량.
    - 이 검정통계량은 카이제곱 분포에 따르며, 자유도는 두 모형 간 모수 개수의 차이.
    - 분석결과가 유의하면 해당 변수의 범주 별 종속변수의 평균값이 선형적으로 증가(혹은 감소)하고 있지 않다는 의미.

### 연속형 설명변수들

- 연속형 설명변수의 회귀계수 추정값은 이 설명변수의 값이 한 단위 증가하는 경우 결과변수에 미치는 영향력의 크기가 얼마나 변하는지를 나타내는 값.
- 회귀모형에 연속형 설명변수들을 포함시키기 위해서는 먼저 이들 설명변수와 종속변수 간에 선형 관계 가정 타당 여부 검토.

`-` 선형성 가정에 대한 검토
- 단순 또는 다중 선형회귀에서 선형성 가정에 관한 검토
    - 연속형 설명변수 $x$와 연속형 종속변수 $y$ 간의 산점도 확인
    - $x$와 모형에서 얻어지는 잔차(residual)들 간의 산점도.
    - $x$와 $y$ 간에 근사적으로 직선적인 경향이 보이거나 또는 $x$와 잔차들 간에 어떠한 특정 패턴도 보이지 않는다면 선형성 가정이 만족.
    - 로지스틱 회귀나 포아송 회귀의 경우 몇 개(주로 5~10개)의 부집단들로 나누어 선형성 가정 검토.
        - 로지스틱 회귀에서는 각 부집단 별로 두 범주 종속변수 값 중 성공에 관한 로그-오즈 값을 계산한 후, 이 값들과 각 부집단 내 $x$값  범위의 가운데 값들 간의 산점도나 막대그림.

### 설명변수의 선택

- 포화모형이 아닌 경우라도 모형에 사용된 설명변수의 수가 지나치게 많으면 언제나 모형이 **과적합(over-fitting)** 위험성 존재.
    - 모형이 과적합되면 실제와는 다른 가짜 결과가 얻어질 수 있으며,
    - 특히 변수들 간 상관성이 높을 때 이러한 가능성이 더 높아짐.
        - 다중 선형회귀모형의 경우, 적정 변수의 수에 관한 대략적인 기준은 전체 관찰값의 수가 설명변수의 수보다 최소한 10배 이상.
        - 로지스틱 회귀모형의 경우에는 결과변수의 두 범주 각각에 대한 관찰값 수가 설명변수의 수보다 최소한 10배 이상.
- 종속 변수와 관련이 있을 것으로 보이는 설명변수의 수가 매우 많은 경우 다변수 분석전략
    - *모형에 사용되는 설명변수들은 생물학적 또는 임상적/보건학적 관점에서종속변수와 관련이 있을 것이라고 믿어지는 변수들이어야만 한다.*
     1. 단변수 분석(univariable analysis)로 각 설명변수들이 개별적으로 종속변수와 관련 여부 평가.
     2. 단변수 분석단계에서 유의하게 나타난 설명변수들을 모두 사용해 다변수 분석(multivariable analysis) 실시

`-` 설명변수의 자동 **-변수선택 절차(variable selection procedure)**들
- *모든 가능한 부분집합*
   - 설명변수들 개수의 모든 가능한 조합을 고려하는 방법
   - 각 모형의 $R^2$ 값이나 LRS 값을 기준으로 최종모형 선택.
- *후진제거*
   - 먼저 모형에 모든 설명변수들 포함.
   - 이들 중 모형 내에서 가장 덜 중요한 변수를 하나씩 제거.
   - 만일 제거되더라도 모형적합에 유의한 영향을 미치지 않게 되는 변수가 더이상 없을 때까지 계속 진행.
- *전진선택*
    - 아무런 설명변수도 선택하지 않은 상태에서 모형적합에 가장 중요한 기여를 하는 변수를 하나씩 추가.
    - 남아있는 변수들 중 더이상 모형적합이 유의하게 향상되는 변수가 없을 때까지 계속 진행.
- *단계적 선택*
    - 전진선택법과 후진제거법을 결합한 방법.
    - 설명변수를 전진적으로 하나씩 선택해 나가는데,
    - 각 단계의 끝에서 이미 선택된 변수들을 후진적으로 다시 평가,
    - 이전 단계에서는 중요했었으나 새로운 변수가 추가됨으로 인해 그 중요성이 사라지는지의 여부를 재검토.

`-` 자동 변수선택 절차들의 단점
- 자료에 대한 적합 정도가 동일한 모형이 두 개 이상 있을 수도 있고, 자료 값이 변하면 최종적으로 선택되는 모형도 달라질 수 있음.
- 변수를 선택해 나가는 절차 내에서 모형들을 서로 반복적으로 비교하기 때문에 다중검정 문제가 발생하게 되고, 따라서 제1종 오류가 상당히 커질 수 있음.
    - 이러한 문제점을 완화시키기 위해 좀 더 엄격한 기준의 유의수준을 사용.
- 변수 선택 시 최종적으로 선택된 변수들로 이루어진 모형을 사용해 자료를 다시 적합시키면 모수 추정값이 변수 선택의 마지막 결과와 다를 수 있음.
    - 변수 선택 결과는 맨 처음에 변수 선택을 위해 투입된 모든 설명변수들에 대해서 결측값이 없는 완전한 자료를 사용한 결과인 반면,
    - 최종 선택된 m개의 변수들을 사용해 재분석하면 이들 m개 변수들에 대해서만 결측값이 없는 완전한 자료를 사용해 분석하게 되므로
    - 실질적으로 더 많은 관찰값이 사용될 가능성이 높아지기 때문.
- 최종적으로 선택된 모형이 수학적으로는 옳은 것이라 할지라도 논리적으로는 맞지 않은 모형일 수 있음.

### 상호작용(interaction)

`-` 상호작용
- 회귀분석에서 하나의 설명변수와 종속변수 간의 관련성 크기가 다른 설명변수의 각 수준별로 다르다는 것, 즉, 두 설명변수가 종속변수에 미치는 영향이 서로 독립적이지 않음.

`-` 상호작용에 관한 검정
- 두 변수 간 상호작용 효과를 반영하기 위한 변수를 생성하기 위해서는 이들 두 변수의 각 관찰값들을 서로 곱함.
- 두 변수 모두 연속형인 경우에는 각 변수를 이분화해서 두 범주 변수로 만든 후, 이들 간의 상호작용 여부를 검토하면 분석결과를 더 편리하게 해석.
- 두 변수 중 하나가 다 범주 범주형 변수인 경우에는 이에 해당하는 일련의 가변수들을 만든 뒤, 각각의 가변수들과 나머지 변수를 서로 곱해서 여러 개의 상호작용 항들을 만듦.
- 두 변수 모두 다 범주 범주형 변수인 경우 역시 동일한 방법을 사용.
- 회귀모형에 상호작용 항을 추가할 때에는 반드시 **주 효과(main effect, 상호작용 항을 만드는 데 사용되는 변수들의 효과)** 항들과 함께 포함.
- 일반적으로 상호작용들에 대한 통계적 검정의 검정력이 낮음.
    - 이 문제는 특히 두 설명변수 모두 범주형 변수이면서, 변수들의 수준 별 조합에 의해 생성되는 부집단들 중 일부 부집단의 전체 관찰값 수나 관심사건의 수가 적으면 더 심각.

### 다중공선성

- 단변수 분석에서는 각 변수가 통계적으로 유의하다고 할지라도, 이들을 다변수 모형에 동시에 포함하면 둘 다 종속변수와 관련이 없는 것으로 나타날 수 있다. 이러한 현상을 **공선성(collinearity)** 또는 **다중공선성(multicollinearity)**.
- 다변수 모형에 공선성이 존재하는지의 여부를 평가하기 위해서는 설명변수들 간의 상관계수를 검토해 보거나, 아니면 다변수 모형의 회귀계수들에 대한 표준오차를 직접 눈으로 확인.

### 교란(혼란 confounding)

- 두 설명변수가 결과변수에 각각 영향을 미치면서, 이들이 서로 관련되어 있다면, 각 변수가 결과변수에 미치는 독립적인 영향력은 파악하기 힘듦. 
    - $\to$ 두 변수는 **교란되어 있다(confounded)**라 표현.

#### 예제 (31,32장 이어서)

- CD8 세포 수 변수가 바이러스학적 치료실패율과 선형적인 관계를 가지고 있다고 볼 수 있는지(선형성 가정) 평가(연속형 변수로 사용하는 것이 타당한가?)
    - CD8 세포 수의 분포를 동일 간격으로 5등분 한 다섯 개 부집단으로 나눈 다음, 각 집단 내에서 치료 실패율을 계산.
        - 각 집단의 로그-율에 관한 막대그림을 그려본 결과, 막대의 크기가 선형적으로 증가하는 경향을 보이지 않음.(그림 33.1_p.185), 따라서 범주형 변수로 사용
    - 기저시점에서의 CD8 세포 수 자료가 결측인 환자들이 많아 이들을 새로운 범주로 구분
- 결과(부록 C_p.315, 표 33.1_p.191)
    - CD8 세포 수에 관한 4개의 가변수들이 포함된 모형의 -2로그-가능도 값인 387.15와 나머지 변수들은 동일하지만 이 가변수들을 제외한 모형의 -2로그-가능도 값인 392.50 간의 차이에 대한 p-value는 0.25였음.(검정 통계량 5.35, 자유도 4)
    - CD8 세포 수와 바이러스학적 치료실패율은 통계적으로 서로 유의한 관련이 없다고 할 수 있음
    - 이것은 CD8 세포 수를 연속형 설명변수로 간주하고 분석한 결과와는 상반된 결과임.
    - 치료 과거력, 성별, 그리고 치료요법에 초기 반응을 한 이후의 경과시간 등이 바이러스학적 치료실패율에 미치는 영향은 CD8 세포 수 변수의 처리방법에 상관없이(즉, 연속형 변수로 분석하든지, 아니면 가변수를 사용한 범주형 변수로 분석하든지 간에) 큰 차이가 없었다.
