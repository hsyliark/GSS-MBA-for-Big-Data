## 8.2 메타분석을 위한 통계 모델


### 8.2.1 가설과 효과 크기(Effect Size)
* $H_0$ : D의 효과는 대조군 P와 차이가 없다. (귀무가설)
* $H_\alpha$ : D의 효과는 대조군 P보다 낫다. (대립가설)
    
    
* 효과 크기($\delta$)는 연속변수인 경우 평균값 차이, 비연속변수인 경우 비율, log 오즈비, 상대적 위험 사용


* $H_0$ : $\delta =0$
* $H_\alpha$ : $\delta >0$


* 분석방법
   * 고정(fixed)효과 모형 : $\hat \delta_i\rightarrow \delta $
   * 변량(random)효과 모형 : $\hat \delta_i\rightarrow \delta_i \rightarrow \delta $

### 8.2.2 고정효과 메타 분석 모형
\begin{align*}
    \hat{\delta}_i&=\delta+\epsilon_i \sim N(\delta,\hat{\sigma}_i^2)\\
    \hat{\delta} &=\sum_{i=1}^C w_i\hat{\delta}_i\\
    \hat{\sigma}^2 &=Var(\hat{\delta})= \sum_{i=1}^C w_i^2\hat{\sigma}_i^2 \mbox{ (under independence of centers)}
\end{align*}
- $\hat{\delta_i}$ : the estimate of treatment effect at $i$th center
- $\hat{\delta}$ : weighted mean or pooled estimate of teatment effect
- $\hat{\sigma}^2$ : variance of $\hat{\delta}$

* $\delta $의 95% 신뢰 구간(CI)
$$
   \hat{\delta}\pm 1.96\times \sqrt{\hat{\sigma}^2} 
$$

* $H_0: \delta=0$검정을 위한 t-type 검정통계량
$$
    T = \frac {\hat{\delta}} {\sqrt{\hat{\sigma}^2} 
$$
    
    
* 가중치($w_i$) 선택 방식 4가지


   (1) 센터의 수로 단순히 나눔 : $ w_i = $ $1 \over C$
       
              
   (2) 센터마다 환자수 고려 : $ w_i = $ $N_i \over N$
   
   
   (3) 각 센터의 환자수와 실험군 수 고려 : $ w_i = $ ${N_{iD} N_{iP}} \over {N_iD + N_iP}$ x $ 1 \over w$
   
   
   (4) 역분산 고려 : $ w_i = $ $ 1 \over {\hat \sigma_i^2}$ x $ 1 \over w$
   
   
* 이항 반응 변수의 경우 
   * 각 center별로 $2\times 2$ 분할표로 요약 
   * 처리와 반응의 연관성은 Mantel and Haenszel 검정 
   * `rmeta`패키지의 `meta.MH` 이용하여 계산  

### 8.2.3 변량효과 메타 분석 모형 : Dersimonian-Laired
\begin{align*}
    \hat{\delta}_i&= \sim N(\delta_i,\sigma_i^2)\\
    {\delta}_i&= \sim N(\delta,\tau_i^2)
\end{align*}

$$
   \hat{\delta}_i=\delta+\nu_i+\epsilon_i \sim N(\delta,\sigma_i^2+\tau^2)
$$
* $\nu_i$ : between-center variation
* `rmeta`의 `meta.DSL` 또는 `meta`의 `metabin, metacont` 이용하여 계산

### 8.2.4 출판 편향
* 약물이 효과가 있다는 연구만 출판할 가능성
* 깔대기 그림(funnel plot)이나 Egger's plot으로 판단 - 샘플이 적은 연구가 대규모 연구보다 출판 편향을 지니기 쉽다는 원리

### 8.3.1 Betablocker 시험 분석

```r
install.packages("rmeta")
```

#### 8.3.1.1 고정효과 메타분석

```r
library("rmeta")
```

```r
data("cochrane")
```

```r
cochrane
```

```r
library("flexmix")
```

```r
data(betablocker)
```

```r
betablocker
```

```r
head(betablocker)
```

```r
# Get the data from the ``beta"
n.trt   = betablocker[betablocker$Treatment=="Treated",]$Total
n.ctrl  = betablocker[betablocker$Treatment=="Control",]$Total
ev.trt  = betablocker[betablocker$Treatment=="Treated",]$Deaths
ev.ctrl = betablocker[betablocker$Treatment=="Control",]$Deaths
```

```r
n.trt
```

```r
n.ctrl
```

```r
ev.trt
```

```r
ev.ctrl
```

```r
?meta.MH
```

```r
# call the meta.MH for calculations

betaOR  = meta.MH(n.trt,n.ctrl,ev.trt,ev.ctrl, names=paste("Center",1:22,sep=" "))
```

```r
# print the summary from meta.MH
summary(betaOR)
```

* OR 값은 0.48 ~ 1.56까지의 범위
* Global OR는 0.77(0.7, 0.85)로 계산됨 - 통계적으로 유의함
* 이질성 검증 결과 p=0.3301로 동질적인 것으로 확인됨

* 숲 그림(forest plot)을 그려봄

```r
plot(betaOR, ylab="center")
```

* 깔대기 그림(funnel plot)을 그려봄

```r
funnelplot(betaOR)
```

* 상부는 sample size가 커서 폭이 좁고 하부는 sample size가 적어서 분산이 클 가능성이 높아 폭이 넓을 것임
* 전체적으로 평균값을 중심으로 편의는 없는 것으로 판단

#### 8.3.1.2 변량효과 메타분석

```r
# Call the meta.DSL for calculations
betaDSL  = meta.DSL(n.trt,n.ctrl,ev.trt,ev.ctrl, names=paste("Center",1:22,sep=" "))
```

```r
# Print the summary from meta.DSL
summary(betaDSL)
```

* Global OR는 0.78(0.7, 0.87)로 계산됨 - 통계적으로 유의함
* 95% 신뢰구간인 (0.7, 0.87)은 고정효과 모형에 비해 약간 넓어짐
* 이질성 검증 결과 p=0.3301로 동질적인 것으로 확인됨

* 숲 그림(forest plot)을 그려봄

```r
plot(betaDSL, ylab="Center")
```

* 복잡한 숲 그림(forest plot)을 그려봄

```r
# Create the ``text" to include all the outputs
text = cbind(c("","Center",betaOR$names,NA,"Summary"),c("Deaths","(Betablockers)",ev.trt,NA,NA), c("Deaths","(Placebo)", ev.ctrl, NA,NA),
	     c("","OR",format(exp(betaOR$logOR),digits=2),
		    NA,format(exp(betaOR$logMH),digits=2)))
```

```r
head(text)
```

```r
# Generate the OR and 95\% CI
mean  = c(NA,NA,betaOR$logOR,NA,betaOR$logMH)
sterr = c(NA,NA,betaOR$selogOR,NA,betaOR$selogMH)
l     = mean-1.96*sterr
u     = mean+1.96*sterr
```

```r
?forestplot
```

```r
# Call forestplot with a few options
forestplot(text,mean,l,u,zero=0,is.summary=c(TRUE,TRUE,rep(FALSE,22),TRUE),
   clip=c(log(0.1),log(2.5)), xlog=TRUE)
```

### 8.2.3 변량효과 메타 분석 모형 : Dersimonian-Laired

```r
# Load the data
data(cochrane)
# print it
cochrane
```

```r
# Fit the fixed-effects model
steroid = meta.MH(n.trt, n.ctrl, ev.trt, ev.ctrl, names=name, data=cochrane)
```

```r
# Print the model fit
summary(steroid)
```

* Global OR는 0.53(0.39, 0.73)으로 계산됨 - 통계적으로 유의함(효과가 있음)
* 이질성 검증 결과 p=0.3303로 동질적인 것으로 확인됨

```r
# Create the ``tabletext" to include all the outputs
tabletext = cbind(c("","Study",steroid$names,NA,"Summary"),
                 c("Deaths","(Steroid)",cochrane$ev.trt,NA,NA),
                 c("Deaths","(Placebo)",cochrane$ev.ctrl, NA,NA),
		 c("","OR",format(exp(steroid$logOR),digits=2),
			NA,format(exp(steroid$logMH),digits=2)))
# Generate the CI
mean   = c(NA,NA,steroid$logOR,NA,steroid$logMH)
stderr = c(NA,NA,steroid$selogOR,NA,steroid$selogMH)
l      = mean-1.96*stderr
u      = mean+1.96*stderr
# Call forestplot
forestplot(tabletext,mean,l,u,zero=0,
	is.summary=c(TRUE,TRUE,rep(FALSE,8),TRUE),
   	clip=c(log(0.1),log(2.5)), xlog=TRUE)
```
