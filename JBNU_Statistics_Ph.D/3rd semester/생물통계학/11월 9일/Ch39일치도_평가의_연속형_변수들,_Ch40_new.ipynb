## 2. 연속형 변수들

### 1) 반복성 척도(measures of repeatability) 및 블랜드-알트만 그림(Bland-Altman diagram)
+ 차이 값들의 표준편차 추정값($s_{d}$)은 개인 별 일치도에 관한 측정값
+ 이 값보다 영국 국가표준 기관 반복성 계수(BSI repeatability coefficient)인 2$s_{d}$ 즉 두 측정 간에 발생 가능한 최대 차이를 더 많이 사용
 - 차이가 정규분포를 한다고 가정하면, 관찰된 차이 값들의 평균을 $\bar{d}$라 할 때, 모집단에 있는 차이 값들 중 약 95는 $\bar{d} \pm 2s_{d}$ 내에 존재
 - 이 구간의 상한과 하한을 일치 한계(limits of agreement)라 한다
 
 > 두 짝진 측정값들의 일치여부를 이 한계를 사용해 판단
 
 > n개의 짝 각각에 대해 평균과 차이를 계산한 뒤, x-축에는 평균을, y-축에는 차이를 사용한 산점도를 그리고, 그 위에 일치 한계를 표시

+ 블랜드-알트만 그림(Bland and Altman diagram)
 - 산점도 상의 점들이 무작위로 흩어져 있으면(만일 각 짝 간에 어떠한 체계적인 차이도 개입되지 않았다면 점들은 y-축의 0을 중심으로 위, 아래에 균일하게 흩어져 있음) 반복성에 관한 척도로 단일 측정값을 사용해도 좋음
 - 블랜드-알트만 그림에서 평균이 작을 때에 비해 평균이 클 때 차이 값들의 변동이 상대적으로 더 심하게 나타나는 깔때기 효과(funnel effect)가 관찰된다면 반복성 척도를 다시 평가

### 2) 신뢰도 지표(index of reliability)
+ 급내상관계수(intraclass correlation coefficient, ICC)
 - 반복성과 재현성을 평가하는 데 주로 사용되는 신뢰도 지표
 - 0(전혀 일치하지 않음)에서 1(완벽하게 일치함) 사이의 값
 - 두 짝진 관찰값들 간의 일치도를 측정하는 경우, ICC는 관찰값들의 전체 변동량 중 짝들 간 차이로 인한 변동량이 차지하는 비율을 나타냄
 
 > 관찰값들의 전체 분산(total variance) 중 짝들 간 분산(between-pair variance)이 차지하는 비율
 
 - 짝들 간에 구조적인 차이가 없을 때 ICC
 
 > n개의 짝을 각각 두 번씩 사용한- 즉, 한 번은 관찰된 짝을, 다른 한 번은 짝 내 값들의 위치를 바꿔 새로 만든 짝을 함께 사용한- 전체 2n개 짝들을 대상으로 피어슨 상관계수(Pearson correlation coefficient)를 계산한 값
 
 - 짝진 자료들 간에 구조적인 차이가 있을 때 ICC
 $$
 ICC = \frac{s_{a}^2 - s_{d}^2}{s_{a}^2 + s_{d}^2 + \frac{2}{n} (n\bar{d}^2 - s_{d}^2)}
 $$
 
  + $s_{a}^2$ = n개 합들의 분산 추정값
  + $s_{d}^2$ = n개 차이들의 분산 추정값
  + $\bar{d}$ = n개 차이들 평균(구조적인 차이에 대한 추정값)

+ 린의 일치상관계수(Lin's concordance correlation coefficient, CCC)
 - 산점도 상의 점들을 가장 잘 적합해 주는 직선과 원점을 통과하는 $45^{\circ}$ 기울기의 직선 간의 차이를 고려하면서, 점들과 적합선 간의 거리를 평가하기 위해 Pearson 상관계수를 수정한 것
 - -1과 1 사이의 값
 - 최대값 1은 짝진 두 값들이 모두 완전히 일치할 때, 즉, 산점도 상의 모든 점들이 원점을 통과하는 $45^{\circ}$ 직선 위애 놓이게 될 때 얻어짐
 $$
 r_{c} = \frac{2rs_{x}s_{y}}{s_{x}^2 + s_{y}^2 + (\bar{x} - \bar{y})} 
 $$
 
 + r은 n개의 짝 $(x_{i}, y_{i})$에 관한 Pearson 상관계수의 추정값
 + $\bar{x}$와 $\bar{y}$는 각각 변수 x와 y의 표본평균
 $$
 s_{x}^2 = \frac{1}{n}\sum_{i=1}^n (x_i - \bar{x})^2, 즉, x에 대한 분산 추정값의 \frac{n-1}{n}배,
 $$
 
 $$
 s_{y}^2 = \frac{1}{n}\sum_{i=1}^n (y_i - \bar{y})^2, 즉, y에 대한 분산 추정값의 \frac{n-1}{n}배
 $$

HPV DNA 검사를 위한 건식추출법은 참고표준 방법인 습식추출법의 대안으로 사용되기에 충분함

### 예제2. 일치도 평가 - 연속형 변수

+ Erikson 방법으로 50개의 수직고경을 측정함

+ Lin의 일치상관계수(CCC)와 동일함

# Chapter 40. 근거-기반 의학

+ 근거-기반 의학(evidence-based medicine, EBM)
 - 개별 환자를 돌보기 위한 의사결정시, 현존하는 가장 최상의 증거를 진실되고, 명확하고, 신중하게 사용하는 것
 - 제기된 문제에 대한 답을 제시할 수 있는 근거 수준(level of evidence)이 각 연구설계 별로 다름
 
          가장 높은 수준부터 가장 낮은 수준의 근거 순서
 
 > 무작위배정 임상시험(RCT)들에 대한 체계적 고찰(systematic review) 또는 메타분석(meta-analysis) $\to$ RCT $\to$ 코호트 연구(cohort study) $\to$ 사례-대조 연구(case-control study) $\to$ 단면조사 연구(cross-sectional study) $\to$ 증례보고(case reports) $\to$ 전문가 의견(expert opinion) $\to$ 일화적 정보(anecdotal information)

 - 결과의 중요도는 관심효과의 신뢰구간을 사용해서 결정(예: 군 간 평균 차이에 대한 신뢰구간)
 
 > (i) (관심효과에 대한 가설검정 결과의 통계적 유의성 여부와는 상관없이) 만일 신뢰구간의 하한이 실제 효과라고 가정한다면, 이 값은 임상적으로 중요한 효과라고 간주할 수 있을 만큼 충분히 큰가?
 
 > (ii) 만일 신뢰구간의 상한이 실제 효과라고 가정한다면, 이 값은 임상적으로 중요한 효과라고 간주할 수 있을 만큼 충분히 큰가?
 
 > (iii) 이들 두 질문에 대한 답이, 해당 연구결과가 모호하지 않고 또 중요한 결과라고 판단해도 좋을 만큼, 서로 충분히 유사한 결론인가?
 
+ 무작위배정 임상시험에서 치료요법의 효능을 평가하고자 하는 경우에는 대조 처치(control treatment) 대비 중재 처치(intervention treatment)에서 '한 명의 환자가 더 치료되기 위해 필요한 환자의 수(number of patients needed to treat, NNT)'를 계산
 
NNT의 의미
 
 > '안 좋은' 결과를 평가하는 연구의 경우, 몇 명의 환자들이 대조 처치에 비해 중재 처치를 더 받아야 안 좋은 결과의 발생이 한 명 예방되는지에 관한 인원 수

## 5. 결과를 임상진료에 적용한다

+ 다른 사람의 연구결과가 내 환자들을 치료하는 데 도움이 되기 위해서 만족되어야 하는 것
 - 내가 진료하는 환자들이 해당 연구결과가 얻어진 환자들과 유사한가?
 - 내가 진료하는 환자들에게 해당 연구결과를 적용할 수 있는가?
 - 임상적으로 중요한 결과들이 모두 고려되었는가?
 - 해당 연구결과를 내가 진료하는 환자들에게 적용했을 때 얻을 수 있는 효과가 잠재적 위험이나 추가로 소요되는 비용 등을 상쇄시킬 수 있을 정도로 충분히 큰가?
