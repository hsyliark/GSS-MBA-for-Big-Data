# Ch 43. 체계적 고찰 및 메타분석

##### 체계적 고찰

`-` `체계적 고찰`: 동일한 또는 유사한 주제에 대해 연구한 모든 적절한 연구결과들의 정보를 일정한 형식에 따라 체계적으로 결합하는 것
  -  동일하거나  유사한 처리를 실시한 임상시험들을 대상으로 실시
  - 임상가의 전문 식견과 함께 최상의 근거에 기초하고 현재 활용가능한 연구결과들을 환자 치료에 사용하는 `근거-기반 의학`의 핵심적인 부분에 해당

`-` `코크란 협력단체` : 근거-기반 의학에서 중요한 체계적 고찰을 위해 결성된 국제적인 네트워크
 - 체계적 고찰에 관한 코크란 데이터베이스 등 근거-기반 보건의료 데이터베이스들을 모은 코크란 라이브러리 운영
     - `참고` https://community.cochrane.org/ 

`-` 체계적 고찰으로 얻을 수 있는 것들
   - **정제와 축소**: 대규모 정보들을 정제해서 사용 가능한 크기의 정보로 요약
   - **효율성**: 새로운 연구를 수행하는 것에 비해 빠른 속도, 적은 비용
   - **일반화 가능성 및 일관성**: 단일연구에서 얻어지는 결과보다 광범위하고 다양한 환자 모집단으로 일반화 가능
   - **신뢰성**: 체계적 고찰의 목적 **오차 줄이기**
       - 종설(체계적이지 않은 고찰)이나 단일 연구에 비해 임상진료지침의 신뢰성 및 정확성 향상
   - **검정력 및 정밀성**: 수량화된 체계적 고찰 수행시 실제 관심효과들을 발견해 낼 수 있는 검정력이 증가 $\to$ 단일연구보다 더 정밀한 추정값 제시 가능

##### 메타분석

`-` **메타분석**: 수치형 결과들에 초점을 맞춘, 체계적 고찰의 특수한 형태
   - 주 목적: 각 개별 연구들의 결과를 결합해서 관심효과에 관한 전반적 혹은 평균적 효과를 추정하고자 하는 것

`-` 수행 방법


**1. 관심효과에 대해 정의하고 만일 각 연구의 원자료들을 얻을 수 있으면, 이를 사용해 관심효과 평가**
- 현실적으로는 원자료 획득이 힘들어서 출판된 논문에서 관심효과 추출
- 두 처리군을 비교하는 임상시험의 경우, 관심효과는 `결과변수의 형태`에 따라 정해짐
     - `연속형`
         - 관심효과는 처리 평균들 간의 차이,(차이가 0이라는 의미는 처리효과가 없다는 뜻)
     - `두 범주`(예; 사망/생존)
         - 주 관심은 처리군에서 결과(예; 사망)가 발생할 위험
         - 관심효과는 군 간 위험 또는 율의 차이(상대위험)(두 집단 간 위험의 차이가 0이라는 의미(RR=1)는 처리효과가 없다는 뜻)

**2. 연구들 간 통계적 동질성 여부를 검토하고, 통계적 이질성에 관한 추정값 계산**

- `코크란의 동질성 카이제곱 검정`
     - $H_0$: k개 연구들 간 집단 처리 효과들은 동질적이다.($\theta_1=\theta_2=...=\theta_k$,자유도 k-1인 카이제곱 분포를 따름) 
     - 메타분석을 실시하고자 하는 연구들의 수가 적으면 검정력이 낮아 연구들 간의 통계적 이질성을 발견해내기 힘듦
     - 연구의 수가 매우 많으면 이들 간의 이질성 여부가 메타분석 결과에 영향을 미치지 않는 경우에도 검정 결과는 매우 유의하게 나올 수 있어, 연구들이 이질적이라는 결론을 내리기 쉬움
     - 따라서 연구들 간 이질성 여부에 대한 평가는 주로 $I^2$ 통계량 많이 사용
                 
- 지표 $I^2$
    - 연구들 간의 전체 변동량 중 이질성으로 인한 변동량이 차지하는 비율
    - 연구의 수나 결과변수의 형태 또는 처리효과의 종류 등에 영향을 받지 않는 지표
        - 연구들 간 일관성 여부 평가
        - 이질성이 메타분석 결과에 미치는 영향력의 크기 정량화하는데 사용
        - 0% ≤ $I^2$ ≤ 100%  (이질성이 없으면 0%)

**3. 결합효과 및 신뢰구간을 추정하고 해당 효과에 대한 가설검정을 수행.**

- 결합효과: 기본적으로 각 연구에서 얻어진 효과 추정량들을 가중평균해서 구함(가중값으로는 각 효과 추정값의 역분산 사용)

`-` 추정값들이 통계적으로 이질적이라는 증거가 없으면 `고정효과메타분석` 실시
- 고정효과모형
    - 모든 연구들의 모집단 처리효과는 같고, 각 연구로부터 관찰되는 추정값들의 변동은 전적으로 표본추출 오차 때문이라 가정
    - 결합효과의 분산은 연구-내 변동


`-` 추정값들이 통계적으로 이질적이라는 증거가 있을때 결합효과 계산 방법


1) `변량효과 메타분석`
- `변량효과 모형`: 각각의 연구들이 어떤 평균 처리효과를 중심으로 분포하고 있는 모집단 내 여러 연구로부터 무작위로 표본추출된 연구들이라는 가정
    - 결합효과의 분산은 연구-내 변동과 연구-간 변동으로 구성됨 $\to$ 해당 추정값에 대한 표준오차 커짐 $\to$ 실제 처리효과에 대한 신뢰구간도 넓어짐 $\to$ $P-value$ 역시 고정효과 메타분석보다 커지게 됨.(=결합효과가 통계적으로 유의하게 될 가능성 $\downarrow$)


2) `부집단 분석`: 연구들을 동일한 특성을 가진 몇 개의 부집단으로 층화한 후, 각 층 별로 분석 실시


3) `메타회귀분석`: 회귀분석방법론을 사용해 연구들 간의 차이가 보정된 결합효과 추정값을 계산하고, 연구들 간에 이질성이 나타나게끔 한 특성(공변량)들은 무엇인지 평가할 수 있는 방법
    - 종속변수: 각 연구에서 얻어진 효과 추정값
    - 독립변수: 개별 연구-수준의 특성들
- `변량효과 메타회귀모형`: 가장 자주 사용되는 메타회귀모형
    - 일반회귀모형에 연구-간 변동을 나타내는 오차항을 추가로 포함시켜 분석 시 이 효과를 반영하기 위한 모형
    - 표본 수:  체계적 고찰에 사용되는 연구들의 수$\to$검정력이 매우 낮아 중요한 효과 발견 어려움
     - 다중공선성 존재시 각 공변량의 효과 독립적 분리 불가$\to$잘못된 결론에 도달할 수 있음
     - 요약값 수준의 변수는 설명변수로 사용$\to$동일 모집단 내 개인 수준에서의 연관성을 반영하지 못하는 생태학적 오류의 발생 가능성

**4. 결과를 해석하고 제시**
- 각 연구들의 효과를 표 형태로 요약
- `숲 그림` 각 연구의 관심효과 크기를 그림으로 표현
    - '처리효과 없음'을 나타내는 수직선 상에 각 연구의 효과 추정값 및 이들의 평균 효과를 가로 선으로 표현하는 방법
    - 각 연구효과의 추정값은 보통 상자로 표시, 상자 면적은 해당 연구의 크기에 비례
    - 효과 추정값의 신뢰구간은 상자를 중심으로 좌우로 그은 선
    - 맨 밑에 있는 다이아몬드는 결과효과 추정값 및 해당 신뢰구간
- 숲 그림의 평가
   1. 수직선을 기준으로 같은 방향으로 놓여 있는지 검토
   2. 신뢰구간들이 겹쳐 있는지 평가
   3. 신뢰구간이 많이 겹치지 않는 연구는 작은 표본 수를 가지고 있지는 않은지 검토
   4. 각 연구의 효과 및 결합 효과에 대한 신뢰구간이 수직선을 포함하고 있는지 확인해서 개별 효과들 및 결합효과의 유의성 여부 평가

```python
from PIL import Image
Image.open('4.jpg').resize((500,300))   #숲 그림
```

- 위에서부터 총 3건의 임상연구는 1을 포함하고 있어 통계적으로 유의하지 않은 것이고, 나머지 아래 2건은 1을 포함하고 있지 않아서 통계적으로 유의한 것이다. 이렇게 총 5개의 임상연구를 메타분석한 결과 A라는 치료가 효과가 있다는 결론을 마름모를 통해서 보여준다. 만일 마름모가 1을 포함하다면 우리는 A라는 치료가 통계학적으로 유의한 효과가 있다고 말할 수 없는 것이다.
- 출처 : http://www.mjmedi.com

#### 장점 및 단점

`장점`:  많은 표본 수 사용하는 메타분석 $\to$ 처리효과 탐지하는 검정력 증가 $\to$ 결합효과 추정시 정밀도 증가

`단점`: 부적절한 사용으로 잘못된 결론
 - 출판 편향: 통계적으로 유의한 결과를 보인 연구들로 메타분석 실시할 시, 발생하는 편향
     - **깔때기 그림**-수직축에는 각 연구의 표본 수를 수평축에는 처리효과를 사용한 산점도
         - 출판 편향 존재시 한 쪽으로 치우쳐진 비대칭 형태
 - 임상적 이질성: 연구들 간에 환자 모집단의 특성, 연구에 사용된 추적기간 등이 서로 다르기 때문에 이들에 대한 메타분석은 비교성을 확보하지 못하는 문제
 - 연구의 질 평가: 각 연구들이 연구설계나 수행과정들이 서로 달라 연구의 질도 서로 다 다름.
 - 종속성 평가: 분석에 포함되는 연구결과들이 서로 독립적이지 않을 수 있음(ex. 한 연구가 여러 개의 논문으로 중복 출판)

#### 민감도 분석

- 메타분석에서 민감도 분석은 결합효과 추정값에 대한 강건성 평가


- 민감도 분석 방법


1. 두 모형 결과 비교평가
    - 고정효과 모형과 변량효과 모형 중 어떤 모형을 사용하는지에 따라 결합 추정값의 유의성 달라질 수 있음
2. 영향력 있는 연구 존재 확인
    - 결합효과 추정값에 유난히 큰 영향을 미치는 연구가 있는지 평가
3. 누적메타분석
    - 연구들을 특정 순서에 따라 하나씩 추가하면서, 누적된 연구들을 대상으로 매번 메타분석을 실시하는 방법

#### 예제

- 8개 관찰연구들을 사용한 메타분석을 실시해서 불안장애를 가진 환자들의 뇌졸중 발생위험 평가
- 전체 환자 수는 950,759명
- 각 연구별 관심효과는 위험비임.

```python
from PIL import Image
Image.open('3.jpg').resize((500,500))   #p268 그림43.1
```

- 깔때기 그림은 비교적 대칭적이기 때문에 메타분석에 포함된 연구들에 출판 편향은 개입되지 않은 것으로 판단.

```python
from PIL import Image
Image.open('1.jpg').resize((1000,500))   #p268 그림43.2
```

- 숲 그림을 보니 개별 연구의 HR 추정값 크기는 서로 상당히 다르지만 신뢰구간들은 모두 상당히 겹침,
- 2개 연구는 불안장애 환자들의 뇌졸증 발생 위험이 감소되는 것으로 나온 반면, 나머지 연구들은 이 위험이 증가되는 것으로 나타났다.

```python
from PIL import Image
Image.open('.jpg').resize((1000,700))   #p269 표43.1
```

- 연구들 간 통계적 이질성을 평가하기 위해 Cochran의 동질성 카이제곱 검정을 실시
- $H_0$: 8개 연구들 간 집단 처리 효과들은 동질적이다.($\theta_1=\theta_2=...=\theta_8$) 
- 검정통계량 Q = 9.55
- 자유도 = 8-1 = 7
- $p-value=0.216$

-> 귀무가설을 기각할 수 없어, 집단 처리 효과들이 동질적인 것으로 나타남

-  $I^2 = \frac{Q-df}Q X 100 = \frac{9.55-7}{9.55} x 100 = 26.7\%$


- Cochran handbook에서는 $I^2$이 25% 정도이면 연구들 간의 이질성 수준이 낮은 것으로 봄

- 따라서 연구결과들 간에 약하지만 작지 않은 수준의 불일치가 있다고 볼 수 있어 결합 HR 추정값을 해석할 때 주의가 필요할 것으로 판단

- 계산된  $I^2$값과 더불어, 연구의 수가 작은 경우 이질성을 평가하기 위한 카이제곱 검정의 검정력이 낮다는 점을 고려해 좀 더 보수적인 방법인 변량효과 메타분석 실시

- HR에 관한 메타분석 결합추정값은 1.24(95% 신뢰구간, 1.09~1.41)이며, 이는 불안이 뇌졸중과 유의미한 관련이 있고,불안장애로 고통받는 환자들은 뇌졸중 발병 위험이 24% 정도 증가하게 됨을 의미
