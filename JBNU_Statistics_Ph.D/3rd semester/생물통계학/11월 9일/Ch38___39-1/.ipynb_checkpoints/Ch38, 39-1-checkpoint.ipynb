## Chapter 38. 진단도구들

`-` **기준구간(reference interval), 진단검사(diagnostic test)**의 사용목적
- 임상검사 결과 상 환자의 질병유무 **진단(diagnosis)** 위함
- 건강인 모집단 내 사람을 **선별(screening)** 할때
- 질병 발견할 수 있는 **일상적인(routine)** 검사

> ex) 혈액검사
- **병기(staging)** 검사

> ex) 암 검사
- 환자 상태의 호전 여부 추적 위한 **점검(monitoring)** 검사

> ex) 혈압

### 기준구간(reference interval)

`-` 단일 연속형 변수에 대한 **기준구간-기준범위(reference range)** or **정상범위(normal range)**
- 정상인의 일반적 관찰 값들의 범위
- 대규모 표본 사용해 계산
- 개인의 값이 기준구간의 상한보다 크거나 하한보다 작음 $\to$ 일반인들에 비해 비정상적으로 높(낮)다.

`-` 기준구간의 계산
- 자료가 정규분포를 따른다고 가정 \$\to$ 자료를 사용해 두 한계 값 계산 

> $\bar{x}\pm 1.96\times s$

- 분포 고려 X $\to$ 자료의 가운데 95%를 포함하는 중앙범위 사용 

> 2.5번째 백분위수, 97.5번째 백분위수

`-` 기준구간에 대한 다른 요인들의 효과
- 연속형 변수 값은 연령, 성별 등 다른 요인들에 의해 영향 받음

- 이런 경우, 먼저 요인 고려 후 값 해석

> ex) 수축기 혈압에 관한 기준구간: 남/여

### 진단검사(diagnostic test)

`-` 정확한 진단 위한 **참고표준 검사(reference standard test, 황금기준 검사, gold_standard test)**  
- 비현실적, 정기적 실시 어려움
- 간단한 진단(대조)검사 필요

`-` 진단검사에 대한 평가 $\to$ 실제 질병상태 유무를 아는 집단을 대상

```python
from PIL import Image
Image.open('8주차_img/img1.png')
```

- **유병률(prevalence)**: n명 중 a+c명이 질병을 가지고 있음 ($\frac{a+c}{n}$)
- **진양성(true positive)**: 질병에 걸린 a+c명 중 a명의 결과가 양성 ($\frac{a}{a+c}$)
- **위음성(false negative)**: 질병에 걸린 a+c명 중 c명의 결과가 음성 ($\frac{c}{a+c}$)
- **진음성(true negative)**: 질병에 걸리지 않은 b+d명 중 d명의 결과가 음성 ($\frac{d}{b+d}$)
- **위양성(false positive)**: 질병에 걸리지 않은 b+d명 중 b명의 결과가 양성 ($\frac{b}{b+d}$)

`-` 신뢰도(reliability) 평가: 민감도(sensitivity)와 특이도(specificity)
- **민감도(진양성률)**: 실제 질병 있는 사람이 검사에서도 양성으로 맞게 판정되는 비율 ($\frac{a}{a+c}$)
- **특이도(진음성률)**: 실제 질병 없는 사람이 검사에서도 음성으로 맞게 판정되는 비율 ($\frac{d}{b+d}$)
- 백분율 표시, 신뢰구간 계산 가능
- 민감도, 특이도 둘다 1.0(즉, 100%)가 베스트
    - 현실: 민감도 증가 $\to$ 특이도 감소, 특이도 증가 $\to$ 민감도 감소
- 어떤 지표가 높은 검사 선택 여부: 위음성 or 위양성률 값 and 평가대상 질병 상태 고려
    - 쉽게 치료되는 질병: 민감도 높은 검사
    - 심각한 질병: 특이도가 높은 검사 ($\because$ 위양성 진단 회피)
- 선별검사 시 중요점
    - 연구대상자에게 양성 진단의 의미 설명
    - 결과가 위양성 or 위음성일 수 있음을 이해시켜야함

`-` 진단검사 결과의 활용: 예측도(predictive values)
- **양성예측도(positive predictive value)**: 양성 결과인 사람이 실제 질병 O 비율 ($\frac{a}{a+b}$)
- **음성예측도(negative predictive value)**: 음성 결과인 사람이 실제 질병 X 비율 ($\frac{d}{c+d}$)
- 백분율 표시, 신뢰구간 계산 가능
- 민감도, 특이도: 검사의 진단력(diagnostic ability)을 정량화 해주는 척도
- 양성예측도, 음성예측도: 검사결과가 주어진 상태에서 질병 유무에 대한 가능성의 크기에 대한 척도
- 예측도의 크기: 모집단 내 유병률 크기에 영향
    - 양성예측도: 희귀질병(rare disease)인 모집단 > 흔한 질병(common disease)인 모집단
    - 음성예측도: 양성예측도와 반대
- 해석: 연구모집단 내에서 이루어져야지, 넘어 다른 모집단으로 일반화 X

`-` 분류기준 값(cut-off value)의 사용
- 측정값들을 사용하여 특정 기준 값 이상 or 이하일 때 질병이 있다고 판단되는 적정 기준이 없는 경우 $\to$ 분류기준 값 설정
- 분류기준 값 설정 방법: 기준구간의 상(하)한 사용
    - 적절성 여부 평가: 민감도, 특이도, 예측도 등의 지표 사용
    - 분류기준 값 변화 $\to$ 지표 변화 $~\therefore$ 최종 분류기준 값 = 지표들이 최적이 되는 값

`-` ROC 곡선 (receiver operating characteristics curve, ROC curve)
- **ROC곡선(수신자 작업특성 곡선)**
    - 유용한 검사인지 평가
    - 두 검사의 수행력 비교
    - 검사의 최적의 분류기준 값 선택

```python
Image.open('8주차_img/img2.png')
```

- **ROC 곡선 아래 면적(area under the curve, AUC or AUROC)**
    - 검사법이 두 개 이상 있는 경우, 이들 간 정확도를 비교
    - **c 통계량(c statistic)**
    - AUC가 높은 검사 $\to$ 질병 유무 잘 판별
        - c = 1: 질병 유무 완벽 판별
        - c = 0.5: 질병 유무 판별 X
        - c < 0.5: 질병여부를 오히려 반대로 예측 (있어서는 안됨)

`-` 검사가 쓸모 있는가?
- **가능도 비(likelihood ratio, LR)**: 검사도구의 유용성 평가에 사용
- **양성 결과에 관한 가능도 비(LR for a positive result)**
    - 실제 질병에 걸린 환자가 양성으로 나타날 확률 / 실제 질병에 걸리지 않은 환자가 양성으로 나타날 확률

$$양성~가능도~비(LR~positive),~LR(+)~=~\frac{a/(a+c)}{b/(b+d)}$$

- 양성 가능도 비와 민감도 및 특이도 간 LR(+) = $\frac{민감도}{1-특이도~~~~}$의 관계 성립
- **음성 가능도 비(음성 결과에 대한 가능도 비(LR for a negative result, LR negative))**

$$음성~가능도~비,~LR(-)~=~\frac{c/(a+c)}{d/(b+d)}$$

- 음의 가능도 비와 민감도 및 특이도 간 LR(-) = $\frac{1-민감도~~~~}{특이도}$의 관계 성립
- 양성 가능도 비가 크다: 이 검사에서 나온 양성 결과는 믿을 만함
- 음성 가능도 비가 크다: 검사에서 음성 결과는 질병이 없는 것으로 판단(rule-out)하기에 충분

## Chapter 39. 일치도 평가

### 측정변동(measurement variability)과 측정오차(measurement error)

`-` 개인으로부터 측정되는 생물학적 변수: 일정량의 변동 존재
- **개인-간 변동(inter-individual variation)**
    - 원인: 개인과 관련된 요인(유전학적, 환경적, 생활 습관 등)들이 측정값에 영향
- **개인-내 변동(intra-individual variation)**
    - 원인: 측정 시점이 측정값에 영향
- **측정변동**: 측정값들 간 변동
- **측정오차**: 변수에 대한 관찰값과 실제 값 간의 차
    - **체계적 오차(systematic error)**
        - 외부 요인이 측정에 영향을 주어 관찰값이 너무 높(낮)게 측정되는 경우
        - 발생하면 편향된 추정값(biased estimate)가 얻어짐
        - 이로 인해 타당도 감소
        - 해결방안: 측정조건 표준화(standardization), 교육(education), 측정도구 눈금 교정(calibration)
    - **무작위 오차(random error)**
        - 관찰값은 실제 값보다 크거나 작을 수 있지만 평균적으론 균형 이룸
        - 우연에 의해(by chance) 발생

### 신뢰도(reliability)

`-` 일치해야만 하는 두 결과를 비교하고자 하는 경우 **일치도(agreement), 신뢰도(reliability)**를 평가, 이를 정량화 해야함
- **재현성(reproducibility)(관찰자/방법 일치도)**
    - 하나의 변수를 측정하는 두 방법이 동일한 결과를 제공하는가?
    - 같은 측정 밥법으로 두 명 이상의 관찰자들은 서로 동일한 결과를 제공하는가?
- **반복성(repeatability)**
    - 한 명의 관찰자가 같은 변수를 반복적으로 측정 시, 동일한 결과가 얻어지는가?
- 재현성, 반복성의 분석 방법은 동일
    - 변수가 **범주형**인지, **연속형**인지에 따라 방법이 결정

### 범주형 변수들

`-` 두 방법 간에 체계적인 오차가 있는가?
- 두 방법이 구조적 차이가 있다 - 둘 중 하나는 다른 방법에 비해 더 많은 환자들을 질병군으로 분류
    - 맥니마 검정(McNemar's test): 질병 분류 범주가 두 개 일때 각 비율 차이를 비교하는 검정 (오차의 존재 여부 평가)
    - 맥니마-바우커 검정(McNemar-Bowker test): 두 범주 이상일때
- 검정결과가 유의 X (non-significance): 두 방법 간 체계적 오차가 개입되었다는 증거 X
- 이때, 한 방법이 '참고표준(reference standard)'검사이면 다른 검사법은 **편향(bias)**되지 않는 검사라는 의미

`-` 일치도를 평가하기 위한 카파 통계량(kappa statistic)

```python
Image.open('8주차_img/img4.jpg')
```

- 평과결과는 이차원 분할표로 요약 가능
- 분할표의 **주대각선**에 있는 값: 두 방법이 일치하는 경우에 대한 빈도
- 각 칸에 대해 무작위로 만들어졌다고 가정하에 기대되는 **기대빈도** 계산
    - 카이제곱 검정 시 귀무가설 하에서의 기대빈도를 구하는 방법과 동일하게 사용
    - 각 칸의 행의 빈도 $\times$ 열의 빈도 / 전체 빈도
- **코헨의 카파(Cohen's kappa)** 통계량 계산: 두 관찰자(혹은 방법)의 평가가 우연히 일치할 가능성이 보정된 두 관찰자/방법 간의 일치도 지표
$$Cohen의 Kappa, \kappa = \frac{\frac{O_d}{m}-\frac{E_d}{m}}{1-\frac{E_d}{m}}$$
    - $m$ = 전체 관찰빈도
    > ex) 전체 환자 수
    - $O_d$ = 주대각선에 있는 관찰빈도들의 합
    - $E_d$ = 주대각선에 해당하는 기대빈도들의 합
    - 분모의 1은 두 관찰자/방법 평가가 완벽하게 일치할 때의 일치도 값
        - $\kappa = 0$: 우연에 의해 얻어진 결과
        - $\kappa\lt0.00$: 일치도가 낮음(poor)
        - $0.00\lt\kappa\lt0.20$: 일치도가 약간 있음(slight)
        - $0.20\lt\kappa\lt0.40$: 일치도가 어느 정도 있음(fair)
        - $0.40\lt\kappa\leq0.60$: 일치도가 적절하게 있음(moderate)
        - $0.60\lt\kappa\leq0.80$: 일치도가 상당히 있음(substantial)
        - $\kappa>80$: 일치도가 적절하게 있음(moderate)
        - $\kappa = 1$: 평가결과 완벽 일치
    - 카파 값에 대한 표준오차 및 신뢰구간 추정 가능: 하지만, 신뢰도 연구에서 카파값이 0인지의 여부를 검정하는 것은 무의미+비현실적
    - 카파 값은 범주의 수에 영향 받음(범주 수 많다 $\to$ 카파 값 낮아짐)
    - 카파 값은 검사의 유병률에서도 영양 받음: 서로 다른 연구의 카파 값들을 비교할 땐 주의를 요함        

`-` **가중 카파(weighted kappa)**
- 결과가 일치하는 정도(주대각선의 빈도), **불일치(disagree)**하는 정도(주대각선 외의 빈도)를 동시 고려
- 연속형 자료에서의 **급내상관계수(intraclass correlation coefficient)**와 유사한 개념값
