# chapter 22 수치형 자료: 세 집단 이상

### 1) 일원분산분석(One-way analysis of variance, One-way ANOVA

   ### 필요 가정

- 단일 요인(factor)의 각 수준(level)들(예를 들면 서로 다른 인종)이 집단으로 정의
- 관심 변수는 모집단에서 각 집단(수준)별로 정규분포
- 이들의 분산은 집단들 간에 서로 동일하다고 가정
- 표본 수는 이 가정들을 검사할 수 있을 정도로 충분히 큰 경우

### 이론적 배경  

- 자료의 전체 변동(total variation)을 개인들의 집단 간 차이에 기인하는 변동(집단-간 변동, between-group variation)과 각 집단 내에서의 개인들 간 무작위 변동(집단-내 변동*, within-group variation)으로 분리하는 방법
- 분석방법은 분산분석(analysis of variance, ANOVA) => 집단-간 변동과 집단-내 변동의 크기는 각기 분산으로 측정되기 때문

  ∴ 분산분석은 집단-간 분산과 집단-내 분산의 비(ratio)를 비교해 집단들 간에 모평균의 차이가 있는지를 검정

> 서로 다른 집단으로부터 얻어진 k개의 독립적인 표본 가정;
> - 표본 수: $n_{i}, i=1,2,..,k$
> - 표본평균: $\bar{x_{i}}, i=1,2,...,k$
> - 표본표준편차: $s_{i}, i=1,2,...,k$
> - 전체 표본 수: $n=n_{1}+n_{2}+\cdots+n_{k}$

---
1. 해당 연구에 관한 귀무가설과 대립가설을 설정  
  $H_{0}$: 모집단에서 집단들의 모평균은 같다. ($\mu_{1}=\mu_{2}=\cdots=\mu_{k}$)  
  $H_{1}$: 모집단에서 집단들의 모평균은 같지 않다. ($not H_{0}$) 
2. 각 표본으로부터 자료를 수집하고 적절한 통계량들을 계산  
3. $H_{0}$에 해당하는 검정통계량 값을 계산  
  - $F=\displaystyle\frac{집단-간 분산}{집단-내 분산}\Rightarrow$  분산분석(ANOVA)  
4. 검정통계량 값을 알려진 확률분포의 값과 비교  
5. P-value를 해석하고 결론을 내림  
  - 결과가 유의하다면 집단들끼리 짝을 지어 어느 집단들 간에 평균 차이가 있는지 사후검정(post-hoc analysis) 실시   
    $\Rightarrow$ 다중비교(multiple comparison)  
---

### 가정이 만족되지 않는 경우

- 분산분석을 실시하기 전,  먼저 자료의 정규성을 검토
- 집단들 간의 분산의 크기가 서로 유사한지 여부 확인
- 만일 이 가정들이 만족되지 않을 때 자료를 변환(ex. sqrt 변환, log 변환 등)
- 비모수적 방법인 크루스칼-왈리스 검정(Kruskal-Wallis test) 실시  
  
    
    

  
    
### 2) 크루스칼-왈리스 검정(Kruskal-Wallis test)

> - 윌콕슨의 순위합 검정 방법을 k개 집단으로 확장한 방법
> - 집단들 간 분포 차이가 없다는 귀무가설이 사실인 경우, k개 집단들의 순위합은 서로 비슷(단, 집단들의 표분 수 차이를 보정하는 경우)

---
1. 해당 연구에 관한 귀무가설과 대립가설을 설정  
 $H_{0}$: 모집단에서 집단들의 분포는 같다. ($median_{1}=median_{2}=...=median_{k}$)  
 $H_{1}$: 모집단에서 집단들의 분포는 같지 않다. (not $H_{0}$)  
2. 각 표본으로부터 자료를 수집하고 적절한 통계량들을 계산  
3. $H_{0}$에 해당하는 검정통계량 값을 계산  
  - 모든  $n(=\Sigma^k_{i=1} n_{i}$)개의 값들을 대상으로 순위를 매기고, 집단 별로 수위의 합을 계산  
  => 이 순위 합들을 $R_{1}, R_{2}, ..., R_{k}$  
  - $H=\displaystyle\frac{12} {n(n+1)} {\Sigma ^{k}_{i=1}}\frac{R^2_{i}}{n_{i}}-3(n+1)$  
4. 검정통계량 값을 알려진 확률분포의 값과 비교  
5. P-value를 해석하고 결론을 내림  
  - 결과가 유의하면 집단들끼리 짝을 지어 두 표본 비모수 검정, 즉 윌콕슨 순위합 검정을 실시한 후, 다중 가설검정의 문제점을 보정하기 위해 본페로니 수정을 적용
---

# Chapter 23 범주형 자료: 단일 비율

### 1) 단일 비율에 관한 검정(test of single proportion)

> - 표본은 모집단에서 무작위로 추출  
> - $p=\frac{r}{n}$  
> - 모집단에서의 비율: $\pi$  
> - $p$는 근사적으로 평균이 $\pi$, 표준편차가 $\sqrt\frac{\pi(1-\pi)}{n}$ 정규분포
> - 이 정규분포의 평균에 대한 추정값 $p$, 표준편차에 대한 추정값 $\sqrt\frac{p(1-p)}{n}$

---
1. 해당 연구에 관한 귀무가설과 대립가설을 설정  
  $H_{0}$: 모비율 $\pi$는 특정 값 $\pi_{1}$일 것이다. ($\pi=\pi_{1}$)  
  $H_{1}$: 모비율 $\pi$는 특정 값 $\pi_{1}$이 아닐 것이다. ($\pi \not= \pi_{1}$)  
2. 표본으로부터 자료를 수집하고 적절한 통계량들을 계산  
3. $H_{0}$에 해당하는 검정통계량 값을 계산  
  $Z=\displaystyle\frac{|p-\pi_{1}|-\frac{1}{2n}}{\sqrt{\frac{\pi_{1}(1-\pi_{1})}{n}}}$  
4. 검정통계량 값을 알려진 확률분포의 값과 비교  
5. P-value를 해석하고 결론을 내림  
  - P-value를 해석하고, 모비율 $\pi$에 대한 신뢰구간을 계산  
  - $\pi$에 대한 근사적 95% 신뢰구간: $p±1.96\sqrt{\frac{p(1-p)}{n}}$
---

### 2) 부호 검정을 비율에 적용

> - 교차설계 임상시험에서 처리 A와 처리 B 중 환자들이 선호하는 방법은 어느 것인가?  
> - 선호도의 차이가 없다면 처리 A를 선호하는 비율은 1/2가 될 것이며, 따라서 부호 검정을 사용해 평가할 수 있다.

---
1. 해당 연구에 관한 귀무가설과 대립가설을 설정  
  $H_{0}: 모집단에서 방법 A를 선호하는 비율 \pi는 \frac{1}{2}이다. \pi=\frac{1}{2}$  
  $H_{1}: 모집단에서 방법 A를 선호하는 비율 \pi는 \frac{1}{2} 아니다. \pi\not=\frac{1}{2}$  
2. 각 표본으로부터 자료를 수집하고 적절한 통계량들을 계산    
3. $H_{0}$에 해당하는 검정통계량 값을 계산  
  - A방법과 B방법 간 선호도 차이가 없는 사람들을 제외하면 표본 수는 n에서 n'로 줄어든다.  
  ∴ $p=\frac{r}{n'}$, r: 방법 A를 선호하는 사람들 수  
    - $n'≤10$, 검정통계량: r  
    - $n'＞10$, 검정통계량: $\displaystyle z'=\frac{|p-\frac{1}{2}|-\frac{1}{2n'}}{\sqrt{\frac{0.5(1-0.5)}{n'}}}$  
    - z'은 정규분포를 따름  
4. 검정통계량 값을 알려진 확률분포의 값과 비교  
5. P-value를 해석하고 결론을 내림  
  - P-value를 해석하고, 신뢰구간을 계산
---
