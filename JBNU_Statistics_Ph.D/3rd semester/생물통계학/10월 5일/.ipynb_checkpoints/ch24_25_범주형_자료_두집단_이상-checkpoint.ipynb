# Part5. 기초적인 자료분석 방법

## Chapter24 범주형 자료 : 두 비율

### 문제 제기

- 서로 독립적인, 서로 관련이 없는 두 집단의 경우

>ex) 입질에 걸렸던 경험이 있는 동성애 남성 집단, 임질 과거력이 없는 동성애 남성 집단

>이들 두 집단 간에 어떤 특성(ex_인간 허피스 바이러스-8)을 가진 비율이 같은지의 여부를 파악하고자 함
  
  
- 서로 관련이 있는 두 집단의 경우
>ex) 연구대상자들을 짝짓기 한 경우 or 서로 다른 시점에서 두 번 측정되는 경우 (즉, 처리 전과 후)

>이들 두 집단 간에 어떤 특성(ex_검사 수치가 높아졌는지의 여부)에 관한 비율이 동일한지를 조사하고자 함

### 독립적인 두 집단 : 카이제곱 검정(chi-squared test, $\chi^2$-test)
 - 용어
 
   - 분할표(contingency table) : 위와 같이 각 칸에 있는 값이 빈도를 나타내는 표
   - 2 $\times$ 2 분할표(2 $\times$ 2 table) : 분할표의 행이 2개이고 열도 2개인 분할표
  #### **표24.1**관찰 빈도
  
|**특성**|**집단1**|**집단2**|**합계**|
|---|---|---|---|
|**있음**|a|b|a+b|
|**없음**|c|d|c+d|
|**합계**|n1=a+c|n2=b+d|n=a+b+c+d|
|---|---|---|---|
|**특성을 가진 비율**|$$p_{1}=\frac {a} {n_{1}}$$|$$p_{2}=\frac {b} {n_{2}}$$|$$p=\frac {a+b} {n}$$|

 
 
   - 각 행과 열 조합에 따른 4개 칸(cell)에 **관찰빈도(observed frequency)**가 주어져 있음
   - 4개의 **주변 합(marginal total)** 및 **전체 합(overall total)** n도 표시되어 있음
   
 4개 칸 각각에 대해서 $H_0$가 사실인 경우에 기대되는 **기대빈도(expected frequency)**를 계산

 - 필요한 가정
 
   - 표의 행과 열들은 **상호 배타적(mutually exclusive)**, 즉, 각 사람은 단지 하나의 행과 하나의 열에만 속하게 됨
   - 일반적인 카이제곱 분석방법을 사용하기 위해서는 `4개의 칸 각각 기대빈도가 최소 5 이상이라는 가정이 필요함`

**1. 해당 연구에 관한 귀무가설과 대립가설을 설정**

$H_0$ : 모집단에서 해당 특성을 가진 사람들의 비율은 서로 같다 (즉, $\pi_{1} = \pi_{2}$)

$H_1$ : 모집단의 비율은 같지 않다 (즉, $\pi_{1} \neq \pi_{2}$)

**2. 표본으로부터 자료를 수집하고 적절한 통계량들을 계산**

자료로부터 검정에 필요한 통계량들(기대빈도,*E*)을 계산

**3. $H_0$에 해당하는 검정통계량 값을 계산**

### $\chi^2 =  \sum \frac{(|O-E|-\frac{1}{2})^2} {E}$

여기서 O와 E는 각각 2 $\times$ 2 표의 4개 칸에 대한 관찰빈도와 기대빈도를 의미

  `예를 들어` 첫 번째 칸의 경우, 관찰빈도는 a이고 기대빈도는 $\frac {(a+b)(a+c)}{n}$임
  
  O - E 앞 뒤에 있는 수직선은 절대값(즉, 부호를 무시하는 것)을 의미
  
  분자 부분의 $\frac {1}{2}$ 은 연속성 수정을 위한 값
  
  이 검정 통계량은 자유도가 1인 카이제곱 분포에 따르게 됨
  
**4. 검정통계량 값을 알려진 확률분포의 값과 비교**

계산된 $X^2$ 값을 부록 표 A3의 자유도 1에 해당하는 카이제곱 분포 백분위 값과 비교

**5. P-value를 해석하고 결론을 내림**

얻어진 P-value를 해석하고, 두 집단 간모비율 차이 $\pi_{1}$ - $\pi_{2}$에 대한 신뢰구간을 계산

$\pi_{1}$- $\pi_{2}$에 대한 근사적 95% 신뢰구간은 다음과 같음

$\displaystyle (p_{1}-p_{2})\pm 1.96 \sqrt{\frac{p_1(1-p_1)}{n_1} + \frac{p_2(1-p_2)}{n_2}}$

이 신뢰구간을 사용해서 연구결과의 임상적/보건학적 중요성을 평가

신뢰구간이 넓다는 것은 표본에서 얻어진 값 $p_{1} - p_{2}$ 가 낮은 정밀성을 가진 추정값이라는 의미

신뢰구간이 0을 포함하고 있지 않으면 두군의 모비율은 통계적으로 유의하게 다르다는 의미

 - 가정이 만족되지 않는 경우
 
   - 최소한 하나의 칸에서 E < 5이면
   - 카이제곱 검정 대신 정확한(exact) P-value를 계산해주는 **피셔의 정확검정(Fisher's exact test)** 을 실시하는 것이 좋음

 - 2 $\times$ 2 분할표의 결합
   
   - 각기 다른 연구에서 얻어지는 분할표들을 합하고자 할 때 서로 대응되는 각 칸들의 빈도를 단순히 합해서는 안됨
   
   > ex) 남성/여성과 같이 하나의 모집단 내 서로 다른 부집단들 or 한국/일본과 같이서로 다른 모집단
   
   - 자료들을 단순히 합해서 얻어지는 표에 대해 카이제곱 검정을 실시하면 **심슨의 역설(Simpson's paradox)** 이 발생될 수 있음
   
   - 심슨의 역설(Simpson's paradox) : 부집단 자료들을 합해서 하나의 집단으로 만드는 경우 연관성의 방향이 바뀌는 것

### 서로 관련이 있는 두 집단 : 맥니아 검정(McNemar's test)

  #### **표24.2**특성 보유 여부에 따른 각 짝의 관찰 빈도

||**상황2**|||
|---|---|---|---|
|**상황1**|**있음**|**없음**|**합계**|
|**있음**|w|x|w+x|
|**없음**|y|z|y+z|
|**합계**|w+y|x+z|m=w+x+y+z|

 - 필요한 가정
 
서로 다른 상황 하에 동일한 사람들을 대상으로 같은 특성을 두 번 측정하는 경우

혹은 동일한 특성을 가진 사람끼리 짝을 짓는 경우 등과 같이 서로 관련이 있거나 또는 서로 종속적인 두 집단이 있고

각 사람들은 해당 특성을 두 상황을 모두 가지고 있거나, 어느 한쪽만 가지고 있거나, 아님 둘 다 가지고 있지 않는 경우 등으로 분류됨

**1. 해당 연구에 관한 귀무가설과 대립가설을 결정**

$H_{0}$ : 두 짝진 군의 모집단 내 관심 특성을 가진 사람들 비율은 같다 (즉, $\pi_{1}$ = $\pi_{2}$)

$H_1$ : 두 짝진 군의 모집단 비율은 같지 않다 (즉, $\pi_{1} \neq \pi_{2}$)

**2. 각 표본으로부터 자료를 수집하고 적절한 통계량들을 계산**

자료로부터 검정에 필요한 통계량들(두 종류 불일치 짝의 수, 즉 x와 y)을 계산

**3. $H_{0}$에 해당하는 검정통계량 값을 계산**

$\displaystyle \chi^2 = \frac{(|x-y|-1)^2}{x+y}$

여기서 x는 (진단1, 진단2) = (있음, 없음)인 불일치 짝의 수, 그리고 y는 (진단1, 진단2) = (없음, 있음)인 불일치 짝의 수를 의미

이 검정통계량은 자유도가 1인 카이제곱 분포를 가진다

분자 부분의 값1은 연속성 수정을 하기 위한 값

**4. 검정통계량 값을 알려진 확률분포의 값과 비교**

계산된 $\chi^2$ 값을 부록 표 A3의 자유도 1에 해당하는 카이제곱 분포 백분위 값과 비교

**5. P-value를 해석하고 결론을 내림**

얻어진 P-value를 해석하고, 두 집단 간의 모비율 차이 $\pi_{1} - \pi_{2}$ 에 대한 신뢰구간을 계산

$\pi_{1} - \pi_{2}$ 에 대한 근사적 95% 신뢰구간은 다음과 같다

$\displaystyle\frac{x-y}{m}\pm\frac{1.96}{m}{\sqrt{{x+y}-\frac{(x+y)^2}{m}}}$

## 범주형 자료:세 범주 이상

### 카이제곱 검정:큰 분할표의 경우

### - 문제 제기

 요인 2개를 사용해 대상자들을 구분하는 경우

  ex) 한 요인은 질병의 중중도(경증, 중등증 및 중증)에 관한 것, 다른 요인은 혈액형(A, B, O, AB)임
  
  이 두 요인이 서로 관련이 있는가 --- 질병의 중증도는 혈액형에 따라 다를까?

### - 필요한 가정

표 안의 값들은 빈도를 나타낸다

요인의 범주들은 서로 배타적

전체 칸 중 80% 이상은 기대빈도가 5보다 크거나 같다

|---|---|---|---|---|---|---|
|행1|$$f_{11}$$|$$f_{12}$$|$$f_{13}$$|...|$$f_{1c}$$|$$R_{1}$$|
|행2|$$f_{21}$$|$$f_{22}$$|$$f_{23}$$|...|$$f_{2c}$$|$$R_{2}$$|

**1. 해당 연구에 관한 귀무가설과 대립가설을 설정**

$H_0$:모집단에서 한 요인의 범주들과 다른 요인의 범주들 간에는 상호 관련이 없다

##### (즉, $\pi_{ij}= \pi_{i.}\pi_{.j},\quad i = 1,2,...,r;j = 1,2,...,c$)

$H_1$:모집단에서 두 요인들은 서로 관련이 있다 (즉, $\pi_{ij}\neq \pi_{i.}\pi_{.j}$)

**2. 표본으로부터 자료를 수집하고 적절한 통계량들을 계산**

자료로부터 검정에 필요한 통계량들(기대빈도, *E*)을 계산

**3. $H_0$에 해당하는 검정통계량 값을 계산**

#### 검정통계량은 $\chi^2 =  \sum \frac{(O-E)^2} {E}$

여기서 *O*와 *E*는 각각 2 $\times$ 2 표의 4개 칸에 대한 관찰빈도와 기대빈도를 의미

이 검정통계량은 자유도가 (r-1) $\times$ (c-1)인 카이제곱 분포에 따름

자유도가 1보다 큰 경우에는 범주형 자료를 연속형 분포인 카이제곱 분포로 근사하는데 큰 문제가 발생되지 않았으므로 연속성 수정을 할 필요는 없다

**4. 검정통계량 값을 알려진 확률분포의 값과 비교**

계산된 $\chi^2$ 값을 부록 표 A3의 자유도 (r-1) $\times$ (c-1)에 해당하는 카이제곱 분포 백분위 값과 비교

**5. P-value를 해석하고 결론을 내림**

## 카이제곱 추세 검정(chi-squared trend test)

### 문제 제기

요인하나는 두 범주(ex> 특성 보유 유무 등)이고, 다른 요인은 상호 배타적이며 순서의 의미를 가지고 있는 k개 범주(즉, 2 $\times$ k 분할표, 표25.2)

이들 간의 상호 관련성을 조사해보자

이 경우 첫 번째 요인의 특성에 관한 비율이 두 번째 요인의 각 범주에 따라 증가 혹은 감소하는 추세가 있는지를 평가하고자 하는 것일 수 있음

ex) 치료에 대한 반응 유무, 4개의 연령구간(65 ~ 69, 70 ~ 74, 75 ~ 79, 80 이상)으로 이루어졌다 할 경우

연령대가 증가함에 따라 치료의 반응률도 증가하는지의 여부가 관심사항일 수 있음

> 참고로 두 요인 간 일반적 관련성에 관한 카이제곱 검정 결과가 유의하지 않더라도 카이 제곱 추세 검정 결과는 유의할 수 있음

**1. 해당 연구에 관한 귀무가설과 대립가설을 설정**

$H_0$:모집단 내 관심 특성 비율들은 추세(경향성)가 없다

$H_1$:모집단 내 비율들 간에는 어떤 추세가 있다

**2. 표본으로부터 자료를 수집하고 적절한 통계량들을 계산**

k개의 열 범주 각각에 대해 행 범주의 특성 비율을 추정하고, 각 열 범주에 점수(score)를 할당함

일반적으로 점수는 1,2,3,...,k 등과 같은 일련의 자연수들을 사용함

그러나 각 열 범주의 내용에 따라 해당 범주의 상대적인 중요도를 반영하는 값들

>ex) 범주화 된 각 연령 구간의 가운데 값

또는 연구자가 평가하고자 하는 경향성

>ex) 선형적 혹은 곡선적 추세

등을 같이 등간격(equally spaced)의 점수를 사용하면 *선형 추세(linear trend)* 여부를 평가하게 됨

반면에 *2차 곡선 추세(quadratic trend)* 를 파악하기 위해서는 1,4,9,...,$k^2$ 등과 같은 점수를 사용

**3. $H_0$에 해당하는 검정통계량 값을 계산**

검정통계량 

### $\chi^2_{trend}$ = $\frac{(\sum_iw_if_{1i}-R_1 {\sum_1\frac {w_iC_i}{n}})^2}{\frac{R_1}{n}(1-\frac{R_1}{n})[{\sum_iw^2_iC_i-n(\frac{\sum_iw_iC_i}{n})^2}]}$

여기서 합($\sum_1$) 기호는 각 칸 별로 계산된 값들을 k개 열 범주 전체에 대해 더한다는 의미. 이 검정 통계량은 자유도가 1인 카이제곱 분표 따름

**4.검정통계량 값을 알려진 확률분포의 값과 비교**

계산된 $\chi^2_{trend}$ 값을 부록 표 A3의 자유도 1에 해당하는 카이제곱 분포 백분위 값과 비교

**5. P-value를 해석하고 결론을 내림**

얻어진 P-value를 해석하고, k개 비율 각각에 대한 신뢰구간 계산
