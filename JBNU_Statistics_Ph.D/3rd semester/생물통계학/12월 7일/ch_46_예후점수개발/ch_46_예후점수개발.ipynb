## CH 46. 예후점수 개발

### 1) 왜 예후점수를 개발하는가?

 개인의 인구학적 또는 임상적 특성들을 알고 있는 상태에서 관심사건을 경험하게 될 가능도(likehood)를 예측(predict)하고자 하는 경우가 있음
 
 `사건이란` 긍정적인 결과 (예, 수술결과가 좋음, 치료됨 등)
 
          부정적인 결과 (예, 질병에 걸림, 사망함 등)
          
**예후점수(prognostic score)** 또는 **예후지수(prognostic index)** : 사건을 경험하게 될 가능성을 점수화한 척도

**위험점수(risk score)** : 부정적인 사건을 예측하는 경우  

예후점수를 생성(generate) 방법
- 가장 단순한 방법
  - 개인이 갖고 있는 위험요인들의 개수를 개인의 예후점수로 사용
  - 해당 사건의 발생에 관한 위험요인들이 잘 알려진 경우 사용 
  > ex) 심혈관 질환 발생여부에 대해서는 잘 알려진 위험요인들인 남성, 고연령, 흡연, 심혈관계 질환 가족력, 당뇨, 고지혈증, 고혈압 중 몇 개나 가지고 있는지를 의미
  
  - 점수는 한 개인이 해당 사건을 경험하게 될 위험에 관한 일차적인 지표로 사용 가능
  
   $\rightarrow$  점수가 높다는 것은 심혈관 질환 발병위험이 높다는 것을 의미
  - 이러한 접근방식은 사건발생에 관한 각요인의 영향력 크기가 모두 같다는 것을 전제로 함

- 일반적인 방법
  - 사건발생과 유의한 관련이 있는 요인들과 이 요인들이 사건을 경험할 가능성을 결정하는 데 상대적으로 얼마나 더 중요한 역할을 하는지 평가할 수 있는 통계적 방법 
   > ex) 로지스틱 회귀나 이와 유사한 판별분석(discriminatanalysis) 등을 사용하는 것이 좋음
  - 각 개인에 대한 예후점수는 모형을 통해 추정되는 회귀계수들을 사용해 요인들의 가중합
   > ex) 30장의 logit(p) = a + $b_1x_1$ + $b_2x_2$ + ... + $b_kx_k$ 값 형태로 계산됨
  - 이 점수의 범위는 해당 점수가 어떤 방식으로 생성되었는지에 따라 달라지지만, 일반적으로 점수가 높다는 것은 사건을 경험하게 될 가능성이 더 크다는 것을 의미
  - 점수를 사용해 각 개인이 사건을 경험하게 될 가능성을 '낮음','중간','높음' 등으로 범주화가 가능
  - 로지스틱 회귀분석을 사용한다면 개인별 점수를 변환해 사건을 경험하게 될 예측확률을 추정할 수도 있으며
   이 값은 확률이므로 0과 1 사이의 값을 가짐

- 예후점수를 생성하기 위해 회귀모형을 사용하는 경우
 - 해당 모형이 자료의 변동량을 많이 설명해준다고 해서 (즉, 모형의 $R^2$ 값이 크다고 해서) 반드시 개인이 질병에 걸릴 가능성도 잘 예측한다고 할 수는 없음
 - 알려진 위험요인들을 기초로 생성된 예후점수라할지라도 개인 별 예후에 대해서 정확하지 않을 수 있음
 - 따라 모형을 사용해 예측점수를 개방할 때는 반드시 모형으로부터 예측된 점수들의 타당성 여부가 평가되어야 함

### 2) 예후점수의 수행력(dperformance) 평가
#### 1. 예후점수가 얼마나 정확(accurat)한가?

이는 점수가 사건을 얼마나 정확히 예측해 주는지에 대해 평가하는 것을 말함
- 점수가 사건이 얼마나 맞게 혹은 틀리게 예측했는지에 대한 **2 $\times$ 2 분류표(classification table)**, 즉, 유사한 형태이 표를 만든 뒤, 표를 사용해 다음의 척도들을 계산
 - 민감도(sensitivity)와 특이도(specificity)
 - 전체 정확도(total accuracy): 전체 인원 수 중 사건을 경험한 경우 및 경험하지 않은 경우가 정확히 예측된 비율. 이 값이 1에 가까울수록 전체 정확도가 높은 점수임
    + 개인별 사건을 100% 정확히 예측해주면 점수는 완벽한 점수(perfect score)라고 함
- 로지스틱 회귀를 사용해 점수를 생성하는 경우 표본 내 n명의 사람들에 대한 **평균 브라이어 점수(mean Brier score)** 를 계산할 수 있음

#### 2. 예후점수가 사건의 경험 여부를 얼마나 잘 판별(discriminate)해주는가?

각 개인이 사건을 경험하게 될 가능성의 크기에 따라 순위를 매겨 점수의 판별력을 평가함
- 개인들의 점수를 범주로 구분한 다음
 > ex) 백분위 수에 근거해 5~10개의 동일 크기 구간들로 나눈 다음

 각 범주 내에 사건 발생률을 계산함
- 점수 구간이 증가함에 따라 사건 발생률 역시 증가하는 경향을 보이면 점수의 판별력은 좋은 것
- **ROC 곡선(수신자 작업특성 곡선)**, 즉, 민감도와 (1-특이도)에 관한 산점도를 그린다
 - 곡선이 그림의 왼쪽 윗부분으로 그려질수록 점수의 판별력이 좋다는 의미
 - 45도 대각선에 가까우면 점수가 사건을 무작위로 판별하는 것과 다를 바 없다는 의미
 - **ROC 곡선 아래 면적(AUROC 또는 AUC)** 은 사건 경험 여부를 예측하는 점수의 판별력 크기를 나타냄
- **해럴의 c-통계량(Harrell's c-statistic)** 을 계산함

#### 3. 점수는 바르게 교정(calibrate)되었는가?
- 예측확률을 로지스틱 회귀를 사용해 계산하는 경우 사건의 발생여부를 예측한 확률이 관찰값(즉, 0 또는 1)과 얼마나 잘 일치하는지 확인
 - 사건 경험 여붕 대한 예측점수의 판별력이 높다고 하더라도, 사건발생 위험을 예측하는 예측력은 낮을 수 있음
 - 이런 상황은 예측점수를 해당 예후점수가 구축된 모집단이 아닌 다른 모집단에 적용할 때 자주 발생하게 됨
 > 조금이라도 예측확률을 사용해 임상적인 의사결정을 하고자 하면, 교정이 잘 되지 않은 예후점수를 사용하는 것은 곧 환자를 부적절하게 치료하는 것을 의미할 수 있음, 예후점수의 예측력을 높이기 위한 교정작업은 매우 중요함
- 모형교정이 필요한지를 판단하기 위해서는 관찰사건의 확률과 이를 점수로 예측한 확률간 일치도를 평가하는 **호스머-르머쇼우 적합도 검정(Hosmer-Lemeshow goodness of fit test)** 을 실시함

#### 4. 점수는 이동가능(transportable)한가, 또는 일반화가 가능(generalizable)한가?
- 예후점수가 해당 점수가 개발된 모집단이 아닌 다른 모집단에 대해서도 잘 적용되는지 알아야 할 때가 있음
 - 모형개발에 사용된 자료와 동일한 자료를 사용해 점수의 타당성을 평가하는 것을 **내적타당도 평가(internal validation)** 라고 함
 - 하지만 점수의 실제 수행력을 평가하기 위해서는 최소한 하나 이상의, 모형개발에 사용된 자료와는 독립적인, 자른 자료를 사용해 점수의 타당성 여부를 평가할 필요가 있음
 - 이를 **외적타당도 평가(external validation)** 라고 함

외적타당도 평가를 실시하는 것이 비현실적인 상황에서는 이에 대한 대안으로 다음과 같은 내적타당도 평가들을 실시할 수도 있음
- 자료를 두 개의 부집단으로 나눈다
 - 점수를 개발하는 데 사용되는 **평가용 표본(training sam-ple)** 
 - **검증용 표본(validation sample)**  : 점수의 타당성을 평가하는 데 사용되는 표본
 - 평가용 표본과 검증용 표본의 선택은 무작위로 진행하되, 일반적으로 검증용 표본보다는 평가용 표본을 더 크게 함
 > (ex, 평가용 표본이 전체 표본의 70%가 되도록 함)
- 자료를 여러 개의 부집단으로 분할해서 **교차타당도 평가(cross-validation)** 를 실시함
 - 이는 하나의 부집단을 사용해 예후점수를 생성한 후, 나머지 부집단들을 사용해 생성된 점수에 대한 타당성을 평가하는 작업임
- 예후점수 추정 및 그 수행력 평가력 위해 **붓스트래핑(Bootstrapping)** 을 실시할 수도 있음
- 다기관 연구로 예후점수를 개발한다면 **내적-외적 교차타당도 평가(internal-exter-nal cross-validation)** 를 실시할 수도 있음
 - 이것은 각 기관의 자료를 제외하면서 예후점수를 개발하는 방법임
