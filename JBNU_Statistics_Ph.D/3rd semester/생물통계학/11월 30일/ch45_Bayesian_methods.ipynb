## Ch 45 베이지안 방법

#### 빈도주의적 접근

실험을 무한히 반복한다고 가정하는 경우에 발생할 것으로 예상되는 사건의 수에 관한 개념을 반영하는 접근 방식

단점
1. 현재 연구와 관련된 정보 외에 우리가 관심효과와 관련해 가진 정보는 사용하지 않음.
    > ex) 임상시험을 실시하기 전에 전문의가 가지고 있는 두 치료요법의 상대적 효용에 관한 임상적 소견 등
         두 약제를 비교할때, 두 약제의 효과가 동일할 경우에 현재와 같은 결과가 관찰된 가능성에 대한 값을 나타냄
         but, 실제 관심사항은 어느 약제가 더 효과가 있는지 파악하는 것임.
2. 결과가 함축하고 있는 의미보다는 가설검정의 역할 및 결과의 통계적 유의성 여부가 지나치게 강조되는 경향 존재

$\downarrow$  대안

#### 베이지안적 접근

가설에 관한 이미 존재하는 정보에 관한 신념을 반영하는 접근 방식 $\to$ 새로운 정보가 추가되면 이 신념이 변할 수 있음
- 새로운 자료가 확보되면 이를 토대로 가설에 관한 사전 의견을 갱신해서 해당 가설이 사실일 확률을 계산하는 방법

`-` **조건부 확률**

베이지안 분석의 근간, 어떤 사건($B$)이 이미 발생했다는 조건 하에 다른 사건($A$)이 발생할 확률

$$P(A|B)=\frac{P(A and B)}{P(B)}=\frac{P(B|A)P(A)}{P(B)}$$

`-` **베이즈 정리**

- 연구를 수행하기 전에 해당 가설에 대해 가지고 있는 개인적 신념인 **사전확률**울 연구 수행 후의 개인적 신념에 해당하는 **사후확률**로 변환하기 위한 이론
    - 여기서 사후확률은 연구결과가 얻어졌다는 조건하에서의 가설에 대한 조건부 확률
- 베이즈 정리에 의하면 사후확률은 사전확률과 **가능도**의 곱에 비례
    - 가능도: 귀무가설이 사실이라는 가정 하에 현재와 같은 결과가 관찰될 가능성
    - $H_0$=어떤 환자가 질병을 가지고 있음

**조건부확률을 이용한 베이즈 정리 설명**

- 어떤 표본공간이 상호 배타적인 두 사건 $E_1$과 $E_2$ 로만 이루어졌다고 가정 (즉, $P(E_1\cap E_2)=0$)
- 어떤 사건 $C$가 주어진 경우에 사건 $E_1$이 발생할 조건부 확률
\begin{align*}
    P(E_1|C)&=\frac{P(C|E_1)P(E_1)}{P(C)} =\frac{P(C|E_1)P(E_1)}{P(C \cap E_1)+P(C \cap E_2)}= \frac{P(C|E_1)P(E_1)}{P(C|E_1)P(E_1)+P(C|E_1)P(E_2)}\\
    & \propto P(C|E_1)P(E_1)
\end{align*}
- 자료에 기초한 가능도 $P(C|E_1)$에 사건이 발생할 것으로 믿어지는 사전확률 $P(E_1)$를 추가적으로 사용해서 해당 사건이 발생할 사후확률 $P(E_1|C)$을 계산할 수 있다는 이론

`-` **베이지안 측면에서의 진단검사**

- 임상의는 병원에 내원한 사람이 질병에 걸려 있을 가능성(사전확률(=사전검사확률))에 대한 개인적인 견해를 가지고 있음.
    - 해당 환자에 대한 사전정보가 전혀 없는 경우라면 사전확률로 해당 질병에 대한 인구집단 유병률을 사용함.
    > 유병률: 어느 한 시점에 특정 인구집단 또는 지역에서, 질병을 가지고 있는 인구의 수를 대응되는 전체 인구의 수로 나눈 것
- 사전확률을 사후확률을 바꾸기 위해 베이즈 정리 사용
- 가장 쉽게 베이즈 정리를 사용할 수 있는 방법: 최근의 조사결과로부터 얻어진 정보에 기초한 **가능도 비** 계산하는 방법

- 양성 가능도 비,$LR(+)$
    - 검사결과가 양성으로 나오게 될 가능도 비는 실제 질병에 걸린 환자의 검사결과가 양성일 확률을 실제로는 질병에 걸리지 않은 환자의 검사 결과가 양성일 확률로 나눈 값

질병의 진단 시 베이즈 정리는 주로 양성 가능도비와 질병발생 오즈를 사용해 다음과 같은 형태로 활용

$$\mbox{odds}(D|T)=LR(+) \times \mbox{odds}(D), D: 질병 있음, T: 검사 결과 양성임$$

$$사전오즈=\frac{사전확률}{1-사전확률}$$

$$사후확률=\frac{사후오즈}{1+사후오즈}$$

- 사후확률(=사후검사 확률): 환자의 검사결과가 양성으로 나왔을 때 그가 실제로 질병에 걸렸을 가능성
    - 사후확률은 양성예측도와 비슷한 개념
        - 양성예측도는 임상의가 환자들의 정확한 질병 상태인 참고표준 검사결과를 모른다면 계산할 수 없음
        - 참고표준 검사결과를 알아도, 양성예측도에 영향을 미치는 주된 요인이 유병률인데, 질병에 대한 각 개인의 기저위험은 일반 인구집단보다 더 높거나 혹은 낮을 수 있기 때문에, 비록 임상의가 이 값을 계산한다고 하더라도 검사 결과를 토대로 환자가 질병에 걸렸는지를 판단하기 위한 믿음의 정도로 사용되기에는 좋은 척도가 아닐 수 있음
     > 즉, 참고표준 검사결과를 알 수 없거나, 환자의 기저위험 정도를 알 수 있는 경우에는 사후확률을 계산하는 것이 좋음.

- **페이건의 노모그램** : 사전확률 값과 가능도 비 값을 서로 연결한 뒤, 해당 직선을 확장해서 사후확률 값을 추정

`-` **베이지안 방법론의 단점**

- 베이지안 분석에 명시되어야 하는 가설에 대한 사전확률은 주관적으로 판단되어야 하기 때문에 결정이 쉽지 않을 수 있음 $\to$ 임의적인 방법론이라는 비판
- 최신 연구결과에 대한 근거가 강할수록(가능도 비가 클수록), 사전확률이 사후확률에 미치는 영향력의 크기는 줄어듦

#### 예제

- (38장)
    - 20세 이하 소아청소년 간염환자의 혈액 내 세룰로플라스민 농도 <16.6mg/dL 기준을 사용할 때, 윌슨병을 예측하는 민감도가 91.2%, 특이도가 94.9%, 양성 가능도 비는 17.9

- 인구집단 내 20세 이하 소아청소년 간염환자의 윌슨병 유병률이 6.4%라 믿는다면, 이 환자들의 윌슨병 사전확률로 0.064를 가정할 수 있음

> $사전오즈 = \frac{0.064}{1-0.064} = \frac{0.064}{0.0936} = 0.068$

$\mbox{사후오즈} = 0.068 \times \mbox{가능도비} = 0.068 \times 17.9 = 1.22 $

$\mbox{사후확률}= \frac{1.22}{1+1.22} = \frac{1.22}{2.22} = 0.550$

- 어떤 환자의 세룰로플라스민 농도가 16.6mg/dL 이하이고, 이 환자의 윌슨병 사전확률로 0.064를 사용한다면, 해당 환자가 윌슨병을 가지고 있을 확률(사후확률)은 55%.

- 페이건의 노모그램
    - 사전검사 확률 축의 6.4%와 가능도 비 축의 17.9에 해당하는 점을 직선으로 연결 후 사후검사 확률 축과 만나는 점(55%) 찾기
       > 윌슨병 유병률이 2%라고 한다면, 환자의 사후검사 확률은 27%

- 두 경우 모두 사전검사 확률에 비해 사후검사 확률이 큼.
- 20세 이하 소아청소년 간염환자의 윌슨병 보유 여부에 대한 지표로 세룰로플라스민 농도 <16.6mg/dL 기준을 사용하는 것이 효과적이라는 것을 의미
