## 11주차 과제



`-` 지진자료에서 1985년 이후의 자료만 고른뒤에 `HeatMapWithTime()`을 이용하여 시각화하라.

```python
import pandas as pd
import numpy as np
import folium
from folium import plugins 
```

```python
df=pd.read_csv('https://raw.githubusercontent.com/plotly/datasets/master/earthquakes-23k.csv')
df
```

```python
list1=df.assign(Year1=list(map(lambda x: x.split('/')[-1], df.Date)))
```

```python
list2=list1.assign(Year=list(map(lambda x: x.split('-')[0], list1.Year1)))
```

```python
df1=list2.iloc[:,[0,1,2,3,5]]
```

```python
df1
```

```python
df1.Year=pd.to_numeric(df1.Year)
df1
```

```python
df2=df1.loc[df1['Year']>=1985,:]
df2
```

```python
list3=[df2.groupby('Year').\
pipe(list)[i][1].iloc[:,[1,2,3]].\
to_numpy().tolist() for i in list(range(2017-1985))]
```

```python
m=folium.Map(scrollWheelZoom=False)
plugins.HeatMapWithTime(list3,radius=5,index=list(range(1985,2017))).add_to(m)
m
```
