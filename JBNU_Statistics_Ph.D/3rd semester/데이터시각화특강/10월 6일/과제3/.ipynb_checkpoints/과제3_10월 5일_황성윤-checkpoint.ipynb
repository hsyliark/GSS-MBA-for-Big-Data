## 과제3

`-` seaborn의 테마를 바꿔서 그래프 그려보기

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
np.random.seed(55364)
x=np.random.normal(size=1000,loc=3,scale=2)
```

```python
from scipy import stats 
np.random.seed(55364)
y=stats.t.rvs(5,size=1000)
```

```python
fig, ((ax1,ax2), (ax3,ax4)) = plt.subplots(2,2)
sns.set_theme(style="whitegrid", palette="pastel")
fig.set_figwidth(6)
fig.set_figheight(6)
```

```python
fig.suptitle('Homework3')
```

```python
xx= (x-np.mean(x)) / np.std(x,ddof=1) 
yy= (y-np.mean(y)) / np.std(y,ddof=1)
```

```python
ax1.boxplot(xx) 
ax1.set_title('boxplot1')
sns.histplot(xx,kde=True,ax=ax2)
ax2.set_title('histogram1')
ax3.boxplot(yy)
ax3.set_title('boxplot2')
sns.histplot(yy,kde=True,ax=ax4)
ax4.set_title('histogram2')
fig.tight_layout()
```
