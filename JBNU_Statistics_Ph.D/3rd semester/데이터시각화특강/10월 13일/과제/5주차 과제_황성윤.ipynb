## 과제

`-` 자유도가 5인 카이제곱분포에서 100개의 랜덤변수를 만들고, boxplot / histogram / qqplot 그리기

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
import pandas as pd 
```

```python
x=np.random.chisquare(df=5, size=100)
```

```python
fig, axs = plt.subplots(1,3)
axs
```

```python
ax1, ax2, ax3 = axs
```

```python
sns.histplot(x,kde=True,ax=ax1)
ax1.set_title('Histogram')
```

```python
ax2.boxplot(x)
ax2.set_title("Boxplot")
```

```python
_ = stats.probplot(x,plot=ax3)
ax3.set_title("Q-Q plot")
```

```python
fig.set_figwidth(15)
fig.set_figheight(7)
fig.tight_layout()
```
