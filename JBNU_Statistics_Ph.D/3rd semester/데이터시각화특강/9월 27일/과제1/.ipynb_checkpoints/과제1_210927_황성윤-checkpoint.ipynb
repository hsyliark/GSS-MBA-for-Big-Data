## 과제1

`-` 임의의 이미지를 cv.imread() 로 불러온뒤에 아래와 같이 blue+green의 조합으로 이미지를 변경해볼것

```python
import cv2 as cv
```

```python
view=cv.imread('hw1_HSY.jpg') 
```

```python
import matplotlib.pyplot as plt 
plt.imshow(view)
```

```python
view.shape
```

```python
import numpy as np
view_red=np.zeros_like(view)
view_green=np.zeros_like(view)
view_blue=np.zeros_like(view)
view_red[:,:,0]=view[:,:,0]
view_green[:,:,1]=view[:,:,1]
view_blue[:,:,2]=view[:,:,2]
```

```python
plt.imshow(view_blue+view_green)
```
