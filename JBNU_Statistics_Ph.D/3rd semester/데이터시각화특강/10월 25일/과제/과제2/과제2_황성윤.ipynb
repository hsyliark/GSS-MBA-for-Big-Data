## 과제2
`-` fifa22 데이터셋에서 `Q==Q4` 이고 `Potentail2>20` 인 선수들의 이름을 출력하라. 

```python
import numpy as np
import pandas as pd
```

```python
fifa22=pd.read_csv('https://raw.githubusercontent.com/guebin/2021DV/master/_notebooks/2021-10-25-FIFA22_official_data.csv')
fifa22
```

```python
fifa22['Potential2'] = fifa22['Potential'] - fifa22['Overall']
```

```python
def f(x): 
    if x>72: y='Q1' 
    elif x>68: y='Q2'
    elif x>63: y='Q3'
    else: y='Q4'
    return y 
```

```python
fifa22['Q']=list(map(f,fifa22.Overall))
fifa22[['Q','Overall']]
```

```python
l=fifa22.groupby(by='Q').mean().Overall.to_list()
def g(x): 
    if x=='Q1': y=l[0]
    elif x=='Q2': y=l[1]
    elif x=='Q3': y=l[2]
    else: y=l[3]
    return y
```

```python
fifa22['Qx']=list(map(g,fifa22.Q))
```

```python
fifa22.query(' Q=="Q4" and Potential2>20 ')
```
