### 3. 아래는 앤스콤의 플랏이다. 옳게 해석한 사람을 모두 고르라

````python
x = [10, 8, 13, 9, 11, 14, 6, 4, 12, 7, 5]
y1 = [8.04, 6.95, 7.58, 8.81, 8.33, 9.96, 7.24, 4.26, 10.84, 4.82, 5.68]
y2 = [9.14, 8.14, 8.74, 8.77, 9.26, 8.10, 6.13, 3.10, 9.13, 7.26, 4.74]
y3 = [7.46, 6.77, 12.74, 7.11, 7.81, 8.84, 6.08, 5.39, 8.15, 6.42, 5.73]
x4 = [8, 8, 8, 8, 8, 8, 8, 19, 8, 8, 8]
y4 = [6.58, 5.76, 7.71, 8.84, 8.47, 7.04, 5.25, 12.50, 5.56, 7.91, 6.89]
fig , ax = plt.subplots(2,2)
(ax1,ax2),(ax3,ax4) = ax
ax1.set_title('(a)')
ax2.set_title('(b)')
ax3.set_title('(c)')
ax4.set_title('(d)')
ax1.plot(x,y1,'o') 
ax2.plot(x,y2,'o') 
ax3.plot(x,y3,'o')  
ax4.plot(x4,y4,'o') 
fig.tight_layout()
````

- (하니) 그림 (a)는 양의 상관계수를 가진다. -> T

- (나애리) 그림 (b)는 산점도가 직선이 아니라 곡선의 모양을 띄고 있으므로 상관계수는 0이다. -> F

- (홍두깨) 그림 (c)에서 상단의 이상치를 제외하면 상관계수는 1이다. -> T

- (고은혜) 그림 (c)와 (d)의 상관계수를 각각 $r_{1},r_{2}$라고 하자. 그리고 (c)와 (d)에서 각각 이상치를 제외한뒤 구한 상관계수를 $\tilde{r_{1}},\tilde{r_{2}}$라고 하자. 그렇다면 $r_{1} r_{2} < \tilde{r_{1}} \tilde{r_{2}}$가 성립한다. -> F

### 4. 아래의 산점도를 보고 옳게 해석한 사람을 모두 고르라.

````python
x=np.arange(0,10,0.1)
y=x+np.random.normal(loc=0,scale=1,size=100)
plt.plot(x,y,'.b')
````

(하니) 두 변수는 양의 상관관계에 있어 보인다. -> T

(나애리) 완전한 직선 모양이 되지는 않았으므로 상관계수는 1보다 작다. -> T

(홍두깨) 산점도를 파악하면 $x$가 증가할때 $y$도 증가하므로 $x$가 $y$를 유발하는 원인이라 해석할 수 있다. -> T

(고은혜) $y$를 유지한채 $x$의 값을 모두 2배로 하면 상관계수의 값도 2배가 된다. -> F

#### (b) 여성의 기본급여 평균과 남성의 기본급여 평균을 각각 구하여라. 즉 `GENDER=Male`인 행들의 `BASE_SALARY`의 평균과 `GENDER=Female`인 행들의 `BASE_SALARY`의 평균을 비교하라.

````python
df.groupby(by='GENDER').agg({'BASE_SALARY':np.mean})
````

- 남성의 기본급여 평균이 여성의 기본급여 평균보다 약 5195.37 정도 더 크다. 

#### (c) 인종과 성별로 자료를 그룹핑하고 기본급여의 평균값 최소값 최대값을 각각 구한뒤 `barplot`으로 시각화 하라.

````python
td=df.groupby(by=['RACE','GENDER']).agg({'BASE_SALARY':['mean','min','max']})
td
````

````python
from plotnine import * 
````

````python
td=td.stack().reset_index().rename(columns={'level_2':'stat'})
td
````

````python
ggplot(td)\
+geom_bar(aes(x='stat',y='BASE_SALARY',fill='GENDER'),stat='identity',position='dodge')\
+coord_flip()+facet_wrap('RACE')
````
