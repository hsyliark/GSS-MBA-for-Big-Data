#### (d) (c)의 결과에 아래의 코드를 활용하여 `Wage`의 값을 적절하게 변환하라. 

````python
def convert_currency(value):
    floatvalue = 0.0
    strvalue=""
    if "M" in value:
        strvalue=value.replace("M","").replace("€","")
        floatvalue=float(float(strvalue)*1000000)
    elif "K" in value:
        strvalue=value.replace("K","").replace("€","")
        floatvalue=float(float(strvalue)*1000)
    else:
        floatvalue=value.replace("€","")
    return floatvalue
````

````python
df2['Wage']=list(map(convert_currency, df2.Wage))
````

````python
td=_df.groupby(by='Best Position').agg({'Value':np.mean}).\
rename(columns={'Value':'mean(Value)'}).reset_index()
td
````

````python
td=td.sort_values('mean(Value)',ascending=False).reset_index(drop=True)
td['Highlight']=td['mean(Value)']>=td['mean(Value)'][2]
td
````

````python
from plotnine import *  
````

````python
ggplot(td)\
+geom_bar(aes(x='Best Position',y='mean(Value)',colour='Highlight',fill='Highlight'),stat='identity',position='dodge')
````
